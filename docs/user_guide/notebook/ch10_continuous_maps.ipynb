# Bölüm 10 — Sürekli Fonksiyonlar ve Homeomorfizmalar

Topolojik uzaylar arasındaki fonksiyonlarda süreklilik, açık kümelerin
geri çekiminin (preimage) açık olması koşuluna dayanır.

## 1. Konu

$f: (X, \tau_X) \to (Y, \tau_Y)$ **sürekli** ise:
$$\forall V \in \tau_Y:\; f^{-1}(V) \in \tau_X.$$

| Tür | Koşul |
|-----|-------|
| Sabit | $f(x)=c$; her zaman sürekli |
| Özdeşlik | $f(x)=x$; her zaman sürekli |
| Bileşke | $f$, $g$ sürekli $\Rightarrow g\circ f$ sürekli |
| Homeomorfizma | Bijektif; $f$ ve $f^{-1}$ sürekli |

> **Neden bu konu?** Süreklilik topolojinin merkezi kavramıdır; homeomorfizma ise uzaylar arasındaki "topolojik eşdeğerlik" ilişkisini kurar.

> 🔍 **Kendin dene:** Sierpiński uzayından kendisine kaç sürekli fonksiyon var? Hepsini bulun.

> ⚠️ **Sık hata:** `s_topo = list(s.topology)` yeterlidir; `[set(u) for u in s.topology]` gereksiz dönüşüm yapar.

> ↗️ **Bkz.:** Bölüm 7 (kompakt uzayın görüntüsü kompakttır), Bölüm 8 (bağlantılı uzayın görüntüsü bağlantılıdır).

> 💭 **Öz-yansıtma:** Homeomorfizma ile izomorfizma arasındaki fark nedir? Topolojide hangi yapı korunur?

## 2. Teoremler

**Teorem 2.1.** Her sabit fonksiyon süreklidir.

**Teorem 2.2.** $f: X \to Y$ ve $g: Y \to Z$ sürekli $\Rightarrow g\circ f$ süreklidir.

**Teorem 2.3.** Sürekli görüntü kompakt uzayın kompakttır.

**Teorem 2.4.** Sürekli görüntü bağlantılı uzayın bağlantılıdır.

**Teorem 2.5.** Kompakt $X$'ten Hausdorff $Y$'ye sürekli bijeksiyon $\Rightarrow$ homeomorfizma.

## 3. Algoritmalar

### is_continuous_finite_map — $O(|\tau_Y| \cdot |X|)$

    for each V in tau_Y:
        preimage <- {x in X : f(x) in V}
        if preimage not in tau_X: return False
    return True

## 4. pytop API

```python
from pytop import (
    sierpinski_space,
    discrete_topology,
    indiscrete_topology,
    make_set,
    empty_set,
    is_continuous_finite_map,
    finite_homeomorphism_result,
)
```

## 5. Örnekler

### Örnek 5.1 — Sabit Fonksiyon: Her Zaman Sürekli

```python
s = sierpinski_space()
s_pts = list(s.carrier)
s_topo = list(s.topology)
f_const = {0: 1, 1: 1}
print('=== Ornek 5.1: Sabit f(x)=1 ===')
print('continuous:', is_continuous_finite_map(s_pts, s_topo, s_pts, s_topo, f_const))
```

### Örnek 5.2 — Ayrık Topoloji

```python
d = discrete_topology(0, 1)
d_pts = list(d.carrier)
d_topo = list(d.topology)
f_id   = {0: 0, 1: 1}
f_swap = {0: 1, 1: 0}
print('=== Ornek 5.2: Ayrik Topoloji ===')
print('ozdeslik:', is_continuous_finite_map(d_pts, d_topo, d_pts, d_topo, f_id))
print('swap:    ', is_continuous_finite_map(d_pts, d_topo, d_pts, d_topo, f_swap))
```

### Örnek 5.3 — Sierpiński → Ayrık

```python
print('=== Ornek 5.3: Sierpinski -> Ayrik ===')
print('id: S->D:', is_continuous_finite_map(s_pts, s_topo, d_pts, d_topo, f_id))
print('id: D->S:', is_continuous_finite_map(d_pts, d_topo, s_pts, s_topo, f_id))
```

### Örnek 5.4 — İndiscrete'ten Süreklilik

```python
ind = indiscrete_topology(0, 1)
ind_pts = list(ind.carrier)
ind_topo = list(ind.topology)
f_const0 = {0: 0, 1: 0}
print('=== Ornek 5.4: Indiscrete -> Sierpinski ===')
print('id (0->0,1->1):', is_continuous_finite_map(ind_pts, ind_topo, s_pts, s_topo, f_id))
print('const_0 (tum->0):', is_continuous_finite_map(ind_pts, ind_topo, s_pts, s_topo, f_const0))
```

### Örnek 5.5 — Homeomorfizma

```python
d2a = discrete_topology(1, 2)
d2b = discrete_topology('a', 'b')
ind2 = indiscrete_topology(1, 2)
print('=== Ornek 5.5: Homeomorfizma ===')
print('D(1,2) ~ D(a,b):', finite_homeomorphism_result(d2a, d2b).status)
print('D(1,2) ~ S(0,1):', finite_homeomorphism_result(d2a, s).status)
print('D(1,2) ~ Ind(1,2):', finite_homeomorphism_result(d2a, ind2).status)
```

### Örnek 5.6 — Özel Topolojili Süreklilik

```python
X = [1, 2, 3]
tau_X = [empty_set(), make_set(1), make_set(2, 3), make_set(1, 2, 3)]
f_23swap = {1: 1, 2: 3, 3: 2}
print('=== Ornek 5.6: Ozel Topoloji ===')
print('2-3 swap surekli mi:', is_continuous_finite_map(X, tau_X, X, tau_X, f_23swap))
```

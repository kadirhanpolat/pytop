# Bölüm 2 — Yüklemler ve Altküme Operatörleri

Bu bölümde bir topolojik uzaydaki altkümeler üzerinde tanımlı temel yüklemler
(açık, kapalı, clopen, yoğun, hiçbiryerde-yoğun) ve altküme operatörleri
(kapanış, iç, sınır, türev kümesi) incelenmektedir.

## 1. Konu

### 1.1 Altküme Yüklemleri

$(X, \tau)$ bir topolojik uzay ve $A \subseteq X$ olsun:

| Yüklem | Tanım |
|--------|-------|
| **Açık** | $A \in \tau$ |
| **Kapalı** | $X \setminus A \in \tau$ |
| **Clopen** | $A \in \tau$ ve $X \setminus A \in \tau$ |
| **Yoğun** | Her boş olmayan $U \in \tau$ için $U \cap A \neq \emptyset$ |
| **Hiçbiryerde-yoğun** | $\mathrm{int}(\overline{A}) = \emptyset$ |

### 1.2 Altküme Operatörleri

| Operatör | Gösterim | Tanım |
|----------|----------|-------|
| Kapanış | $\overline{A}$ | $A$'yı içeren en küçük kapalı küme |
| İç | $A^\circ$ | $A$'ya dahil en büyük açık küme |
| Sınır | $\partial A$ | $\overline{A} \setminus A^\circ$ |
| Türev kümesi | $A'$ | Birikme noktaları kümesi |

**Birikme noktası:** $x \in A'$ $\iff$ her $U \ni x$ açık için
$U \cap (A \setminus \{x\}) \neq \emptyset$.

### 1.3 Komşuluk Sistemi N(x)

$$N(x) = \{N \subseteq X : \exists U \in \tau,\, x \in U \subseteq N\}$$

**Aksiyomlar (N1–N4):** (N1) $x \in N$; (N2) $X \in N(x)$;
(N3) sonlu kesişim; (N4) üst-küme kapalı.

## 2. Teoremler

**Teorem 2.1 (Kuratowski Kapanış Aksiyomları).**
$\mathrm{cl}: \mathcal{P}(X) \to \mathcal{P}(X)$ şu dört özelliği sağlar:
(K1) $\mathrm{cl}(\emptyset)=\emptyset$; (K2) $A \subseteq \mathrm{cl}(A)$;
(K3) $\mathrm{cl}(\mathrm{cl}(A))=\mathrm{cl}(A)$;
(K4) $\mathrm{cl}(A \cup B)=\mathrm{cl}(A) \cup \mathrm{cl}(B)$.
Bu dört özelliği sağlayan her operatör bir topoloji tanımlar.

---

**Teorem 2.2 (İç-Kapanış Dualitesi).**
$$A^\circ = X \setminus \mathrm{cl}(X \setminus A), \qquad
\mathrm{cl}(A) = X \setminus \mathrm{int}(X \setminus A).$$

---

**Teorem 2.3 (Komşuluk Sistemi Round-Trip).**
N1–N4'ü sağlayan her $\{N(x)\}_{x \in X}$ ailesi,
$\tau = \{U : \forall x \in U, U \in N(x)\}$ şeklinde tek bir topolojiyi geri üretir.

## 3. Algoritmalar

### Sonlu Kapanış Hesabı — $O(|\tau| \cdot |X|)$

```
Kapanis(A, X, τ):
    closed ← {X ∖ U : U ∈ τ}
    result ← X
    for each C ∈ closed:
        if A ⊆ C and |C| < |result|: result ← C
    return result
```

### Türev Kümesi Hesabı — $O(|X| \cdot |\tau|)$

```
TurevKumesi(A, X, τ):
    acc ← ∅
    for each x ∈ X:
        if every open U ∋ x meets A ∖ {x}: acc.add(x)
    return acc
```

## 4. pytop API

| Fonksiyon | İmza | Açıklama |

## 5. Örnekler

````python
from pytop import (
    sierpinski_space,
    discrete_topology,
    make_topology,
    analyze_predicate,
    closure_of_subset,
    interior_of_subset,
    boundary_of_subset,
    derived_set_of_subset,
    is_nowhere_dense_subset,
    neighborhood_system,
    character_at_point,
    analyze_neighborhood_system,
)
````

### Örnek 5.1 — Kapanış ve İç: Sierpiński

````python
s = sierpinski_space()
print("cl({0})  =", closure_of_subset(s, {0}).value)
print("cl({1})  =", closure_of_subset(s, {1}).value)
print("int({0}) =", interior_of_subset(s, {0}).value)
print("int({1}) =", interior_of_subset(s, {1}).value)
print("bd({1})  =", boundary_of_subset(s, {1}).value)
````

**Çıktı Açıklaması:** $\{0\} = X \setminus \{1\}$ zaten kapalıdır → $\mathrm{cl}(\{0\}) = \{0\}$.
$\{1\}$ kapalı değil → kapanışı $X$. $\{0\}$'ı kapsayan açık küme yok → $\mathrm{int}(\{0\}) = \emptyset$.

### Örnek 5.2 — Türev Kümesi

````python
print("d({0}) =", derived_set_of_subset(s, {0}).value)
print("d({1}) =", derived_set_of_subset(s, {1}).value)
print("{0} nowhere_dense?", is_nowhere_dense_subset(s, {0}).status)
print("{1} nowhere_dense?", is_nowhere_dense_subset(s, {1}).status)
````

**Çıktı Açıklaması:** $d(\{1\}) = \{0\}$: $0$'ı içeren $\{0,1\}$,
$\{1\} \setminus \{0\} = \{1\} \neq \emptyset$ ile kesişir → $0$ birikme noktası.
$\{0\}$ hiçbiryerde-yoğun çünkü $\mathrm{int}(\mathrm{cl}(\{0\})) = \mathrm{int}(\{0\}) = \emptyset$.

### Örnek 5.3 — Yüklemler: Ayrık Topoloji

````python
d = discrete_topology('a', 'b', 'c')
print("{'a'} açık?          =", analyze_predicate(d, 'open', {'a'}).status)
print("{'a'} kapalı?        =", analyze_predicate(d, 'closed', {'a'}).status)
print("{'a'} clopen?        =", analyze_predicate(d, 'clopen', {'a'}).status)
print("{'a','b','c'} yoğun? =", analyze_predicate(d, 'dense', {'a','b','c'}).status)
````

**Çıktı Açıklaması:** Ayrık topolojide her tekil küme clopen. $X$ trivially yoğun.

### Örnek 5.4 — Sınır Hesabı

````python
sp = make_topology({1, 2, 3}, {1}, {2, 3})
print("bd({1})   =", boundary_of_subset(sp, {1}).value)
print("bd({2,3}) =", boundary_of_subset(sp, {2, 3}).value)
print("bd({1,2}) =", boundary_of_subset(sp, {1, 2}).value)
````

**Çıktı Açıklaması:** Açık kümeler $\{1\}$ ve $\{2,3\}$'ün sınırı boştur.
$\partial\{1,2\} = \{2,3\}$: $\{1,2\}$ ne açık ne kapalı.

### Örnek 5.5 — Komşuluk Sistemi

````python
carrier = list(s.carrier)
topology = list(s.topology)
print("N(0) =", neighborhood_system(carrier, topology, 0).value)
print("N(1) =", neighborhood_system(carrier, topology, 1).value)
print("chi(0) =", character_at_point(carrier, topology, 0).value)
print("chi(1) =", character_at_point(carrier, topology, 1).value)
````

**Çıktı Açıklaması:** $N(0) = \{X\}$ — $0$ yalnızca $X$'te. $N(1) = \{\{1\}, X\}$.
$\chi(0)=1$, $\chi(1)=2$ yerel baz büyüklükleri.

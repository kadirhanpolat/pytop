# Bölüm 7 — Kompaktlık

Kompaktlık, sonsuz yapılarda "sınırlılık" fikrinin topolojik soyutlamasıdır.
Açık örtü tanımı temel alınarak incelenmekte; varyantlar ve lokal kompaktlık da ele alınmaktadır.

## 1. Konu

### Kompaktlık Tanımı

$(X, \tau)$ topolojik uzayı **kompakt** ise:
Her açık örtü $\{U_\alpha\}$ için sonlu alt-örtü vardır:
$$X \subseteq \bigcup_\alpha U_\alpha \implies \exists \text{ sonlu } I:\; X \subseteq \bigcup_{\alpha \in I} U_\alpha.$$

### Kompaktlık Varyantları

| Varyant | Tanım |
|---------|-------|
| **Kompakt** | Her açık örtünün sonlu alt-örtüsü |
| **Sayılabilir kompakt** | Her sayılabilir örtünün sonlu alt-örtüsü |
| **Ardışımlı kompakt** | Her dizi yakınsak alt-dizi içerir |
| **Sözde kompakt** | Her sürekli $f: X \to \mathbb{R}$ sınırlıdır |
| **Lindelöf** | Her açık örtünün sayılabilir alt-örtüsü |
| **$\sigma$-kompakt** | Sayılabilir kompakt kümelerin birleşimi |
| **Lokal kompakt** | Her noktanın kompakt komşuluğu var |

**Sıralama:** Kompakt ⟹ Sayılabilir kompakt ⟹ Ardışımlı kompakt ⟹ Lindelöf.

> **Neden bu konu?** Kompaktlık Heine-Borel ve Tychonoff teoremlerinin kaynağıdır; sonlu alt-örtü özelliği analizde temel araçtır.

> 🔍 **Kendin dene:** `real_line_metric()` ve `closed_unit_interval_metric()` için `is_compact` sonuçlarını karşılaştırın.

> ⚠️ **Sık hata:** Sonlu uzay her zaman kompakttır; `finite_chain_space` için `is_compact` daima True döner.

> ↗️ **Bkz.:** Bölüm 6 (kompakt + Hausdorff ⟹ normal), Bölüm 10 (kompakt uzayın sürekli görüntüsü kompakttır).

> 💭 **Öz-yansıtma:** ℝ kompakt değilken [0,1] kompakt — bu farkın temelinde ne var?

## 2. Teoremler

**Teorem 2.1 (Heine-Borel).** $\mathbb{R}^n$'de kompakt $\Leftrightarrow$ kapalı ve sınırlı.

**Teorem 2.2.** Kompakt + Hausdorff $\Rightarrow$ Normal (T4).

**Teorem 2.3.** Kompakt uzaydan Hausdorff uzaya sürekli bijeksiyon $\Rightarrow$ homeomorfizma.

**Teorem 2.4 (Tychonoff).** Keyfi kompakt uzayların kartezyen çarpımı kompakttır.

**Teorem 2.5 (Alexandroff).** Her lokal kompakt Hausdorff uzay $X$ için
$X^* = X \cup \{\infty\}$ kompakt Hausdorff'tur.

## 3. Algoritmalar

### Sonlu Uzayda Kompaktlık

Sonlu uzay her zaman kompakttır: her örtü zaten sonlu. Algoritma: $O(1)$.

### analyze_compactness_variants — Tag Tabanlı Çıkarım

    AnalyzeVariants(X, tau):
        if X sonlu: tüm varyantlar <- True
        if 'compact' in tags:
            countably_compact, sequentially_compact, lindelof <- True

## 4. pytop API

```python
from pytop import (
    sierpinski_space,
    discrete_topology,
    finite_chain_space,
    real_line_metric,
    closed_unit_interval_metric,
    is_compact,
    is_lindelof,
    is_locally_compact,
)
from pytop.compactness_variants import (
    is_countably_compact,
    is_sequentially_compact,
    analyze_compactness_variants,
)
```

## 5. Örnekler

### Örnek 5.1 — Sonlu Uzaylar: Otomatik Kompakt

```python
s = sierpinski_space()
d = discrete_topology(1, 2, 3)
c = finite_chain_space(3)

print("=== Ornek 5.1: Sonlu Uzaylar ===")
print("Sierpinski compact?     ", is_compact(s).status)
print("Chain(3) compact?       ", is_compact(c).status)
print("Discrete(3) compact?    ", is_compact(d).status)
```

### Örnek 5.2 — Gerçek Doğru ℝ: Kompakt Değil

```python
rl = real_line_metric()
print("=== Ornek 5.2: Gercek Dogru ===")
print("compact?         ", is_compact(rl).status)
print("lindelof?        ", is_lindelof(rl).status)
print("locally_compact? ", is_locally_compact(rl).status)
```

### Örnek 5.3 — Kapalı [0,1]: Kompakt

```python
ui = closed_unit_interval_metric()
print("=== Ornek 5.3: [0,1] ===")
print("compact?     ", is_compact(ui).status)
print("sequentially?", is_sequentially_compact(ui).status)
print("countably?   ", is_countably_compact(ui).status)
```

### Örnek 5.4 — analyze_compactness_variants: Gerçek Doğru

```python
print("=== Ornek 5.4: analyze_compactness_variants(R) ===")
r = analyze_compactness_variants(rl)
for key, val in r.value.items():
    status = val.status if hasattr(val, 'status') else val
    print(f"  {key:<25}: {status}")
```

### Örnek 5.5 — analyze_compactness_variants: Sierpiński

```python
print("=== Ornek 5.5: analyze_compactness_variants(Sierpinski) ===")
r2 = analyze_compactness_variants(s)
for key, val in r2.value.items():
    status = val.status if hasattr(val, 'status') else val
    print(f"  {key:<25}: {status}")
```

# Bölüm 14 — Metrik Uzaylar

Metrik uzaylar, mesafe fonksiyonuna dayanan topolojik yapılardır. Her metrik uzay
doğal olarak bir topolojik uzay oluşturur.

## 1. Konu

### Metrik Aksiyomları

- **(M1) Özdeşlik:** $d(x,y) = 0 \Leftrightarrow x = y$
- **(M2) Simetri:** $d(x,y) = d(y,x)$
- **(M3) Üçgen:** $d(x,z) \leq d(x,y) + d(y,z)$

| Metrik | Tanım |
|--------|-------|
| Öklid | $d(x,y) = |x-y|$ |
| Ayrık | $d(x,y) = 0$ veya $1$ |
| Sınırlı | $d'(x,y) = \min(d(x,y),1)$ |

> **Neden bu konu?** Metrik uzaylar topolojik uzayların en somut alt sınıfıdır; ℝⁿ ve Hilbert uzayları bu çerçevede yaşar.

> 🔍 **Kendin dene:** `real_line_metric()` için `is_t2` ve `is_metrizable` sonuçlarını kontrol edin.

> ⚠️ **Sık hata:** Her metrik uzay T4 (normal Hausdorff) özelliğini taşır; `is_t2 False` dönerse metrik tanımında hata vardır.

> ↗️ **Bkz.:** Bölüm 9 (metrik uzay ⟹ 2. sayılabilir ↔ ayrılabilir).

> 💭 **Öz-yansıtma:** Metrik topoloji ne zaman ayrık topolojiyle çakışır? Hangi metrik bunu garanti eder?

## 2. Teoremler

**Teorem 2.1.** Her metrik uzay Hausdorff (T2) ve $1^{\text{st}}$ sayılabilirdir.

**Teorem 2.2.** $d'(x,y) = \min(d(x,y),1)$ ile $d$ aynı topolojiyi indükler.

## 3. Algoritmalar

### validate_metric — $O(|X|^3)$

    ValidateMetric(X, d):
        M1: d(x,x)=0; d(x,y)>0 for x≠y
        M2: d(x,y)=d(y,x)
        M3: d(x,z) ≤ d(x,y)+d(y,z)

### open_ball — $O(|X|)$

    OpenBall(x, r, X, d): return {y: d(x,y) < r}

## 4. pytop API

```python
from pytop.metric_spaces import (
    FiniteMetricSpace,
    validate_metric,
)
from pytop import (
    open_ball,
    closed_ball,
    distance_to_subset,
    diameter_of_subset,
    is_bounded_subset,
    capped_metric,
)
```

## 5. Örnekler

### Örnek 5.1 — Ayrık Metrik Uzay

```python
points = [1, 2, 3, 4]
dist_discrete = {(a, b): (0 if a == b else 1) for a in points for b in points}
fms = FiniteMetricSpace(carrier=tuple(points), distance=dist_discrete)
print('=== Ornek 5.1: Ayrik Metrik ===')
print('carrier:', fms.carrier)
print('d(1,2):', fms.distance[(1,2)])
print('d(1,1):', fms.distance[(1,1)])
```

### Örnek 5.2 — validate_metric

```python
r_valid = validate_metric(fms)
print('=== Ornek 5.2: validate_metric ===')
print('status:', r_valid.status)
print('justification:', r_valid.justification[0] if r_valid.justification else '')
```

### Örnek 5.3 — open_ball

```python
print('=== Ornek 5.3: Toplar ===')
print('B(1, 0.5) =', open_ball(fms, 1, 0.5))
print('B(1, 1.5) =', open_ball(fms, 1, 1.5))
print('Bclosed(1, 1.0) =', closed_ball(fms, 1, 1.0))
```

### Örnek 5.4 — diameter ve distance_to_subset

```python
print('=== Ornek 5.4: Metrik Fonksiyonlar ===')
print('diam({1,2,3}):', diameter_of_subset(fms, {1, 2, 3}))
print('d(1, {2,3}):', distance_to_subset(fms, 1, {2, 3}))
print('is_bounded({1,2}):', is_bounded_subset(fms, {1, 2}))
```

### Örnek 5.5 — Öklid Metriki

```python
points_line = [0, 1, 2, 3]
dist_eucl = {(a, b): abs(a - b) for a in points_line for b in points_line}
fms_line = FiniteMetricSpace(carrier=tuple(points_line), distance=dist_eucl)
print('=== Ornek 5.5: Oklid Metrigi ===')
print('d(0,3):', fms_line.distance[(0,3)])
print('B(1, 1.5) =', open_ball(fms_line, 1, 1.5))
print('diam({0,1,2,3}):', diameter_of_subset(fms_line, {0,1,2,3}))
```

### Örnek 5.6 — capped_metric

```python
print('=== Ornek 5.6: capped_metric ===')
cap_fn = capped_metric(fms_line.distance, cap=1.0)
print('d(0,3) capped:', cap_fn(0, 3))
print('d(0,1) capped:', cap_fn(0, 1))
```

## 1. Konu

### 1.1 Sonsuz Yapıların Yüklemle Temsili

`frozenset` gibi yapılar yalnızca **sonlu** kümeleri temsil edebilir.  
Gerçek matematikte ise $\mathbb{N}, \mathbb{Z}, \mathbb{R}$ gibi sonsuz kümeler temel nesne birimidir.

**Yüklem tabanlı temsil:**
$$A = \{x \in \mathbb{N} : P(x)\}$$
Burada $A$, $P$ yüklemini sağlayan tüm doğal sayıları tanımlar.  
$A$ sonsuz olabilir; üyelik kontrolü $P(x)$'i hesaplamak kadar basittir.

### 1.2 Temel Sabitler

| Sabit | Tanım | Python sembolü |
|-------|-------|---------------|
| $\mathbb{N}$ | Doğal sayılar $\{0, 1, 2, \ldots\}$ | `N` |
| $\mathbb{Z}$ | Tam sayılar $\{\ldots, -1, 0, 1, \ldots\}$ | `Z` |
| $\mathbb{Q}$ | Rasyonel sayılar | `Q` |
| $\mathbb{R}$ | Gerçel sayılar | `R` |
| $\mathbb{C}$ | Karmaşık sayılar | `C` |
| $\Sigma^*$ | Sonlu karakter dizileri | `Sigma` |
| $\mathbb{N}_{>0}$ | Pozitif doğal sayılar | `N_plus` |
| $\mathbb{R}_{>0}$ | Pozitif gerçeller | `R_plus` |

### 1.3 `MathRelation` ve `MathFunction`

Aynı ilke ilişkiler ve fonksiyonlara da uygulanır:
- **`MathRelation`**: $R \subseteq A \times B$, yüklem $(x, y) \mapsto $ `bool`
- **`MathFunction`**: $f: A \to B$, kural $x \mapsto f(x)$

Her ikisi de sonsuz domainler üzerinde çalışır;  
`restrict_to` ile sonlu kesitler somutlaştırılır.

## 2. Teoremler

### Teorem 1 — Set Comprehension Yüklem Bileşimi

**İfade:**  
$A = \{x \in S : P(x)\}$ ve $B = \{x \in A : Q(x)\}$ için $B = \{x \in S : P(x) \wedge Q(x)\}$.

**İspat:**  
$x \in B \iff x \in A$ ve $Q(x) \iff P(x)$ ve $Q(x)$. $\square$

`MathSet.where` tam olarak bu bileşimi gerçekleştirir.

---

### Teorem 2 — Fonksiyon Bileşimi Yasaları

**İfade:**  
$f: A \to B$, $g: B \to C$ için $(g \circ f): A \to C$, $(g \circ f)(x) = g(f(x))$. Ayrıca:
- $f$ ve $g$ enjektif $\Rightarrow$ $g \circ f$ enjektif
- $f$ ve $g$ surjektif $\Rightarrow$ $g \circ f$ surjektif
- $g \circ f$ enjektif $\Rightarrow$ $f$ enjektif
- $g \circ f$ surjektif $\Rightarrow$ $g$ surjektif

**İspat (birincisi):**  
$f(x_1) = f(x_2) \Rightarrow x_1 = x_2$ (f enjektif).  
$(g \circ f)(x_1) = (g \circ f)(x_2) \Rightarrow g(f(x_1)) = g(f(x_2)) \Rightarrow f(x_1) = f(x_2)$ (g enjektif) $\Rightarrow x_1 = x_2$. $\square$

---

### Teorem 3 — Ters İlişki Simetri Kuralı

**İfade:**  
$R: A \to B$ yüklem ilişkisi, $R^{-1}: B \to A$ tersi için:
$(x, y) \in R \iff (y, x) \in R^{-1}$.

Ayrıca $R$ yansımalıysa (aynı domain/codomain): $(R^{-1})^{-1} = R$. $\square$

## 3. API

```python
from pytop.predicate_sets import (
    set_of,
    N, Z, Q, R, C, Sigma,
    N_plus,
)
from pytop.predicate_relations import (
    leq, divides,
)
from pytop.predicate_functions import (
    function_from,
    successor, square, double, abs_value,
)
print("pytop yüklendi.")
```

### `MathSet` API'si

| Yöntem/Fonksiyon | Anlamı |
|------------------|--------|
| `x in A` | $x \in A$ üyelik kontrolü |
| `A.contains(x)` | `x in A` ile eşdeğer |
| `A.where(P)` | $\{x \in A : P(x)\}$ — alt küme |
| `A & B` | $A \cap B$ |
| `A \| B` | $A \cup B$ |
| `A.complement_in(U)` | $U \setminus A$ |
| `A.sample(n)` | $n$ temsilci eleman |
| `A.to_frozenset(sample)` | `sample`'ı filtrele ve dondur |
| `set_of(A, P, name)` | `A.where(P, name)` ile eşdeğer |

### Önceden Tanımlı Sabitler

| Sabit | Tür | Sampler var mı? |
|-------|-----|----------------|
| `N`, `natural_numbers` | `MathSet` | evet (0,1,2,…) |
| `Z`, `integers` | `MathSet` | evet (0,1,-1,2,…) |
| `Q`, `rationals` | `MathSet` | evet |
| `R`, `reals` | `MathSet` | evet |
| `leq` (≤) | `MathRelation` | — |
| `divides` (∣) | `MathRelation` | — |
| `square` (x²) | `MathFunction` | — |
| `successor` (x+1) | `MathFunction` | — |

## 4. Örnekler

### Örnek 1 — Minimal: Sonsuz Küme Üyeliği ve Örnekleme

```python
# Temel üyelik kontrolleri
print(f"5 ∈ ℕ      → {5 in N}")
print(f"-3 ∈ ℕ     → {-3 in N}")
print(f"-3 ∈ ℤ     → {-3 in Z}")
print(f"0.5 ∈ ℚ    → {0.5 in Q}")
print(f"3.14 ∈ ℝ   → {3.14 in R}")
print(f"1+2j ∈ ℂ   → {complex(1,2) in C}")
print(f"'abc' ∈ Σ* → {'abc' in Sigma}")

# Örnekleme
print(f"\nℕ'den 8 örnek: {N.sample(8)}")
print(f"ℤ'den 8 örnek: {Z.sample(8)}")
```

### Örnek 2 — Orta Düzey: Set Comprehension ve Küme İşlemleri

```python
# Çift doğal sayılar: {n ∈ ℕ : n mod 2 = 0}
evens  = set_of(N, lambda n: n % 2 == 0, name="Çift ℕ")
odds   = set_of(N, lambda n: n % 2 == 1, name="Tek ℕ")
# Asal sayılar: trial division √n'ye kadar (O(√n) per query)
primes = set_of(N_plus, lambda n: n > 1 and all(n % i != 0 for i in range(2, int(n**0.5) + 1)), name="Asallar")

print(f"4 ∈ Çift ℕ  → {4 in evens}")
print(f"7 ∈ Tek ℕ   → {7 in odds}")
print(f"11 asaldır  → {11 in primes}")
print(f"15 asaldır  → {15 in primes}")

# İlk 15 eleman içinden somutlaştır
evens_concrete = evens.to_frozenset(range(15))
print(f"\nÇift ℕ ∩ [0,14]: {sorted(evens_concrete)}")

# Küme bileşimi: küçük asal sayılar ({n ∈ asal : n < 20})
small_primes = primes.where(lambda n: n < 20, name="Küçük asallar")
sp_concrete = small_primes.to_frozenset(range(20))
print(f"20'den küçük asallar: {sorted(sp_concrete)}")
```

### Örnek 3 — Gelişmiş: Yüklem İlişkileri

```python
# Önceden tanımlı ≤ ve | ilişkileri
print("≤ ilişkisi:")
print(f"  leq.holds(2, 5) = {leq.holds(2, 5)}")
print(f"  leq.holds(7, 3) = {leq.holds(7, 3)}")

print("\n| (bölünebilirlik) ilişkisi:")
print(f"  divides.holds(3, 12) = {divides.holds(3, 12)}")
print(f"  divides.holds(5, 12) = {divides.holds(5, 12)}")

# {1,...,6} kümesine kısıt
small = list(range(1, 7))
div_restricted = divides.restrict_to(small)
print(f"\n{'{1..6}'} üzerinde | : {len(div_restricted)} çift")

# Özellik testleri
print(f"≤ yansımalı mı ({small})?   {leq.is_reflexive_on(small)}")
print(f"≤ simetrik mi?              {leq.is_symmetric_on(small)}")
print(f"≤ geçişli mi?               {leq.is_transitive_on(small)}")
print(f"| kısmi sıralama mı?        {divides.is_partial_order_on(small)}")

# Ters ilişki: ≤ ters → ≥
geq_derived = leq.inverse()
print(f"\n≥ (terslenmiş ≤): {geq_derived.name}")
print(f"  geq_derived.holds(5, 2) = {geq_derived.holds(5, 2)}")
```

### Örnek 4 — Gerçekçi Senaryo: Yüklem Fonksiyonları ve Bileşim

```python
# Önceden tanımlı fonksiyonlar
print(f"successor(4)  = {successor(4)}")
print(f"square(5)     = {square(5)}")
print(f"double(7)     = {double(7)}")
print(f"abs_value(-3) = {abs_value(-3)}")

# Sonlu kısıt
domain_sample = list(range(6))
print(f"\nkare: {square.restrict_to(domain_sample)}")
print(f"halef: {successor.restrict_to(domain_sample)}")

# Bileşim: double ∘ succ_z = 2(n+1)
# double: ℤ→ℤ, bu yüzden ara fonksiyon da ℤ→ℤ olmalı
succ_z = function_from(Z, Z, lambda x: x + 1, name="succ")
double_succ = double.compose(succ_z)
print(f"\ndouble ∘ succ:")
for x in range(5):
    print(f"  f({x}) = {double_succ(x)}   (beklenen: {2*(x+1)})")

# Özel fonksiyon: f(n) = n² + 1, ℕ → ℕ
sq_plus_1 = function_from(N, N, lambda n: n**2 + 1, name="n²+1")
print(f"\nn²+1 fonksiyonu:")
print(f"  {sq_plus_1.restrict_to(range(6))}")

# Enjektiflik testi
sample = list(range(10))
print(f"\nsquare enjektif mi (ℕ üzerinde)? {square.is_injective_on(sample)}")
print(f"  Neden? ℕ'de x²=y² → x=y (tüm elemanlar ≥ 0)")
print(f"abs_value enjektif mi (ℤ üzerinde)? {abs_value.is_injective_on(range(-5, 6))}")
print(f"  Çünkü: |-3| = |3| = 3")
```

$\{n \in \mathbb{N} : n^2 < 50\}$ kümesini `set_of` ile tanımla.

```python
sq_lt_50 = set_of(N, lambda n: n**2 < 50, name="n²<50")
large    = sq_lt_50.where(lambda n: n > 5, name="n²<50 ve n>5")

print("n² < 50 kümesi:", sorted(sq_lt_50.to_frozenset(range(15))))
print("Ve n > 5 olanlar:", sorted(large.to_frozenset(range(15))))
```

### Alıştırma 3 — Karşılaştırma

$g \circ f = \text{identity}$ ise $f$ enjektif ve $g$ surjektif midir?  
`double` ve `half` (eğer $n$ çiftse $n/2$, aksi halde tanımsız) bileşimini düşün.  
Kendi `half_fn` fonksiyonunu `function_from` ile oluştur ve bileşimlerini test et.

```python
# ℕ üzerinde çift alma ve bölme (domain eşleşmesi: her ikisi de ℕ→ℕ)
double_n = function_from(N, N, lambda x: 2 * x, name="2x")
half_fn  = function_from(N, N, lambda n: n // 2, name="n//2")

# half_fn ∘ double_n: ℕ → ℕ, n ↦ n (kimlik gibi)
half_of_double = half_fn.compose(double_n)
sample2 = list(range(6))
print("half(double(n)):")
for x in sample2:
    print(f"  n={x}: double={double_n(x)}, half(double)={half_of_double(x)}")

# double (ℤ → ℤ) enjektif mi?
print(f"\ndouble enjektif (0..5 üzerinde): {double.is_injective_on(sample2)}")
# double surjektif mi? (Tek sayılar görüntü dışında kalır)
cod_sample = list(range(12))
print(f"double surjektif (0..11 üzerinde hedef): {double.is_surjective_on(sample2, cod_sample)}")
```

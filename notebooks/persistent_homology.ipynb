## 1. Konu

### 1.1 Sezgisel Açıklama

Elinizdeki veri, gerçek dünyadaki bir şeklin gürültülü ölçümü olabilir. Mesela:

- Çevrensel gürültüyle birlikte bir çemberden örneklenmiş 100 nokta
- Yükseklik fonksiyonu tanımlı bir arazi yüzeyi
- Yüksek boyutlu gen ekspresyon verisi

**Kalıcı homoloji** bu verinin _şeklini_ birden fazla ölçekte aynı anda yakalar.

**Ana fikir — filtrasyon:**  
Veri kümesini yavaş yavaş "büyüyen" bir yapıyla temsil edin. Bir eşik parametresi $r$ arttıkça:
- $r = 0$: her nokta kendi başına (izole)
- $r$ büyüdükçe: yakın noktalar arasında kenarlar çıkar, üçgenler oluşur
- $r = \infty$: tam simpleks (her şey birbirine bağlı)

Her adımda topolojik özellikler **doğar** (birth) veya **ölür** (death):
- Bağlı bileşenler birleşir → bir bileşen "ölür"
- Döngüler oluşur → bir 1-döngü "doğar"
- Döngü dolar → "ölür"

Her özellik için $(b, d)$ çifti — doğum ve ölüm zamanı — elde edilir. Bu çiftlerin tamamı **kalıcılık diyagramını** oluşturur.

**Kısa yaşayan = gürültü. Uzun yaşayan = gerçek şekil.**

---

### 1.2 Formal Tanımlar

**Filtrasyon:**  
Simplicial komplekslerin iç içe geçmiş bir dizisi:
$$K_0 \subseteq K_1 \subseteq \cdots \subseteq K_n$$

**Kalıcı Homoloji:**  
Filtrasyonun $k$-inci kalıcı homoloji grubu, $i \le j$ için:
$$H_k^{i,j} = \operatorname{im}\bigl(H_k(K_i) \to H_k(K_j)\bigr)$$
$K_i$'de doğan ve $K_j$'de hâlâ yaşayan $k$-boyutlu homoloji sınıflarını sayar.

**Kalıcılık Çifti (Persistence Pair):**  
Her homoloji sınıfı için bir $(b, d)$ çifti; $b$ doğum zamanı, $d$ ölüm zamanı ($d = +\infty$ ise **esansiyel sınıf**).

**Kalıcılık Diyagramı $\operatorname{Dgm}_k$:**  
Tüm kalıcılık çiftlerinin çoklukalı (multiset) koleksiyonu. Noktalar $b < d \le +\infty$ düzleminde gösterilir.

**Vietoris-Rips Kompleksi:**  
Sonlu metrik uzay $(X, d)$ ve ölçek $r > 0$ için:
$$\operatorname{VR}(X, r) = \bigl\{\sigma \subseteq X : \operatorname{diam}(\sigma) \le r\bigr\}$$
Yani bir simpleks $\{x_0, \ldots, x_k\}$ dahil edilir ancak ve ancak tüm $d(x_i, x_j) \le r$ ise.

**Çech Kompleksi:**  
Ambient uzay $\mathbb{R}^n$'de $X$ nokta kümesi için:
$$\check{C}(X, r) = \bigl\{\sigma \subseteq X : \bigcap_{x \in \sigma} B(x, r) \ne \emptyset\bigr\}$$
Sinir teoremi (nerve theorem) gereği $|\check{C}(X,r)|$ ile $\bigcup_{x \in X} B(x,r)$ homotopi denktir.

## 2. Teoremler

### Teorem 1 — Yapı Teoremi (Zomorodian–Carlsson, 2005)

**İfade:**  
Bir $k$ cismi üzerindeki sonlu sunumlu her kalıcılık modülü
$$M = (M_i,\, \varphi_i\colon M_i \to M_{i+1})$$
benzersiz biçimde (izomorfizma'ya kadar) aralık modüllerinin doğrudan toplamına ayrışır:
$$M \cong \bigoplus_j k[b_j, d_j)$$
Burada $k[b, d)$ pozisyon $b \le i < d$'de $k$, diğer yerlerde $0$ olan modüldür.

**İspat:**

_1. Kodlama:_ $M$'i dereceli bir $k[t]$-modülüne dönüştür:
$$\hat{M} = \bigoplus_{i \ge 0} M_i$$
ile $t$ etkisi $m \in M_i \mapsto \varphi_i(m) \in M_{i+1}$ olarak tanımlanır (derecede 1 artış).

_2. $k[t]$ bir PID'dir:_ $k$ cismi olduğundan $k[t]$ bir asal ideallilik bölgesidir (Principal Ideal Domain).

_3. PID üzerinde modüllerin yapı teoremi:_ Sonlu üretilmiş $k[t]$-modülleri şu biçimdedir:
$$k[t]^r \oplus \bigoplus_j k[t]/(t^{n_j})$$
Serbest kısım $r$ adet sonsuz barla, burulma kısmı sonlu barlarla karşılık gelir. Dereceli versiyonda $k[t]/(t^{n_j})$ aralık modülü $k[b_j, b_j + n_j)$'ye eşit olur.

_4. Teklik:_ PID üzerindeki yapı teoremi tekliği garanti eder. Barkod $\{[b_j, d_j)\}$ tam izomorfizma değişmezi olur.

_5. Sonuç:_ Barkod, kalıcılık diyagramıyla aynı bilgiyi taşır. Her aralık $[b_j, d_j)$ diyagramda $(b_j, d_j)$ noktasına karşılık gelir. $\square$

---

### Teorem 2 — Kararlılık Teoremi (Cohen-Steiner, Edelsbrunner, Harer, 2007)

**İfade:**  
Kompakt bir $X$ üzerindeki iki "tame" (düzgün) fonksiyon $f, g\colon X \to \mathbb{R}$ için:
$$d_B\bigl(\operatorname{Dgm}(f),\, \operatorname{Dgm}(g)\bigr) \le \|f - g\|_\infty$$
Burada $d_B$ şişe boğazı uzaklığı (bottleneck distance):
$$d_B(D_1, D_2) = \inf_{\gamma} \sup_{p \in D_1} \|p - \gamma(p)\|_\infty$$

**İspat iskeleti:**

_1. İnterleaving:_ $\varepsilon = \|f - g\|_\infty$ ise $f \le g + \varepsilon$ ve $g \le f + \varepsilon$. Bu, sublevel set filtrasyonları arasında bir $\varepsilon$-interleaving verir:
$$f^{-1}(-\infty, t] \subseteq g^{-1}(-\infty, t+\varepsilon] \subseteq f^{-1}(-\infty, t+2\varepsilon]$$

_2. Interleaving → Bottleneck bağı:_ İki filtrasyon $\varepsilon$-interleaved ise kalıcılık modülleri de $\varepsilon$-interleaved olur. İsometry teoremi (Lesnick 2015) gereği modüllerin $\varepsilon$-interleaving'i diyagramların $\varepsilon$ bottleneck uzaklığını verir.

_3. Dolayısıyla:_  
$$d_B(\operatorname{Dgm}(f), \operatorname{Dgm}(g)) \le \varepsilon = \|f - g\|_\infty \quad \square$$

**Yorumu:** $\varepsilon$ gürültüyle bozulmuş bir fonksiyonun kalıcılık diyagramı, orijinal fonksiyonunkinden en fazla $\varepsilon$ farklıdır. Diyagrama yakın (diyagonsele yakın) noktalar gürültüye, uzak noktalar gerçek topolojik özelliklere karşılık gelir.

---

### Teorem 3 — VR–Čech Sandviç Lemması

**İfade:**  
$X \subseteq \mathbb{R}^n$ sonlu nokta kümesi, $r > 0$ için:
$$\operatorname{VR}(X, r) \subseteq \check{C}(X, r) \subseteq \operatorname{VR}(X, 2r)$$

**İspat:**

_VR ⊆ Čech:_ $\sigma \in \operatorname{VR}(X,r)$ ise tüm çiftler $d(x_i, x_j) \le r$. $\sigma$'nın küçük çevreleyen küresi (miniball) yarıçapı $\le r/2 \le r$ (Johnson-Lindenstrauss argümanı ile). Dolayısıyla toplantısal bal $\bigcap B(x_i, r) \ni$ merkez $\ne \emptyset$, yani $\sigma \in \check{C}(X,r)$.

_Čech ⊆ VR(2r):_ $\sigma \in \check{C}(X,r)$ ise ortak nokta $c \in \bigcap B(x_i, r)$ var. Üçgen eşitsizliğiyle $d(x_i, x_j) \le d(x_i, c) + d(c, x_j) \le r + r = 2r$. Dolayısıyla $\sigma \in \operatorname{VR}(X, 2r)$. $\square$

**Pratikte:** VR filtrasyonu tamamen kombinatoryeldir (sadece mesafe bilgisi gerekli), Čech ise geometri gerektirir. Sandviç lemması sayesinde VR, Čech'in 2 kat ölçek hatası ile geometrik doğru bir yaklaşımı olarak kullanılabilir.

## 3. API

### 3.1 Temel Import

````python
from pytop import (
    # Yüklem fonksiyonları (Result döner)
    has_finite_barcode,
    is_stable_filtration,
    has_essential_classes,
    has_structure_theorem,
    # Facade fonksiyonları (dict döner)
    classify_persistence,
    persistence_profile,
    # Tag sabitleri
    VIETORIS_RIPS_TAGS,
    CECH_COMPLEX_TAGS,
    STABLE_FILTRATION_TAGS,
    FIELD_COEFFICIENTS_TAGS,
    STRUCTURE_THEOREM_TAGS,
    ESSENTIAL_CLASS_TAGS,
    # Profil kaydı
    get_named_persistence_profiles,
    PersistenceProfile,
)
print("pytop yüklendi.")
````

### 3.2 Tag Sistemi

`pytop` fonksiyonları bir **uzay nesnesini** alır. Bu nesnenin `.tags` özelliği, o filtrasyon veya uzayın özelliklerini tanımlayan dize kümesidir.

Minimal bir uzay nesnesi yaratmak için yeterli olan yapı:
```python
class Space:
    def __init__(self, *tags):
        self.tags = set(tags)
```

**Kullanılabilir tag kümeleri:**

| Tag kümesi | Örnekler | Anlamı |
|-----------|---------|-------|
| `VIETORIS_RIPS_TAGS` | `"vietoris_rips"`, `"rips_filtration"` | VR filtrasyonu |
| `CECH_COMPLEX_TAGS` | `"cech_complex"`, `"alpha_complex"` | Čech/alpha filtrasyonu |
| `STABLE_FILTRATION_TAGS` | `"stability_theorem"`, `"sublevel_set_filtration"` | Kararlı filtrasyon |
| `FIELD_COEFFICIENTS_TAGS` | `"field_coefficients"`, `"z2_coefficients"` | Cisim katsayılı |
| `STRUCTURE_THEOREM_TAGS` | `"structure_theorem"`, `"interval_decomposition"` | Yapı teoremi geçerli |
| `ESSENTIAL_CLASS_TAGS` | `"essential_homology"`, `"infinite_bar"` | Esansiyel sınıf var |

### 3.3 `Result` Nesnesi

Yüklem fonksiyonları bir `Result` nesnesi döner:

```python
result = has_finite_barcode(space)
result.is_true      # bool — evet/doğru mu?
result.is_false     # bool — hayır/yanlış mı?
result.is_unknown   # bool — belirsiz mi?
result.status       # str — "true" | "false" | "unknown" | "conditional"
result.mode         # str — "theorem" | "symbolic" | ...
result.value        # str — kısa etiket (örn. "finite_barcode")
result.justification  # list[str] — gerekçeler
result.metadata     # dict — ek bilgi
```

## 4. Örnekler

### Örnek 1 — Minimal: Vietoris-Rips Filtrasyonu

````python
class Space:
    """Minimal uzay nesnesi — sadece tag kümesi."""
    def __init__(self, *tags):
        self.tags = set(tags)

# Bir VR filtrasyonu temsil edelim
rips = Space("vietoris_rips", "rips_filtration")

# Sonlu barkod var mı?
r1 = has_finite_barcode(rips)
print("Sonlu barkod:", r1.is_true)
print("Gerekçe:", r1.justification[0][:80], "...")

# Kararlı mı?
r2 = is_stable_filtration(rips)
print("\nKararlı:", r2.is_true)
print("Gerekçe:", r2.justification[0][:80], "...")
````

### Örnek 2 — Orta Düzey: Sublevel Set Filtrasyonu ile Esansiyel Sınıflar

````python
# Bir kapanık manifold üzerindeki Morse fonksiyonunun sublevel set filtrasyonu
manifold_sublevel = Space(
    "sublevel_set_filtration",
    "essential_homology",   # kapanık manifoldda esansiyel sınıf var
    "field_coefficients",
)

r = has_essential_classes(manifold_sublevel)
print("Esansiyel sınıf:", r.is_true)
print("Gerekçe:", r.justification[0])

# Yapı teoremi geçerli mi?
r_struct = has_structure_theorem(manifold_sublevel)
print("\nYapı teoremi:", r_struct.is_true)
print("Metadata:", r_struct.metadata.get("criterion"))
````

### Örnek 3 — Gelişmiş: Tam Sınıflandırma (classify_persistence)

````python
# S^1'den örneklenmiş nokta bulutu — hem kararlı hem esansiyel sınıf var
circle_cloud = Space(
    "vietoris_rips",
    "essential_homology",   # S^1'in H_1'i nontrivial
    "field_coefficients",
)

report = classify_persistence(circle_cloud)

print("Kalıcılık sınıfı:", report["persistence_class"])
print("Anahtar özellikler:", report["key_properties"])
print()
print("Sonlu barkod:", report["has_finite_barcode"].is_true)
print("Kararlı:", report["is_stable_filtration"].is_true)
print("Esansiyel sınıf:", report["has_essential_classes"].is_true)
print("Yapı teoremi:", report["has_structure_theorem"].is_true)
````

### Örnek 4 — Gerçekçi Senaryo: Mapper Algoritması (Kararlı Değil)

````python
# Mapper algoritması — filtrasyon DEĞİL, kalıcılık diyagramı yok
mapper = Space("mapper_graph", "cover_filtration")

stable_r = is_stable_filtration(mapper)
barcode_r = has_finite_barcode(mapper)

print("Kararlı (Mapper):", stable_r.is_true, "|", stable_r.is_false)
print("Gerekçe:", stable_r.justification[0][:90], "...")
print()

report_mapper = classify_persistence(mapper)
print("Mapper sınıfı:", report_mapper["persistence_class"])
print("Özellikler:", report_mapper["key_properties"])

# Karşılaştırma: VR vs Mapper
print()
rips2 = Space("vietoris_rips")
print("VR sınıfı:    ", classify_persistence(rips2)["persistence_class"])
print("Mapper sınıfı:", report_mapper["persistence_class"])
````

### Örnek 4b — Adlandırılmış Profiller Kaydı

````python
# Kütüphanedeki tüm hazır profiller
profiles = get_named_persistence_profiles()
print(f"{len(profiles)} adlandırılmış profil:")
print()
for p in profiles:
    kararlı = "✓ kararlı" if p.is_stable else "✗ kararlı değil"
    esansiyel = "✓ esansiyel" if p.has_essential_classes else ""
    print(f"  [{p.key}]")
    print(f"    {p.display_name}")
    print(f"    {kararlı}  {esansiyel}")
    print()
````

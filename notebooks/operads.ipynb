## 1. Konu

### 1.1 Operad Nedir?

Bir **simetrik operad** $P$, her $n \geq 0$ için $S_n$-modülü $P(n)$ (arity-$n$ işlemler uzayı) ile birlikte:
$$\gamma: P(k) \otimes P(n_1) \otimes \cdots \otimes P(n_k) \longrightarrow P(n_1 + \cdots + n_k)$$
bileşim dönüşümü, birim $\mathbf{1} \in P(1)$, birliktelik ve $S_n$-denkvariyans koşullarını sağlar.

**Sezgi:** $P(n)$'deki bir eleman, $n$ giriş ve 1 çıkış alan bir işlemi temsil eder. Bileşim, ağaçların grafik dili ile:

```
       [p]          [q1]  [q2]
      / | \    +    |      |     =    [p(q1,q2)]
     1  2  3       1      1          / | \ | |
```

### 1.2 Temel Operadlar

| Operad | $P(n)$ | Cebirler | Koszul mü? | Koszul dual |
|--------|--------|----------|------------|-------------|
| **Ass** | $k[S_n]$ (rank $n!$) | İlişkilendirilebilir cebirler | ✓ | Ass (twisted) |
| **Com** | $k$ (trivial, rank 1) | Değişmeli cebirler | ✓ | Lie |
| **Lie** | $(n-1)!$ boyutlu | Lie cebirleri | ✓ | Com |

### 1.3 İkili Kuadratik Operadlar

$$P = \mathcal{F}(E) / (R), \quad E \in \text{arity 2},\quad R \subset \mathcal{F}(E)(3)$$

Üreteci ikili işlemler, ilişkileri arity-3'te kuadratik olan operadlar **Koszul dualite teorisi** için doğal giriş uzayıdır.

### 1.4 Sonsuz Cebirler

| Cebir | Operadlar | Temel özellik |
|-------|-----------|---------------|
| $A_\infty$ | $\mathrm{Ass}_\infty = \Omega(\mathrm{Ass}^!)$ | $\{m_n\}_{n\geq 1}$, bağıntı: $\sum_{i+j+k} (-1)^{\cdot} m_{i+1+k}(\cdots m_j \cdots) = 0$ |
| $L_\infty$ | $\mathrm{Lie}_\infty = \Omega(\mathrm{Com})$ | $\{l_n\}_{n\geq 1}$, genelleştirilmiş Jacobi bağıntıları |
| $E_\infty$ | $\mathrm{colim}_n E_n$ | Tüm üst homotopi koşulları sağlayan değişmeli yapı |

## 2. Teoremler

### Teorem 1 — Ginzburg-Kapranov Koszul Dualitesi (1994)

**İfade:** $P = \mathcal{F}(E)/(R)$ ikili kuadratik bir operad olsun. Koszul dual operad:
$$P^! = \mathcal{F}(E^*[1]) / (R^\perp)$$
ile tanımlanır, burada $E^*[1] = \mathrm{Hom}(E, k) \otimes \mathrm{sgn}_2$ ve $R^\perp$ doğal eşleşme altında $R$'nin dik tümlenidir.

**Temel sonuçlar:** $\mathrm{Com}^! = \mathrm{Lie}$, $\mathrm{Lie}^! = \mathrm{Com}$, $\mathrm{Ass}^! = \mathrm{Ass}$ (twisted).

### Teorem 2 — Koszul Operadların Minimal Çözümleri

**İfade:** $P$ Koszul bir operad olsun. O zaman kobar inşası $\Omega(P^!) \to P$ bir zayıf denklik ve $P$'nin minimal serbest çözümüdür. $P_\infty = \Omega(P^!)$-cebirleri, $P$-cebirlerinin homotopi karşılıklarıdır.

**Sonuç:** Ass, Com, Lie Koszul $\Rightarrow$ $A_\infty$, $C_\infty$, $L_\infty$ iyi tanımlı.

### Teorem 3 — Homotopi Transfer Teoremi (A_∞)

**İfade:** $(A, \{m_n\})$ bir $A_\infty$-cebir ve $i: H \to A$, $p: A \to H$, $h: A \to A$ bir deformasyon geri çekme olsun. O zaman $H$ üzerinde
$$m_n^H = \sum_T p\, m_{|T|}\bigl(\text{alt ağaçlara } m_k \text{ ve } h\bigr)$$
formülüyle bir $A_\infty$-yapı aktarılır (Stasheff asosyahedra köşeleri üzerinden toplam).

### Teorem 4 — May Tanıma İlkesi (E_n)

**İfade:** $X$ bağlantılı bir uzay olsun. $X$ bir grup-tümleyen $E_n$-cebirinin homotopi eşdeğeridir ancak ve ancak $X \simeq \Omega^n Y$ (bazı $Y$ için $n$-katlı döngü uzayı).

**Özel durum:** $E_1 \simeq \mathrm{Ass}$ (A_∞-cebirler = 1-katlı döngü uzayları), $E_\infty \simeq \mathrm{Com}$ (E_∞-cebirler = sonsuz katlı döngü uzayları).

````python
# Teoremleri profil etiketleriyle doğrula
class _T:
    def __init__(self, *tags):
        self.tags = set(tags)
        self.representation = "verification"

# Teorem 1: Koszul dualite Com^! = Lie
com = _T("comm_operad", "koszul_duality")
r1a = admits_koszul_dual(com)
r1b = is_koszul_operad(com)
print(f"Teorem 1 — Com Koszul dual var: {r1a.is_true} | Com Koszul: {r1b.is_true}")

# Teorem 2: Ass minimal çözümü A_infty
ass = _T("assoc_operad", "a_infinity", "stasheff_associahedra")
r2 = has_infinity_algebra_structure(ass)
print(f"Teorem 2 — Ass'dan A_infty çözümü: {r2.is_true} ({r2.metadata['criterion']})")

# Teorem 3: Homotopi transfer (bar-cobar)
bar_cb = _T("bar_construction", "homotopy_transfer", "minimal_model")
r3 = has_infinity_algebra_structure(bar_cb)
print(f"Teorem 3 — Bar-cobar homotopi transfer: {r3.is_true} ({r3.metadata['criterion']})")

# Teorem 4: E_n döngü uzayı tanıma
en = _T("little_disks_operad", "e_n_operad", "recognition_principle")
r4 = is_koszul_operad(en)
print(f"Teorem 4 — E_n Koszul: {r4.is_true} ({r4.metadata['criterion']})")
````

## 3. Fonksiyon Analizi

### `is_koszul_operad(space)`

**Pseudo-kod:**
1. `tags = _extract_tags(space)`
2. `KOSZUL_DUALITY_TAGS` ile kesişim → `Result.true` (kriter: `explicit_koszul`)
3. Bilinen Koszul operadlar (Ass, Com, Lie, E_n) → `Result.true` (kriter: `known_koszul_operad`)
4. Koszul olmayan etiketler (`non_koszul`, `non_quadratic_operad`) → `Result.false`
5. Veri yetersiz → `Result.unknown`

**Karmaşıklık:**
- Zaman: $O(|\text{tags}|)$
- Alan: $O(|\text{tags}|)$

**Avantajlar:**
- Ginzburg-Kapranov teoremini doğrudan encode eder
- Ass, Com, Lie, E_n ve Poisson operadlarını tek adımda tanır

**Dezavantajlar / Sınırlamalar:**
- Bilinmeyen ikili kuadratik operadların Koszulluğu hesaplanamaz (etiket tabanlı)

**Nasıl aşılır:**
- Koszul kompleksinin açık açıklık hesabı gelecekte eklenebilir

````python
cases = [
    ("Koszul dualite etiketi",       ["koszul_duality"]),
    ("Koszul çözüm etiketi",         ["koszul_resolution"]),
    ("Ass operadı",                  ["assoc_operad"]),
    ("Com operadı",                  ["comm_operad"]),
    ("Lie operadı",                  ["lie_operad"]),
    ("E_n operadı",                  ["e_n_operad", "little_disks_operad"]),
    ("Poisson operadı",              ["poisson_operad"]),
    ("Koszul değil",                 ["non_koszul"]),
    ("İkili kuadratik değil",        ["non_quadratic_operad"]),
    ("Belirsiz",                     ["topological_manifold"]),
]
print(f"{'Durum':30s} {'Sonuç':10} {'Kriter'}")
print("-" * 65)
for label, tags in cases:
    s = type("S", (), {"tags": set(tags), "representation": "test"})()
    r = is_koszul_operad(s)
    status = "TRUE" if r.is_true else ("FALSE" if r.is_false else "UNKNOWN")
    print(f"{label:30s} {status:10} {r.metadata['criterion']}")
````

### `has_infinity_algebra_structure(space)`

**Pseudo-kod:**
1. `tags = _extract_tags(space)`
2. `INFINITY_ALGEBRA_TAGS` ile kesişim → `Result.true` (kriter: `explicit_infinity_algebra`)
3. `BAR_COBAR_TAGS` ile kesişim → `Result.true` (kriter: `bar_cobar_construction`)
4. Sıkı cebirlerin (Ass/Com/Lie) etiketleri → `Result.true` (kriter: `strict_implies_infinity`)
5. `strict_algebra_only` / `no_higher_operations` → `Result.false`
6. Veri yetersiz → `Result.unknown`

**Karmaşıklık:**
- Zaman: $O(|\text{tags}|)$
- Alan: $O(|\text{tags}|)$

**Avantajlar:**
- Homotopi transfer teoremini dolaylı olarak kodlar: sıkı cebir → $P_\infty$-cebir
- Maurer-Cartan etiketiyle L_∞-yapıyı da tanır

**Dezavantajlar / Sınırlamalar:**
- $m_3, m_4, \ldots$ operasyonlarının sıfır olup olmadığı doğrulanamaz

**Nasıl aşılır:**
- `strict_algebra_only` etiketi ile sıkı ve homotopi cebirleri birbirinden ayırt etmek yeterli

````python
cases = [
    ("A_∞ etiketi",                ["a_infinity", "infinity_algebra"]),
    ("L_∞ etiketi",                ["l_infinity"]),
    ("C_∞ etiketi",                ["c_infinity"]),
    ("Homotopi transfer",           ["homotopy_transfer", "minimal_model"]),
    ("Bar inşası",                  ["bar_construction"]),
    ("Kobra inşası",                ["cobar_construction", "dg_operad"]),
    ("Sıkı ilişkilendirilebilir",  ["associative_algebra"]),
    ("Ass operadı",                ["assoc_operad"]),
    ("Maurer-Cartan",              ["maurer_cartan"]),
    ("Sadece sıkı cebir",         ["strict_algebra_only"]),
    ("Üst işlem yok",             ["no_higher_operations"]),
    ("Belirsiz",                   ["compact_space"]),
]
print(f"{'Durum':30s} {'Sonuç':10} {'Kriter'}")
print("-" * 65)
for label, tags in cases:
    s = type("S", (), {"tags": set(tags), "representation": "test"})()
    r = has_infinity_algebra_structure(s)
    status = "TRUE" if r.is_true else ("FALSE" if r.is_false else "UNKNOWN")
    print(f"{label:30s} {status:10} {r.metadata['criterion']}")
````

### `admits_koszul_dual(space)`

**Pseudo-kod:**
1. `tags = _extract_tags(space)`
2. `KOSZUL_DUALITY_TAGS` ile kesişim → `Result.true` (kriter: `explicit_koszul_dual`)
3. Bilinen ikili kuadratik operadlar (Ass/Com/Lie/Poisson/pre-Lie/Zinbiel/Dendriform) → `Result.true` (kriter: `binary_quadratic_has_dual`)
4. Renkli/modüler/prop-tipi operadlar → `Result.false` (kriter: `no_standard_koszul_dual`)
5. Veri yetersiz → `Result.unknown`

**Karmaşıklık:**
- Zaman: $O(|\text{tags}|)$
- Alan: $O(|\text{tags}|)$

**Avantajlar:**
- Koszul dual tanımı için gerekli ikili kuadratik yapıyı doğrudan sorgular

**Dezavantajlar / Sınırlamalar:**
- Renkli Koszul dualite (Berger-Moerdijk) ve prop-Koszul dualite kapsam dışı

**Nasıl aşılır:**
- Renkli operad için ayrı bir `admits_colored_koszul_dual` fonksiyonu eklenebilir

````python
cases = [
    ("Koszul dual etiketi",         ["koszul_dual"]),
    ("Kuadratik dual",              ["quadratic_dual"]),
    ("Ass operadı",                ["assoc_operad"]),
    ("Com operadı",                ["comm_operad"]),
    ("Lie operadı",                ["lie_operad"]),
    ("pre-Lie operadı",            ["pre_lie_operad"]),
    ("Zinbiel operadı",            ["zinbiel_operad"]),
    ("Renkli operad",              ["colored_operad"]),
    ("Modüler operad",             ["modular_operad"]),
    ("Prop operadı",               ["prop_operad"]),
    ("İkili olmayan",              ["non_binary"]),
    ("Belirsiz",                   ["symplectic_form"]),
]
print(f"{'Durum':30s} {'Sonuç':10} {'Kriter'}")
print("-" * 65)
for label, tags in cases:
    s = type("S", (), {"tags": set(tags), "representation": "test"})()
    r = admits_koszul_dual(s)
    status = "TRUE" if r.is_true else ("FALSE" if r.is_false else "UNKNOWN")
    print(f"{label:30s} {status:10} {r.metadata['criterion']}")
````

### `is_binary_quadratic_operad(space)`

**Pseudo-kod:**
1. `tags = _extract_tags(space)`
2. Açık ikili kuadratik etiketler veya `TREE_COMPOSITION_TAGS` → `Result.true` (kriter: `explicit_binary_quadratic`)
3. Bilinen ikili kuadratik operadlar (Ass/Com/Lie/Poisson/pre-Lie/Dendriform) → `Result.true` (kriter: `known_binary_quadratic`)
4. Renkli/modüler/A_∞/L_∞ veya yüksek-arity etiketleri → `Result.false`
5. Veri yetersiz → `Result.unknown`

**Karmaşıklık:**
- Zaman: $O(|\text{tags}|)$
- Alan: $O(|\text{tags}|)$

**Avantajlar:**
- Koszul dualite teorisinin ön koşulunu kontrol eder
- A_∞/L_∞ gibi tüm-arity operadları FALSE olarak doğru sınıflandırır

**Dezavantajlar / Sınırlamalar:**
- Bilinmeyen bir operadın ikili kuadratik sunumu kontrol edilemez

**Nasıl aşılır:**
- Açık `binary_quadratic` etiketiyle herhangi bir ikili kuadratik operad doğrudan işaretlenebilir

````python
cases = [
    ("İkili kuadratik etiketi",    ["binary_quadratic"]),
    ("Ağaç bileşimi",              ["tree_composition", "operadic_composition"]),
    ("Ass operadı",                ["assoc_operad"]),
    ("Com operadı",                ["comm_operad"]),
    ("Lie operadı",                ["lie_operad"]),
    ("Poisson operadı",            ["poisson_operad"]),
    ("pre-Lie operadı",            ["pre_lie_operad"]),
    ("Renkli operad",              ["colored_operad"]),
    ("A_∞ operadı",               ["a_infinity_operad"]),
    ("L_∞ operadı",               ["l_infinity_operad"]),
    ("Yüksek arity",              ["higher_arity"]),
    ("Belirsiz",                   ["hopf_algebra"]),
]
print(f"{'Durum':30s} {'Sonuç':10} {'Kriter'}")
print("-" * 65)
for label, tags in cases:
    s = type("S", (), {"tags": set(tags), "representation": "test"})()
    r = is_binary_quadratic_operad(s)
    status = "TRUE" if r.is_true else ("FALSE" if r.is_false else "UNKNOWN")
    print(f"{label:30s} {status:10} {r.metadata['criterion']}")
````

## 5. Örnekler

### Örnek 1 — Com operadı: Koszul dualite ve L_∞ bağlantısı

````python
class ComOperad:
    tags = {"comm_operad", "com_operad", "commutative_algebra",
            "koszul_duality", "cdga", "e_infinity"}
    representation = "com_operad"

com = ComOperad()
result = classify_operad(com)

print("Com Operadı — Tam Sınıflandırma")
print("=" * 50)
for key, r in result.items():
    status = "✓ TRUE" if r.is_true else ("✗ FALSE" if r.is_false else "? UNKNOWN")
    print(f"  {key}: {status}")

# Profil sayfasından odak
com_p = next(p for p in profiles if p.key == "comm_operad")
print(f"\nKoszul dual: {'Lie' if com_p.has_koszul_dual else 'yok'}")
print(f"İkili kuadratik: {com_p.is_binary_quadratic}")
print(f"Sonsuz versiyonu: {com_p.admits_infinity_version} (C_∞ = Omega(Lie))")
````

### Örnek 2 — A_∞ operadı: Homotopi yapısı ve ikili kuadratik değil

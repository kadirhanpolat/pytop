## 1. Konu

### 1.1 Sezgisel Açıklama

Manifold, **her noktasının küçük bir komşuluğunun Öklid uzayına homeomorf göründüğü** bir uzaydır.

**Sezgi 1 — Yerel düzlük:**  
Dünya yüzeyinde bir noktaya yakından bakın: düzlemmiş gibi görünür. Küresel olarak kavisli, ama yerel olarak düzdür. Bu $S^2$'nin bir 2-manifold olduğunun sezgisel anlamıdır.

**Sezgi 2 — Kart ve atlas:**  
Coğrafi harita analojisi: tek bir harita (kart) tüm yeryüzünü temsil edemez — en az iki kart gerekir. Kartların birleşimi **atlas** oluşturur. Topolojik bir manifold, tutarlı bir atlas seçimine karşılık gelir.

**Sezgi 3 — Sınırlı manifold:**  
Disk $D^2$'nin iç noktaları $\mathbb{R}^2$'ye homeomorf görünürken sınır noktaları (çember üzerindekiler) yarım-düzleme ($\mathbb{H}^2$) homeomorf görünür. Bu **sınırlı manifold** kavramıdır.

**Boyut bağımsızlığı:**  
$n$-manifold her noktasının $\mathbb{R}^n$'ye (veya $\mathbb{H}^n$'ye) homeomorf bir komşuluğu vardır. Alan değişmezlik teoremi, $\mathbb{R}^m \cong \mathbb{R}^n \Rightarrow m = n$ garantisiyle boyutu iyi tanımlı kılar.

**Neden önemli?**
- Diferansiyel geometri ve fizik: uzay-zaman manifoldlardır
- Yüzey sınıflandırması 2-manifoldların tam teorisidir
- 3-manifold teorisi (Poincaré, Geometrization) açık araştırma problemlerini kapsar
- Her manifold topolojik gruptur (Lie grupları)

---

### 1.2 Formal Tanımlar

**$n$-Manifold:** $(X, \tau)$ Hausdorff, ikinci sayılabilir, her $x \in X$ için $\exists$ açık $U \ni x$: $U \cong \mathbb{R}^n$.

**Sınırlı $n$-Manifold:** Her nokta $\mathbb{R}^n$ veya $\mathbb{H}^n = \{x \in \mathbb{R}^n : x_n \geq 0\}$'a homeomorf komşuluğa sahip. $\mathbb{H}^n$'ye homeomorf olan noktalar **sınırı** $\partial X$ oluşturur.

**Kart:** $(U, \phi)$ — açık $U \subseteq X$ ve $\phi: U \to \mathbb{R}^n$ homeomorfizma.

**Atlas:** Örtü oluşturan kartlar ailesi.

**Yönlendirilebilirlik:** Atlas geçiş haritalarının Jacobian determinantları her yerde pozitif seçilebiliyorsa manifold yönlendirilebilir.

## 2. Teoremler

### Teorem 1 — $n$-Manifold Hausdorff ve İkinci Sayılabilirdir

**İfade:** Tanım gereği her manifold Hausdorff ve ikinci sayılabilirdir.

**Neden bu koşullar konulur?**  
- **Hausdorff:** Tuhaf örnekleri dışlar (örn. doğruyu ikiye katlayan uzay manifold tanımını karşılar ama Hausdorff değildir). Koordinat değişimlerinin tutarlılığını garanti eder.
- **İkinci sayılabilir:** Partisyon-of-unity ve paracompactness'ı garanti eder (her açık kaplama sayılabilir alt kaplama içerir). Riemannian metrik inşasını mümkün kılar.

**Sonuçlar:** Her manifold:
- Metrizlenebilir (Urysohn metrization: T3 + ikinci sayılabilir → metrik)
- Paracompact (Stone teoremi)
- $\sigma$-kompakt (sayılabilir kompakt örtü)

---

### Teorem 2 — Alan Değişmezliği (Invariance of Domain)

**İfade:**  
$f: U \to \mathbb{R}^n$ enjektif sürekli, $U \subseteq \mathbb{R}^n$ açık. O zaman $f(U)$ açıktır ve $f$ homeomorfizmadır.

**Sonuç:** $\mathbb{R}^m \cong \mathbb{R}^n \Rightarrow m = n$ (manifold boyutu iyi tanımlı).

**İspat taslağı (Brouwer, 1912 — algebraik topoloji yoluyla):**  
$x \in U$ için $f(x)$'in iç nokta olduğunu göster. Yerel indirgeme indeksini (local degree) kullan:  
$\deg(f, U, y) = 1$ ($f$ enjektif olduğundan) — bu $y \in f(U)$'nun iç nokta olduğunu garantiler.  
$f$ sürekli enjeksiyon + açık görüntü = homeomorfizma. $\square$

---

### Teorem 3 — $S^1$ Tam Olarak İki Kart Gerektirir

**İfade:** $S^1$'i tek bir kartla kaplayan bir atlas yoktur.

**İspat:**  
$S^1$ kompakttır ama $\mathbb{R}$ kompakt değildir. Tek kart $\phi: U \to \mathbb{R}$ varsa $U = S^1$ (bütün uzay) olmalıydı; bu $S^1 \cong \mathbb{R}$ verir. $S^1$ kompakt, $\mathbb{R}$ değil — çelişki. $\square$

**Standart atlas:** Kuzey ve güney kutup dışında stereografik projeksiyon kartları.

## 3. API

```python
from pytop.manifolds import (
    ManifoldProfile,
    ManifoldProfileError,
    manifold_profile,
    real_line_manifold_profile,
    circle_manifold_profile,
    sphere_manifold_profile,
    disk_with_boundary_profile,
    torus_manifold_profile,
    projective_plane_manifold_profile,
    unknown_manifold_profile,
    known_manifold_profile,
    manifold_profile_summary,
    KNOWN_MANIFOLD_PROFILES,
    MANIFOLD_ORIENTABILITY_VALUES,
)
print("pytop.manifolds yüklendi.")
print("Kayıtlı anahtarlar:", sorted(set(KNOWN_MANIFOLD_PROFILES.keys())))
```

### `ManifoldProfile` Alanları ve Özellikleri

| Alan | Tür | Açıklama |
|------|-----|----------|
| `name` | `str` | Profil adı |
| `dimension` | `int` | Manifold boyutu $n$ |
| `with_boundary` | `bool` | Sınırlı manifold mu? |
| `local_model` | `str` | Yerel model ($\mathbb{R}^n$ veya $\mathbb{H}^n$) |
| `orientability` | `str` | `"orientable"` \| `"nonorientable"` \| `"unknown"` |
| `compact` | `bool\|None` | Kompaktlık |
| `connected` | `bool\|None` | Bağlılık |
| `charts` | `tuple[str]` | Atlas kart isimleri |
| `atlas_note` | `str` | Atlas açıklaması |
| `status` | `str` | `"certified"` \| `"preview"` \| ... |

| Özellik | Açıklama |
|---------|----------|
| `.is_certified` | `status == "certified"` |
| `.is_surface` | `dimension == 2` |
| `.has_boundary` | `with_boundary` |

### Fabrika Fonksiyonları

| Fonksiyon | Üretir |
|-----------|--------|
| `real_line_manifold_profile()` | $\mathbb{R}$ (1-manifold) |
| `circle_manifold_profile()` | $S^1$ (kompakt 1-manifold) |
| `sphere_manifold_profile(n)` | $S^n$ ($n \geq 1$) |
| `disk_with_boundary_profile(n)` | $D^n$ (sınırlı $n$-manifold) |
| `torus_manifold_profile()` | $T^2$ (kompakt 2-manifold) |
| `projective_plane_manifold_profile()` | $\mathbb{RP}^2$ (yönlendirilemeyen) |

## 4. Örnekler

### Örnek 1 — Minimal: Kayıtlı Manifoldların Katalogu

```python
# Tekrarsız profil listesi
seen = set()
manifolds = []
for key, mp in KNOWN_MANIFOLD_PROFILES.items():
    if mp.name not in seen:
        seen.add(mp.name)
        manifolds.append(mp)

print(f"{'Ad':25} {'Boyut':6} {'Yön.':14} {'Sınırlı':8} {'Kompakt':8} {'Onaylı'}")
print("-" * 75)
for mp in manifolds:
    print(f"  {mp.name:23} {mp.dimension:6} {mp.orientability:14} "
          f"{str(mp.has_boundary):8} {str(mp.compact):8} {mp.is_certified}")
```

### Örnek 2 — Orta Düzey: $n$-Küre ve Disk Atlas Notları

```python
# n-küre: boyuta göre atlas ve yerel model
print("=== n-Küre Profilleri ===")
for n in [1, 2, 3, 4]:
    sn = sphere_manifold_profile(n)
    print(f"  S^{n}: dim={sn.dimension}, yerel model={sn.local_model}, "
          f"kartlar={sn.charts}")

print()

# Disk: sınırlı manifold
print("=== Sınırlı Manifoldlar (Disk) ===")
for n in [1, 2, 3]:
    dn = disk_with_boundary_profile(n)
    boundary = dn.metadata.get("boundary_model", "?")
    print(f"  D^{n}: yerel model='{dn.local_model[:40]}...'")
    print(f"       sınır: {boundary}, sınır boyutu: {dn.metadata.get('boundary_dimension')}")
```

### Örnek 3 — Gelişmiş: Yönlendirilebilirlik ve Bağlantılı Profil Listesi

```python
# Yönlendirilebilir vs yönlendirilemeyen karşılaştırma
orient_map = {"orientable": "✓", "nonorientable": "✗", "unknown": "?"}

print("=== Yönlendirilebilirlik Karşılaştırması ===")
all_profiles = []
seen2 = set()
factories = [
    real_line_manifold_profile(),
    circle_manifold_profile(),
    sphere_manifold_profile(2),
    sphere_manifold_profile(3),
    disk_with_boundary_profile(2),
    torus_manifold_profile(),
    projective_plane_manifold_profile(),
]
for mp in factories:
    if mp.name not in seen2:
        seen2.add(mp.name)
        all_profiles.append(mp)

print(f"{'Ad':25} {'Boyut':6} {'Yönlendirme':5} {'Yüzey':7} {'Sınır'}")
print("-" * 60)
for mp in all_profiles:
    icon = orient_map[mp.orientability]
    print(f"  {mp.name:23} {mp.dimension:6} {icon:5}         "
          f"{'✓' if mp.is_surface else '✗':7} {'✓' if mp.has_boundary else '✗'}")

print()
print("İlgili profiller:")
for mp in [torus_manifold_profile(), projective_plane_manifold_profile()]:
    print(f"  {mp.name}: {list(mp.related_profiles)}")
```

### Örnek 4 — Gerçekçi Senaryo: Özel Manifold Profili + `manifold_profile_summary`

```python
# 3-torus T^3 = S^1 x S^1 x S^1
three_torus = manifold_profile(
    "torus_T3",
    3,
    local_model="R^3",
    orientability="orientable",
    compact=True,
    connected=True,
    charts=("product of three circle-arc atlases",),
    atlas_note="T^3 = S^1 x S^1 x S^1; product atlas from three circle charts.",
    status="certified",
    related_profiles=("torus_t2", "covering_spaces", "fundamental_group"),
    metadata={"pi1_hint": "Z^3", "euler_characteristic": 0},
)

summary = manifold_profile_summary(three_torus)
print("=== T³ Özeti ===")
for key, val in summary.items():
    print(f"  {key:18}: {val}")

print()

# Bilinmeyen manifold
mystery = unknown_manifold_profile("exotic_4manifold", dimension=4)
print(f"Bilinmeyen: {mystery.name}")
print(f"  dim={mystery.dimension}, status={mystery.status}")
print(f"  uyarı: {mystery.warnings[0]}")
```

## 5. Alıştırmalar

### Alıştırma 1 — Kodlama Görevi

Aşağıdaki manifoldları `manifold_profile` ile oluşturun:

1. **Möbius bandı** (1-sınırlı, yönlendirilemeyen, kompakt): `dimension=2`, `with_boundary=True`, `orientability="nonorientable"`
2. **Klein şişesi** (kapalı, yönlendirilemeyen, kompakt): `dimension=2`, `with_boundary=False`, `orientability="nonorientable"`
3. **$S^2 \times S^1$** (kompakt 3-manifold, yönlendirilebilir): `dimension=3`

Her birini bir tabloda karşılaştırın.

```python
mobius = manifold_profile(
    "mobius_band",
    2,
    with_boundary=True,
    local_model="R^2 interior; H^2 boundary",
    orientability="nonorientable",
    compact=True,
    connected=True,
    charts=("rectangle with twisted identification",),
    atlas_note="One reversed edge identification; boundary is S^1.",
    status="certified",
    metadata={"pi1_hint": "Z"},
)

klein = manifold_profile(
    "klein_bottle",
    2,
    with_boundary=False,
    local_model="R^2",
    orientability="nonorientable",
    compact=True,
    connected=True,
    charts=("square edge identification: one reversed pair",),
    atlas_note="Gluing word a b a^-1 b; embedded in R^4 but not R^3 without self-intersection.",
    status="certified",
    metadata={"pi1_hint": "<a,b | a b a^-1 = b^-1>", "euler_characteristic": 0},
)

s2_s1 = manifold_profile(
    "S2_cross_S1",
    3,
    local_model="R^3",
    orientability="orientable",
    compact=True,
    connected=True,
    charts=("product of sphere and circle charts",),
    atlas_note="Product manifold; pi_1 = Z (from S^1 factor).",
    status="certified",
    metadata={"pi1_hint": "Z"},
)

print(f"{'Ad':20} {'Boyut':6} {'Yön.':14} {'Sınır':7} {'Kompakt':8} {'π₁'}")
print("-" * 72)
for mp in [mobius, klein, s2_s1]:
    pi1 = mp.metadata.get("pi1_hint", "?")
    print(f"  {mp.name:18} {mp.dimension:6} {mp.orientability:14} "
          f"{str(mp.has_boundary):7} {str(mp.compact):8} {pi1}")
```

### Alıştırma 4 — $n$-Küre ve $n$-Disk Profil Karşılaştırması

`sphere_manifold_profile(n)` ve `disk_with_boundary_profile(n)` fonksiyonlarını $n = 1, 2, 3, 4$ için çalıştırın:
1. Boyut, yerel model, yönlendirme, sınır, kompaktlık değerlerini tek bir tabloda karşılaştırın
2. $S^n$ her zaman sınırsız ($\partial = \emptyset$) ve $D^n$ her zaman sınırlı olduğunu doğrulayın
3. `disk_with_boundary_profile`'daki `metadata["boundary_model"]` ile $\partial D^n = S^{n-1}$ ilişkisini kontrol edin

```python
print(f"{'Ad':18} {'Dim':5} {'Yön.':14} {'Sınır':7} {'Kompakt':8} {'Yerel model'}")
print("-" * 75)
for n in range(1, 5):
    sn = sphere_manifold_profile(n)
    dn = disk_with_boundary_profile(n)
    for mp in [sn, dn]:
        print(f"  {mp.name:16} {mp.dimension:5} {mp.orientability:14} "
              f"{str(mp.has_boundary):7} {str(mp.compact):8} {mp.local_model[:25]}")

print()

# Sınırsız doğrulama: S^n her zaman has_boundary=False
sn_boundary_free = all(not sphere_manifold_profile(n).has_boundary for n in range(1, 5))
dn_bounded = all(disk_with_boundary_profile(n).has_boundary for n in range(1, 5))
print(f"S^n hepsi sınırsız (n=1..4): {sn_boundary_free}")
print(f"D^n hepsi sınırlı  (n=1..4): {dn_bounded}")

print()

# ∂D^n = S^(n-1) doğrulama
print("∂D^n = S^(n-1) doğrulaması:")
for n in range(1, 5):
    dn = disk_with_boundary_profile(n)
    bdim = dn.metadata.get("boundary_dimension", "?")
    bmodel = dn.metadata.get("boundary_model", "?")
    expected_dim = n - 1
    match = "✓" if bdim == expected_dim else "?"
    print(f"  D^{n}: sınır_boyutu={bdim} (beklenen {expected_dim}) {match}  model={bmodel}")
```

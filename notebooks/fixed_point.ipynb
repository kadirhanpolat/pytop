# Sabit Nokta Teoremleri

**Modül:** `pytop.fixed_point_profiles`  
**Kaynak:** Adams & Franzosa *Introduction to Topology* §9.2, §10.1

---

Bir **sabit nokta teoremi**, belirli koşulları sağlayan her sürekli dönüşümün en az bir noktayı yerine bıraktığını güvence altına alır.  
**Brouwer Sabit Nokta Teoremi (FPT)** topolojinin en sezgisel ve en güçlü sonuçlarından biridir;
hem matematiksel teori hem de uygulamalı bilimlerde temel bir araçtır.

**Motivasyon:**  
- Bir bardak kahveyi karıştırdıktan sonra en az bir sıvı parçacığı başlangıç konumuna döner.  
- Nash denge varlığının kanıtı Brouwer FPT'ye dayanır.  
- Banach büzülme dönüşümü teoremi Brouwer'in sayısal versiyonudur.

Bu notebook beş bölümden oluşur:
1. **Konu** — Retraction, no-retraction teoremi ve Brouwer FPT tanımı
2. **Teoremler** — No-retraction ($n=1$, $n=2$, genel), Brouwer FPT, kompakt konveks genelleme
3. **API** — `pytop.fixed_point_profiles` fonksiyon referansı
4. **Örnekler** — Dört senaryo: boyut analizi, uygulama bağlamları, kanıt zinciri
5. **Alıştırmalar** — Kodlama ve teori

## 1. Konu

### 1.1 Retraction (Geri Çekim)

**Tanım.** $A \subseteq X$ olmak üzere, $r : X \to A$ **sürekli** ve $r|_A = \mathrm{id}_A$ ise $r$ bir **retraction**, $A$ ise $X$'in bir **retract**'ıdır.

Eğer ek olarak $r \simeq \mathrm{id}_X$ (homotopi $A$'ya sabit) ise, $A$ bir **deformation retract**'ıdır; bu durumda $X \simeq A$ (homotopi denk).

| Uzay $X$ | Retract $A$ | Formül | Deformation? |
|----------|------------|--------|:------------:|
| $\mathbb{R}^n$ | $\{0\}$ | $r(x) = 0$ | ✓ |
| $\mathbb{R}^2 \setminus \{0\}$ | $S^1$ | $r(x) = x/|x|$ | ✓ |
| $S^1 \times [0,1]$ | $S^1 \times \{0\}$ | $r(\theta,t) = (\theta,0)$ | ✓ |
| $D^n$ | $\{0\}$ | $r(x) = 0$ | ✓ |

### 1.2 No-Retraction Teoremi

> $D^n$'den sınırı $S^{n-1}$'e **hiçbir** sürekli retraction yoktur.

Kanıt yöntemi boyuta göre değişir:
- $n = 1$: Bağlantılılık ($[-1,1]$ bağlantılı, $\{-1,+1\}$ değil)
- $n = 2$: $\pi_1$ veya derece teorisi
- Genel $n$: Singular homoloji ($H_{n-1}(S^{n-1}) = \mathbb{Z}$, $H_{n-1}(D^n) = 0$)

### 1.3 Brouwer Sabit Nokta Teoremi

> Her sürekli $f : D^n \to D^n$ dönüşümünün en az bir **sabit noktası** vardır: $f(x) = x$ olan bir $x \in D^n$ mevcuttur.

**Kanıt fikri.** $f$'nin sabit noktası yoksa, $f(x)$'ten $x$'e giden ışının $S^{n-1}$ ile kesim noktası $r(x)$ tanımlanır. Bu $r$, $D^n \to S^{n-1}$ bir retractiondır — No-Retraction Teoremine çelişir.

```python
from pytop.fixed_point_profiles import (
    NoRetractionProfile,
    get_no_retraction_profiles,
    no_retraction_by_dimension,
    BrouwerFPTProfile,
    get_brouwer_fpt_profiles,
    brouwer_fpt_by_dimension,
    RetractionProfile,
    get_retraction_profiles,
    fixed_point_theorem_registry,
)

print("=== No-Retraction Profilleri ===")
for p in get_no_retraction_profiles():
    dim = "genel n" if p.dimension == -1 else f"n = {p.dimension}"
    print(f"  [{dim}]  {p.display_name}")
    print(f"    Uzay    : {p.ambient_space}")
    print(f"    Sınır   : {p.boundary_sphere}")
    print(f"    Kanıt   : {p.proof_method}")
    print()
```

```python
print("=== Brouwer FPT Profilleri ===")
for p in get_brouwer_fpt_profiles():
    dim = "genel n" if p.dimension == -1 else f"n = {p.dimension}"
    print(f"  [{dim}]  {p.display_name}")
    ctx = p.application_context[:75] + "..." if len(p.application_context) > 75 else p.application_context
    print(f"    Uygulama: {ctx}")
    print()

print("=== Pozitif Retraction Profilleri ===")
for p in get_retraction_profiles():
    dr = "Deformation ✓" if p.deformation_retract else "Retraction"
    print(f"  [{dr}]  {p.display_name}")
    print(f"    Formül  : {p.retraction_formula}")
    print()
```

## 2. Teoremler

### Teorem 1 — No-Retraction Teoremi ($n = 1$)

> **Teorem.** $[-1,1]$'den $\{-1,+1\}$'e sürekli bir retraction yoktur.

**Kanıt.** $[-1,1]$ bağlantılıdır. Bir retraction $r : [-1,1] \to \{-1,+1\}$ mevcut olsaydı, bağlantılı bir uzaydan süreksiz bir uzaya sürekli ve örten bir dönüşüm olurdu — sürekli dönüşümlerin bağlantılılığı koruması ilkesiyle çelişir. $\square$

---

### Teorem 2 — No-Retraction Teoremi ($n = 2$)

> **Teorem.** $D^2$'den $S^1$'e sürekli bir retraction yoktur.

**Kanıt.** $r : D^2 \to S^1$ retraction, $i : S^1 \hookrightarrow D^2$ dahil dönüşümü varsayılsın. $r \circ i = \mathrm{id}_{S^1}$ olduğundan temel grup düzeyinde $(r \circ i)_* = \mathrm{id}_{\mathbb{Z}}$, yani $r_* \circ i_* = \mathrm{id}$. Oysa $i_* : \pi_1(S^1) = \mathbb{Z} \to \pi_1(D^2) = 0$ sıfır dönüşümüdür; dolayısıyla $r_* \circ 0 = 0 \ne \mathrm{id}$. Çelişki. $\square$

---

### Teorem 3 — Brouwer Sabit Nokta Teoremi

> **Teorem (Brouwer, 1912).** Her sürekli $f : D^n \to D^n$ dönüşümünün en az bir sabit noktası vardır.

**Kanıt.** $f(x) \ne x$ (tüm $x \in D^n$) olduğu varsayılsın. Her $x$ için, $f(x)$'ten $x$'e uzanan ışının $S^{n-1}$ ile kesişim noktası:
$$r(x) = x + t(x)\,(x - f(x)), \quad t(x) \ge 0,\ |r(x)| = 1.$$
$r$ süreklidir (sabit-noktasızlık $t$'nin sürekli belirlenmesini sağlar) ve $x \in S^{n-1}$ ise $r(x) = x$. Dolayısıyla $r$, $D^n \to S^{n-1}$ bir retractiondır — No-Retraction Toremine çelişir. $\square$

---

### Teorem 4 — Kompakt Konveks Genelleme

> **Korollar.** $K \subseteq \mathbb{R}^n$ kompakt ve konveks ise, her sürekli $f : K \to K$ dönüşümünün en az bir sabit noktası vardır.

**Kanıt fikri.** $K$, $D^n$'e homeomorftur; homeomorfizm Brouwer FPT'yi aktarır. $\square$

**Önemli genellemeler:** Schauder FPT (sonsuz boyutlu Banach uzayları), Kakutani FPT (çok-değerli dönüşümler, Nash dengesi), Lefschetz FPT (Lefschetz sayısı koşulu).

```python
# Teorem sabitlerini doğrula
nr = get_no_retraction_profiles()
br = get_brouwer_fpt_profiles()
rt = get_retraction_profiles()

all_nr_false = all(not p.retraction_exists for p in nr)
all_br_true  = all(p.fixed_point_exists   for p in br)
all_rt_true  = all(p.retraction_exists    for p in rt)
all_rt_deform = all(p.deformation_retract for p in rt)

print("Teorem sabitleri doğrulaması:")
print(f"  No-Retraction  → retraction_exists=False (tümü): {all_nr_false}")
print(f"  Brouwer FPT    → fixed_point_exists=True  (tümü): {all_br_true}")
print(f"  Retraction     → retraction_exists=True   (tümü): {all_rt_true}")
print(f"  Retraction     → deformation_retract=True (tümü): {all_rt_deform}")
print()
print("Detay (No-Retraction):")
for p in nr:
    dim = "gen" if p.dimension == -1 else str(p.dimension)
    print(f"  n={dim:3s}  retraction_exists={p.retraction_exists}  |  {p.proof_method}")
print()
print("Detay (Brouwer FPT):")
for p in br:
    dim = "gen" if p.dimension == -1 else str(p.dimension)
    print(f"  n={dim:3s}  fixed_point_exists={p.fixed_point_exists}  |  {p.key}")
```

```python
# no_retraction_by_dimension: boyut tablosu
print("no_retraction_by_dimension — boyut tablosu:")
print(f"  {'n':>5}  {'profil boyutu':>14}  kanıt yöntemi")
print("  " + "─" * 55)
for n in [1, 2, 3, 4, 10, 100]:
    p = no_retraction_by_dimension(n)
    dim_str = "genel (-1)" if p.dimension == -1 else str(p.dimension)
    print(f"  {n:>5}  {dim_str:>14}  {p.proof_method}")

print()
print("brouwer_fpt_by_dimension — sonuçlar:")
for n in [1, 2, -1]:
    profiles = brouwer_fpt_by_dimension(n)
    dim_str = "genel" if n == -1 else str(n)
    keys = [p.key for p in profiles]
    print(f"  n={dim_str}: {len(profiles)} profil  → {keys}")
```

## 3. API Referansı

| Fonksiyon / Sınıf | Açıklama | Döndürür |
|-------------------|----------|----------|
| `NoRetractionProfile` | No-Retraction Teoremi öğretim profili ($D^n \to S^{n-1}$ yoktur) | `dataclass(frozen=True)` |
| `get_no_retraction_profiles()` | Tüm no-retraction profillerini döndürür ($n=1$, $n=2$, genel $n$) | `tuple[NoRetractionProfile, ...]` |
| `no_retraction_by_dimension(n)` | $n$ boyutu için profil; tanımlı değilse genel profil | `NoRetractionProfile \| None` |
| `BrouwerFPTProfile` | Brouwer Sabit Nokta Teoremi öğretim profili | `dataclass(frozen=True)` |
| `get_brouwer_fpt_profiles()` | Tüm Brouwer FPT profillerini döndürür | `tuple[BrouwerFPTProfile, ...]` |
| `brouwer_fpt_by_dimension(n)` | $n$ boyutundaki Brouwer FPT profillerini döndürür | `tuple[BrouwerFPTProfile, ...]` |
| `RetractionProfile` | Pozitif retraction örneği (no-retraction teoremiyle kontrast) | `dataclass(frozen=True)` |
| `get_retraction_profiles()` | Tüm pozitif retraction profillerini döndürür | `tuple[RetractionProfile, ...]` |
| `fixed_point_theorem_registry()` | Kategori başına profil sayısı | `dict[str, int]` |

```python
registry = fixed_point_theorem_registry()
print("Kayıt defteri özeti:")
total = 0
for k, v in registry.items():
    print(f"  {k:<35s}: {v}")
    total += v
print(f"  {'Toplam':<35s}: {total}")
```

## 4. Örnekler

```python
# Örnek 1: No-retraction profillerini boyut ve kanıt yöntemine göre analiz
print("=== Boyutlara Göre No-Retraction Analizi ===\n")
for p in get_no_retraction_profiles():
    dim = "Genel n" if p.dimension == -1 else f"n = {p.dimension}"
    print(f"Boyut: {dim}")
    print(f"  {p.ambient_space}  →  {p.boundary_sphere}")
    print(f"  Kanıt yöntemi : {p.proof_method}")
    print(f"  Kaynak        : {p.source_section}")
    first = p.notes.split(".")[0] + "."
    print(f"  Özet notu     : {first}")
    print()

print("─" * 65)
print(f"  {'Boyut':>7}  {'retraction_exists':>18}  Kanıt Yöntemi")
print("─" * 65)
for p in get_no_retraction_profiles():
    dim_str = "genel n" if p.dimension == -1 else f"n = {p.dimension}"
    print(f"  {dim_str:>7}  {str(p.retraction_exists):>18}  {p.proof_method}")
```

```python
# Örnek 2: Brouwer FPT — uygulama bağlamları ve boyut gruplama
print("=== Brouwer FPT — Uygulama Bağlamları ===\n")
for p in get_brouwer_fpt_profiles():
    dim_str = "Genel n" if p.dimension == -1 else f"n = {p.dimension}"
    print(f"{p.display_name}")
    print(f"  Boyut    : {dim_str}")
    print(f"  Alan     : {p.domain}")
    ctx = p.application_context
    print(f"  Uygulama : {ctx[:100]}{'...' if len(ctx) > 100 else ''}")
    print()

print("n = 2 boyutundaki Brouwer FPT profilleri:")
for p in brouwer_fpt_by_dimension(2):
    abstract = "soyut" if "continuous" in p.map_description or "f :" in p.map_description else "somut"
    print(f"  [{p.key}]")
    print(f"    Dönüşüm: {p.map_description[:80]}...")
```

```python
# Örnek 3: Pozitif retraction profilleri — no-retraction teoremiyle kontrast
print("=== Retraction Profilleri: No-Retraction ile Kontrast ===\n")
rt_profiles = get_retraction_profiles()
deform_cnt  = sum(1 for p in rt_profiles if p.deformation_retract)
print(f"Toplam retraction profili         : {len(rt_profiles)}")
print(f"Bunların deformation retract olanı: {deform_cnt}\n")

for p in rt_profiles:
    dr_tag = "[Deformation]" if p.deformation_retract else "[Retraction only]"
    print(f"  {dr_tag}  {p.display_name}")
    print(f"    X       : {p.total_space}")
    print(f"    A       : {p.retract}")
    print(f"    r(x)    : {p.retraction_formula}")
    print()

print("─" * 65)
print("KONTRAST:")
p_dn = next(p for p in rt_profiles if p.key == "retraction_Dn_to_point")
print(f"  Dⁿ → {{0}}   : deformation retraction VARDIR  →  {p_dn.retraction_formula}")
p_nr = no_retraction_by_dimension(2)
print(f"  Dⁿ → Sⁿ⁻¹  : retraction YOKTUR  →  kanıt: {p_nr.proof_method}")
```

```python
# Örnek 4: Kanıt stratejilerini ve uygulama bağlamlarını karşılaştır
print("=== Kanıt Stratejileri Karşılaştırması ===\n")

print("No-Retraction — Kanıt Yöntemleri:")
for p in get_no_retraction_profiles():
    dim = "Genel" if p.dimension == -1 else str(p.dimension)
    print(f"  n = {dim:5s}: {p.proof_method}")

print()
print("Brouwer FPT — Kanıt Stratejisi Özetleri (ilk 90 karakter):")
for p in get_brouwer_fpt_profiles():
    dim = "Genel" if p.dimension == -1 else str(p.dimension)
    strat = p.proof_strategy[:90] + "..." if len(p.proof_strategy) > 90 else p.proof_strategy
    print(f"  n = {dim:5s}: {strat}")

print()
print("Kanıt Zinciri:")
print("  No-Retraction Teoremi  ──→  Brouwer FPT")
print("  (Dⁿ → Sⁿ⁻¹ yok)              (sabit-noktasız f → retraction → çelişki)")
print()
p_nash = next(p for p in get_brouwer_fpt_profiles() if p.key == "brouwer_fpt_Dn")
print("Nash dengesi bağlantısı:")
print(f"  {p_nash.application_context[:130]}...")
```

## Özet

| Kavram | Ana Sonuç |
|--------|-----------|
| Retraction | $r : X \to A$ sürekli, $r|_A = \mathrm{id}_A$; deformation retract ise $X \simeq A$ |
| No-Retraction Teoremi | $D^n$'den $S^{n-1}$'e retraction yoktur; kanıt boyuta göre değişir |
| Brouwer FPT | Her $f : D^n \to D^n$ sürekli dönüşümün sabit noktası vardır |
| Kanıt zinciri | Sabit-noktasız $f$ → retraction $D^n \to S^{n-1}$ → No-Retraction ile çelişki |
| Kompakt konveks genelleme | $K \subseteq \mathbb{R}^n$ kompakt, konveks → Brouwer FPT $K$ üzerinde geçerli |
| Uygulamalar | Nash dengesi, diferansiyel denklem varlık teoremleri, sıvı dinamiği |

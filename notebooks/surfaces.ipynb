## 1. Konu

### 1.1 Sezgisel Açıklama

Yüzey teorisi, iki boyutlu kompakt manifoldları sınıflandırır. Temel soru: **bu yüzey hangi standart modele homeomorf?**

**Sezgi 1 — Yapıştırma ile oluşturma:**  
Bir kare alın ve kenarlarını belirli kurallara göre yapıştırın:
- $aba^{-1}b^{-1}$ → torus $T^2$ (karşıt kenarlar aynı yönde)
- $abab^{-1}$ → Klein şişesi (bir çift ters yönde)
- $aa$ → projektif düzlem $\mathbb{RP}^2$

**Sezgi 2 — Cins (genus):**  
Yönlendirilebilir yüzeyler için cins = "sap sayısı". Küre $g=0$, torus $g=1$, çift torus $g=2$. Topolojik olarak farklı cinsler homeomorf değildir.

**Sezgi 3 — Yönlendirilebilirlik:**  
Yüzeyin içinde bir Möbius bandı varsa yönlendirilemeyen yüzeydir. Klein şişesi ve $\mathbb{RP}^2$ yönlendirilemeyen kompakt yüzeylerin standart örnekleridir.

**Temel sınıflandırma:**  
Sınır içermeyen bağlantılı kompakt yüzey üçten birinde:
1. **Küre** $S^2$
2. **Bağlantılı torus toplamı** $T^2 \# T^2 \# \cdots$ (cins $g \geq 1$, yönlendirilebilir)
3. **Bağlantılı projektif düzlem toplamı** $\mathbb{RP}^2 \# \cdots$ (cross-cap sayısı $k \geq 1$, yönlendirilemeyen)

**Neden önemli?**
- İlk tam sınıflandırma teoremi (boyut 2): düzgün ifade, tam ispat mümkün
- Euler karakteristiği her kompakt manifoldu sayısal olarak karakterize eder
- 3-manifold teorisinin ve cebirsel topolojinin temel örnek kaynağı

---

### 1.2 Formal Tanımlar

**Yüzey:** Bağlantılı, kompakt, sınırsız (veya sınırlı) 2-manifold.

**Yönlendirilebilirlik:** $X$ yönlendirilebilir $\iff$ $X$'e bir Möbius bandı gömülemiyor $\iff$ her noktanın komşuluğunda tutarlı bir "saat yönü" seçilebiliyor.

**Cins $g$:** Yönlendirilebilir kapalı yüzey için $\chi(X) = 2 - 2g$.  
**Cross-cap sayısı $k$:** Yönlendirilemeyen kapalı yüzey için $\chi(X) = 2 - k$.

**Euler Karakteristiği:** Herhangi bir CW-ayrışımı $\chi = V - E + F$ (köşe - kenar + yüz); topoloji değişmezidir.

**Bağlantılı Toplam $A \# B$:** Her yüzeyde küçük açık disk çıkar, sınır çemberlerini homeomorfizmayla yapıştır.

## 2. Teoremler

### Teorem 1 — Kompakt Yüzeylerin Sınıflandırması

**İfade:**  
Her bağlantılı, kompakt, sınırsız yüzey aşağıdakilerden birine homeomorf:
$$S^2, \quad T_g = \underbrace{T^2 \# \cdots \# T^2}_{g} \;(g \geq 1), \quad N_k = \underbrace{\mathbb{RP}^2 \# \cdots \# \mathbb{RP}^2}_{k} \;(k \geq 1)$$
Ve bu liste kesişimsizdir (hiçbiri diğerine homeomorf değildir).

**İspat taslağı:**  
_(Varlık)_ Her kompakt yüzey bir çokgen yapıştırmasıyla temsil edilebilir. Sistematik kenar çiftleme işlemleriyle $aa^{-1}$ çiftleri iptal edilerek, örtüşen köşeler birleştirilerek standart form elde edilir:
- Yönlendirilebilir: $a_1 b_1 a_1^{-1} b_1^{-1} \cdots a_g b_g a_g^{-1} b_g^{-1}$
- Yönlendirilemeyen: $c_1 c_1 c_2 c_2 \cdots c_k c_k$

_(Teklik)_ Euler karakteristiği $\chi$ ve yönlendirilebilirlik tam sınıflandırıcıdır:
- $S^2$: $\chi = 2$, yönlendirilebilir
- $T_g$: $\chi = 2-2g$, yönlendirilebilir ($g \geq 1$)
- $N_k$: $\chi = 2-k$, yönlendirilemeyen ($k \geq 1$)

Bu üçü ayrık değer kümeleri kapsar. $\square$

---

### Teorem 2 — Euler Karakteristiğinin Topoloji Değişmezi Olması

**İfade:**  
Kompakt yüzey $X$'in herhangi iki CW-ayrışımı aynı $\chi(X) = V - E + F$ değerini verir.

**İspat (kısmi — barisentrik alt bölünme ile):**  
İki ayrışım verilsin; ikisini de refine eden ortak bir barisentrik alt bölünme var.  
Bir kenarı eklediğimizde: $E' = E+1$, ya iki yüzü böldük ($F' = F+1$, $V$ sabit) ya da iki tepe birleşti ($V' = V-1$, $F$ sabit). Her iki durumda $V - E + F$ değişmez.  
Refine işlemi sonlu adımda tamamlanır; $\chi$ her adımda sabit kalır. $\square$

**Özet tablo:**

| Yüzey | $\chi$ | Cins/Cross-cap | Yönlendirilebilir |
|-------|--------|----------------|-------------------|
| $S^2$ | 2 | $g=0$ | ✓ |
| $T^2$ | 0 | $g=1$ | ✓ |
| $T_g$ | $2-2g$ | $g$ | ✓ |
| $\mathbb{RP}^2$ | 1 | $k=1$ | ✗ |
| Klein şişesi | 0 | $k=2$ | ✗ |
| $N_k$ | $2-k$ | $k$ | ✗ |

---

### Teorem 3 — Klein Şişesi $\mathbb{RP}^2 \# \mathbb{RP}^2$'ye Homeomorf

**İfade:** $K^2 \cong \mathbb{RP}^2 \# \mathbb{RP}^2$.

**İspat:**  
Klein şişesinin yapıştırma sözcüğü $abab^{-1}$'dir. Standart yönlendirilemeyen form $c_1 c_1 c_2 c_2$ elde etmek için:
$abab^{-1} \xrightarrow{c_1 = ab} c_1 c_1 b^{-1} a^{-1}$ — uygun köşe özdeşleşmeleriyle tam yeniden yazma $c_1 c_1 c_2 c_2$ formuna girer.  
Bu $N_2 = \mathbb{RP}^2 \# \mathbb{RP}^2$'nin standart formudur. $\square$

## 3. API

```python
from pytop.surfaces import (
    SurfaceProfile,
    SurfaceProfileError,
    surface_profile,
    sphere_surface_profile,
    torus_surface_profile,
    double_torus_surface_profile,
    disk_surface_profile,
    annulus_surface_profile,
    mobius_band_surface_profile,
    projective_plane_surface_profile,
    klein_bottle_surface_profile,
    unknown_surface_profile,
    known_surface_profile,
    as_manifold_profile,
    surface_profile_summary,
    KNOWN_SURFACE_PROFILES,
    SURFACE_ORIENTABILITY_VALUES,
)
print("pytop.surfaces yüklendi.")
print("Kayıtlı anahtarlar:", sorted(set(KNOWN_SURFACE_PROFILES.keys())))
```

### `SurfaceProfile` Alanları ve Özellikleri

| Alan | Tür | Açıklama |
|------|-----|----------|
| `name` | `str` | Profil adı |
| `orientability` | `str` | `"orientable"` \| `"nonorientable"` \| `"unknown"` |
| `genus` | `int\|None` | Yönlendirilebilir yüzey için cins |
| `nonorientable_genus` | `int\|None` | Cross-cap sayısı |
| `boundary_component_count` | `int` | Sınır bileşen sayısı |
| `compact` | `bool\|None` | Kompaktlık |
| `model` | `str` | Sezgisel model açıklaması |
| `gluing_word` | `str` | Yapıştırma sözcüğü |
| `fundamental_group_hint` | `str` | $\pi_1$ ipucu |
| `status` | `str` | `"certified"` \| `"preview"` \| ... |

| Özellik | Açıklama |
|---------|----------|
| `.dimension` | Her zaman 2 |
| `.has_boundary` | `boundary_component_count > 0` |
| `.is_closed_surface` | Kompakt + sınırsız |
| `.is_certified` | `status == "certified"` |
| `.classification_label` | İnsan okunur sınıf etiketi |

**`known_surface_profile(key)`** ile erişilebilen anahtarlar:  
`"sphere"`, `"torus"`, `"double_torus"`, `"disk"`, `"annulus"`, `"mobius_band"`, `"projective_plane"`, `"klein_bottle"`

## 4. Örnekler

### Örnek 1 — Minimal: Tüm Kapalı Yüzeylerin Katalogu

```python
closed_surfaces = [
    sphere_surface_profile(),
    torus_surface_profile(),
    double_torus_surface_profile(),
    projective_plane_surface_profile(),
    klein_bottle_surface_profile(),
]

print(f"{'Ad':30} {'Yön.':14} {'Cins/CC':9} {'χ':5} {'Kapalı'} {'π₁ ipucu'}")
print("-" * 95)
for sp in closed_surfaces:
    euler = sp.metadata.get("euler_characteristic", "?")
    g = sp.genus if sp.genus is not None else f"k={sp.nonorientable_genus}"
    closed_icon = "✓" if sp.is_closed_surface else "✗"
    print(f"  {sp.name:28} {sp.orientability:14} {str(g):9} {str(euler):5} {closed_icon:8} {sp.fundamental_group_hint}")
```

### Örnek 2 — Orta Düzey: Sınırlı Yüzeyler ve Deformasyon Retraktları

```python
# Sınırlı yüzeyler: disk, annulus, Möbius bandı
bounded_surfaces = [
    disk_surface_profile(),
    annulus_surface_profile(),
    mobius_band_surface_profile(),
]

print("=== Sınırlı Yüzeyler ===")
print(f"{'Ad':20} {'Yön.':14} {'Sınır':6} {'Kompakt':8} {'χ':5} {'π₁'}")
print("-" * 70)
for sp in bounded_surfaces:
    euler = sp.metadata.get("euler_characteristic", "?")
    print(f"  {sp.name:18} {sp.orientability:14} {sp.boundary_component_count:6} {str(sp.compact):8} {str(euler):5} {sp.fundamental_group_hint}")

print()

# Deformasyon retraktı bağlantısı
print("Deformasyon retraktları:")
print("  Disk D² → {pt}  (büzülebilir, π₁ = trivial)")
print("  Annulus → S¹    (π₁ = Z, yapıştırma sözcüğü olmadan)")
print("  Möbius bandı → S¹ (sınırı değil orta çemberi, π₁ = Z)")

print()
# Yapıştırma sözcükleri
print("Yapıştırma sözcükleri (yönlendirilemeyen yüzeyler):")
for sp in bounded_surfaces + [klein_bottle_surface_profile(), projective_plane_surface_profile()]:
    if sp.gluing_word:
        print(f"  {sp.name:25} → {sp.gluing_word}")
```

### Örnek 3 — Gelişmiş: Euler Karakteristiği ile Sınıflandırma

```python
# Kapalı yüzeyler için χ ve sınıflandırma etiketi
all_closed = [
    known_surface_profile("sphere"),
    known_surface_profile("torus"),
    known_surface_profile("double_torus"),
    known_surface_profile("projective_plane"),
    known_surface_profile("klein_bottle"),
]

print(f"{'Yüzey':28} {'χ':5} {'Etiket':40} {'Onaylı'}")
print("-" * 85)
for sp in all_closed:
    euler = sp.metadata.get("euler_characteristic", "?")
    print(f"  {sp.name:26} {str(euler):5} {sp.classification_label:40} {sp.is_certified}")

print()

# χ = 2 - 2g (yönlendirilebilir) veya 2 - k (yönlendirilemeyen)
print("χ formülleri:")
for sp in all_closed:
    euler = sp.metadata.get("euler_characteristic", "?")
    if sp.orientability == "orientable" and sp.genus is not None:
        formula = f"2 - 2×{sp.genus} = {2 - 2*sp.genus}"
    elif sp.orientability == "nonorientable" and sp.nonorientable_genus is not None:
        formula = f"2 - {sp.nonorientable_genus} = {2 - sp.nonorientable_genus}"
    else:
        formula = "?"
    print(f"  {sp.name:26} χ = {formula}  (kayıt: {euler})")
```

### Örnek 4 — Gerçekçi Senaryo: `surface_profile_summary` ile Tam Profil

```python
# Torus için tam özet
torus = torus_surface_profile()
summary = surface_profile_summary(torus)

print("=== Torus T² Özeti ===")
for key, val in summary.items():
    print(f"  {key:28}: {val}")

print()

# ManifoldProfile'a dönüşüm
mp = as_manifold_profile(torus)
print(f"=== ManifoldProfile ===")
print(f"  Boyut        : {mp.dimension}")
print(f"  Sınırlı mı?  : {mp.with_boundary}")
print(f"  Yerel model  : {mp.local_model}")
print(f"  Yönlendirme  : {mp.orientability}")

print()

# Bilinmeyen yüzey
mystery = unknown_surface_profile("exotic_surface", reason="No registered model for this space.")
print(f"Bilinmeyen profil: {mystery.name}, status={mystery.status}, uyarı: {mystery.warnings[0]}")
```

### Alıştırma 3 — Karşılaştırma: Sınıflandırma Tam mı?

Sınıflandırma teoremini tabloya dökün: her $(\chi, \text{yönlendirme})$ çifti en fazla bir yüzey tipini belirlemelidir.

Kapalı yüzeyler için aşağıdaki soruları yanıtlayın:
- $\chi = 0$, yönlendirilebilir → ?
- $\chi = 0$, yönlendirilemeyen → ?
- $\chi = -2$, yönlendirilebilir → ?
- $\chi = 1$, yönlendirilemeyen → ?

```python
# χ ve yönlendirme → yüzey
queries = [
    (0,  "orientable"),
    (0,  "nonorientable"),
    (-2, "orientable"),
    (1,  "nonorientable"),
    (2,  "orientable"),
    (-4, "orientable"),
]

# Kayıtlı profillerden χ → profil eşlemesi oluştur
chi_map: dict[tuple[int, str], str] = {}
seen_names = set()
for sp in KNOWN_SURFACE_PROFILES.values():
    if sp.name in seen_names:
        continue
    seen_names.add(sp.name)
    euler = sp.metadata.get("euler_characteristic")
    if euler is not None and sp.is_closed_surface:
        chi_map[(euler, sp.orientability)] = sp.classification_label

print(f"{'χ':5} {'Yönlendirme':16} {'Sınıflandırma'}")
print("-" * 60)
for chi, orient in queries:
    label = chi_map.get((chi, orient), "(kayıt yok — cins hesapla)")
    # Kayıt yoksa formülden hesapla
    if label == "(kayıt yok — cins hesapla)":
        if orient == "orientable":
            g = (2 - chi) // 2
            label = f"orientable genus {g} surface (formülden)"
        else:
            k = 2 - chi
            label = f"nonorientable genus {k} surface (formülden)"
    print(f"  {chi:3}  {orient:16} {label}")
```

### Alıştırma 4 — Euler Karakteristiği Doğrulaması

Hem yönlendirilebilir hem yönlendirilemeyen yüzeyler için Euler karakteristiğini hesaplayıp formülü doğrulayın:
1. Cins $g = 0, 1, 2, 3$ için `surface_profile` ile yönlendirilebilir yüzeyler oluşturun; $\chi = 2 - 2g$ formülünü kontrol edin
2. Cross-cap $k = 1, 2, 3, 4$ için yönlendirilemeyen yüzeyler oluşturun; $\chi = 2 - k$ formülünü kontrol edin
3. Tüm sonuçları tek tabloda gösterin ve hesaplanan ile kayıt edilen χ değerlerinin eşit olduğunu doğrulayın

```python
print(f"{'Tür':16} {'g/k':4} {'χ formül':10} {'χ kayıt':10} {'Eşleşme':9} {'Etiket'}")
print("-" * 80)

# Yönlendirilebilir: g = 0,1,2,3
for g in range(4):
    chi = 2 - 2 * g
    sp = surface_profile(
        f"orientable_g{g}",
        orientability="orientable",
        genus=g,
        compact=True,
        connected=True,
        model=f"genus-{g} orientable surface",
        status="certified",
        metadata={"euler_characteristic": chi},
    )
    chi_rec = sp.metadata.get("euler_characteristic", "?")
    match = "✓" if chi_rec == chi else "✗"
    print(f"  Yönlendir.  {g:4}  {chi:10}  {str(chi_rec):10}  {match:9}  {sp.classification_label[:28]}")

# Yönlendirilemeyen: k = 1,2,3,4
for k in range(1, 5):
    chi = 2 - k
    sp = surface_profile(
        f"nonorientable_k{k}",
        orientability="nonorientable",
        nonorientable_genus=k,
        compact=True,
        connected=True,
        model=f"nonorientable k={k} surface",
        status="certified",
        metadata={"euler_characteristic": chi},
    )
    chi_rec = sp.metadata.get("euler_characteristic", "?")
    match = "✓" if chi_rec == chi else "✗"
    print(f"  Yönlend.yok {k:4}  {chi:10}  {str(chi_rec):10}  {match:9}  {sp.classification_label[:28]}")

print()
all_match = all(
    (2 - 2*g) == (2 - 2*g) for g in range(4)
)  # trivially True; real check is the ✓/✗ above
print("Tüm formül eşleşmeleri yukarıda ✓ görünüyorsa doğrulama başarılı.")
```

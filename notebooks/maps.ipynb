## 1. Konu

### 1.1 Sezgisel Açıklama

Sürekli fonksiyon = **şekli korumadan noktaları hareket ettir**. Daha kesin:

- **Analizde:** $|x - a| < \delta \Rightarrow |f(x) - f(a)| < \varepsilon$ — yakın noktalar yakın gider
- **Topolojide:** Her açık kümenin ön görüntüsü açıktır — bu daha genel ve mesafeye bağlı değil

**Homeomorfizma = topolojik eşdeğerlik:**  
İki uzay $X \cong Y$ homeomorfsa, topolojik bakışta birbirinden ayırt edilemezler. Tüm topolojik özellikler (bağlılık, kompaktlık, Hausdorff, ...) korunur.

**Hiyerarşi:**
```
Homeomorfizma
  ↑  (hem açık hem sürekli bijeksiyon)
Açık gömme
  ↑  (sürekli injektif açık)
Sürekli fonksiyon
```

---

### 1.2 Formal Tanımlar

**Süreklilik:**  
$f: (X, \tau_X) \to (Y, \tau_Y)$ süreklidir $\iff$ her $V \in \tau_Y$ için $f^{-1}(V) \in \tau_X$.

**Nokta Başına Süreklilik:**  
$f$, $x \in X$'te süreklidir $\iff$ $f(x)$'in her $V$ açık komşuluğu için $f^{-1}(V)$, $x$'in açık komşuluğudur.

**Açık Fonksiyon:**  
Her $U \in \tau_X$ için $f(U) \in \tau_Y$.

**Kapalı Fonksiyon:**  
Her kapalı $F \subseteq X$ için $f(F)$ kapalı.

**Gömme (Embedding):**  
$f: X \to Y$ sürekli injektif ve $f: X \to f(X)$ homeomorfizma.

**Bölüm Fonksiyonu (Quotient Map):**  
$f: X \to Y$ sürekli surjektif; $V \subseteq Y$ açık $\iff$ $f^{-1}(V)$ açık.

**Homeomorfizma:**  
$f: X \to Y$ sürekli bijeksiyon ve $f^{-1}: Y \to X$ de sürekli.

---
## 2. Teoremler

### Teorem 1 — Sürekliliğin Dört Karakterizasyonu

**İfade:**  
$f: X \to Y$ için şu dört ifade denktir:
1. $f$ süreklidir
2. Her açık $V \subseteq Y$ için $f^{-1}(V)$ açıktır
3. Her kapalı $F \subseteq Y$ için $f^{-1}(F)$ kapalıdır
4. Her $A \subseteq X$ için $f(\bar{A}) \subseteq \overline{f(A)}$

**İspat ($2 \Leftrightarrow 3$):**  
$f^{-1}(Y \setminus F) = X \setminus f^{-1}(F)$. Tümleyen açık $\iff$ küme kapalı. Dolayısıyla (2) ↔ (3) tümleyenle.

**İspat ($1 \Leftrightarrow 4$):**  
$(1 \Rightarrow 4)$: $x \in \bar{A}$, yani $x$'in her açık $U$ komşuluğu $A$ ile kesişir. $f$ sürekli: $f^{-1}(V)$ açık $f(x)$'in her komşuluğu $V$ için ön görüntü $x$'i içerir. $f^{-1}(V) \cap A \ne \emptyset \Rightarrow V \cap f(A) \ne \emptyset \Rightarrow f(x) \in \overline{f(A)}$.  
$(4 \Rightarrow 1)$: $F \subseteq Y$ kapalı. $A = f^{-1}(F)$ için (4) gereği $f(\bar{A}) \subseteq \overline{f(A)} \subseteq F$. Dolayısıyla $\bar{A} \subseteq f^{-1}(F) = A$, yani $A$ kapalı. $\square$

---

### Teorem 2 — Kompaktan Hausdorff'a Sürekli Bijeksiyon = Homeomorfizma

**İfade:**  
$f: X \to Y$ sürekli bijeksiyon, $X$ kompakt, $Y$ Hausdorff. O zaman $f$ homeomorfizmadır.

**İspat:**  
$f^{-1}$ sürekli olduğunu göstermek için: $f$ kapalı fonksiyon olduğunu göster.  
$F \subseteq X$ kapalı olsun. $X$ kompakt, $F$ kapalı → $F$ kompakt (Teorem: kompaktta kapalı alt küme kompakttır). $f$ sürekli → $f(F)$ kompakt. $Y$ Hausdorff → $f(F)$ kapalıdır (kompakt küme Hausdorff uzayda kapalıdır).  
Dolayısıyla $f$ kapalı fonksiyondur. Kapalı bijeksiyon → $(f^{-1})^{-1}(F) = f(F)$ kapalı → $f^{-1}$ sürekli. $\square$

---

### Teorem 3 — Ara Değer Teoremi

**İfade:**  
$f: X \to \mathbb{R}$ sürekli, $X$ bağlı, $f(a) < c < f(b)$. O zaman $\exists x \in X$ ile $f(x) = c$.

**İspat:**  
Sürekli fonksiyon bağlıyı bağlıya götürür (sürekli görüntü bağlıdır). $X$ bağlı olduğundan $f(X) \subseteq \mathbb{R}$ bağlıdır. $\mathbb{R}$'deki bağlı kümeler aralıklardır. $f(a), f(b) \in f(X)$ ve $f(X)$ bir aralık olduğundan, aralarındaki her değer $c$ de $f(X)$'dedir. $\square$

## 4. API

````python
import itertools
from pytop.spaces import TopologicalSpace
from pytop.maps import FiniteMap, is_continuous_map, map_report, preimage_of_subset, image_of_subset

# is_continuous_map: O(|τ_Y| * n) — ön görüntü testi
X = frozenset({1, 2, 3})
tau = frozenset([frozenset(), frozenset({1}), frozenset({2,3}), X])
sp  = TopologicalSpace(carrier=X, topology=tau)

# f = id: her ön görüntü açık → sürekli
fm_id = FiniteMap(domain=sp, codomain=sp, mapping={1:1, 2:2, 3:3})
r_cont = is_continuous_map(fm_id)
print(f"Kimlik sürekli: {r_cont.is_true}")
print(f"|τ_Y| = {len(sp.topology)} → {len(sp.topology)} ön görüntü testi")

# map_report: 9 özelliği toplu test
report = map_report(fm_id)
print("\nmap_report (kimlik fonksiyonu):")
for k, v in report.items():
    icon = "✓" if v else ("✗" if v is False else "?")
    print(f"  {k:15} {icon}")

# preimage / image: O(n) / O(|A|)
pre = preimage_of_subset(fm_id, frozenset({1, 2}))
img = image_of_subset(fm_id, frozenset({2, 3}))
print(f"\nf⁻¹({{1,2}}) = {sorted(pre)}")
print(f"f({{2,3}})   = {sorted(img)}")
````

### Örnek 2 — Orta Düzey: Sürekli Ama Homeomorfizma Olmayan Bijeksiyon

````python
# Kaynak: ayrık topoloji, Hedef: indirgenmiş topoloji
X = frozenset({1, 2, 3})

disc_tau = frozenset(
    frozenset(s) for r in range(len(X)+1)
    for s in itertools.combinations(X, r)
)
indisc_tau = frozenset([frozenset(), X])

sp_disc  = TopologicalSpace(carrier=X, topology=disc_tau)
sp_indisc = TopologicalSpace(carrier=X, topology=indisc_tau)

# f: ayrık → indirgenmiş, f(x) = x
f = FiniteMap(domain=sp_disc, codomain=sp_indisc, mapping={1:1, 2:2, 3:3})

print("Sürekli:", is_continuous_map(f).is_true)       # True: her şey açık → ön görüntü açık
print("Bijektif:", is_bijective_map(f).is_true)        # True
print("Homeomorfizma:", is_homeomorphism_map(f).is_true)  # False: ters sürekli değil
````

### Örnek 3 — Gelişmiş: Ön Görüntü ve Görüntü

````python
# f: {1,2,3} → {a,b} bölüm fonksiyonu: 1→a, 2→a, 3→b
X_src = frozenset({1, 2, 3})
X_tgt = frozenset({'a', 'b'})

tau_src = frozenset([
    frozenset(), frozenset({1,2}), frozenset({3}), X_src
])
tau_tgt = frozenset([
    frozenset(), frozenset({'a'}), frozenset({'b'}), X_tgt
])

sp_src = TopologicalSpace(carrier=X_src, topology=tau_src)
sp_tgt = TopologicalSpace(carrier=X_tgt, topology=tau_tgt)

f = FiniteMap(
    domain=sp_src, codomain=sp_tgt,
    mapping={1:'a', 2:'a', 3:'b'}
)

# Ön görüntü
pre_a = preimage_of_subset(f, frozenset({'a'}))
print("f⁻¹({a}) =", sorted(pre_a))   # {1, 2}

# Görüntü
img_12 = image_of_subset(f, frozenset({1, 2}))
print("f({1,2}) =", img_12)           # {'a'}

# Sürekli mi? Bölüm fonksiyonu mu?
print("Sürekli:", is_continuous_map(f).is_true)
print("Bölüm:  ", is_quotient_map(f).is_true)
````

Beklenen sonuçlar: sürekli=True, homeomorfizma=False.

````python
# Üç farklı fonksiyonu karşılaştır:
# 1) Kimlik (homeomorfizma)
# 2) Sabit (ne sürekli de olabilir)
# 3) Birebir değil

X = frozenset({1, 2, 3})
tau = frozenset([frozenset(), frozenset({1}), frozenset({2,3}), X])
sp  = TopologicalSpace(carrier=X, topology=tau)

maps = {
    "kimlik":      FiniteMap(domain=sp, codomain=sp, mapping={1:1, 2:2, 3:3}),
    "sabit(1→1)": FiniteMap(domain=sp, codomain=sp, mapping={1:1, 2:1, 3:1}),
}

props = ["continuous", "injective", "surjective", "homeomorphism"]
print(f"{'Özellik':15}", end="")
for name in maps: print(f"  {name:12}", end="")
print()
print("-" * 50)

for prop in props:
    print(f"{prop:15}", end="")
    for fm in maps.values():
        v = map_report(fm).get(prop)
        icon = "✓" if v else "✗"
        print(f"  {icon:12}", end="")
    print()
````

b) Aşağıdaki iki bölüm fonksiyonunu oluşturun ve `map_report` ile karşılaştırın:
   - $q_1$: $\{1,2,3,4\} \to \{a,b\}$, $q_1(1) = q_1(2) = a$, $q_1(3) = q_1(4) = b$ (simetrik bölüm)
   - $q_2$: $\{1,2,3,4\} \to \{a,b,c\}$, $q_2(1) = a$, $q_2(2) = q_2(3) = b$, $q_2(4) = c$ (asimetrik)
   
   Her ikisi sürekli midir? Bölüm fonksiyonu mudur?

c) Kompaktan Hausdorff'a sürekli bijeksiyon = homeomorfizma teoremine dayalı bir örnek kurun. `is_homeomorphism_map` ile doğrulayın.

````python
import itertools
from pytop.spaces import TopologicalSpace
from pytop.maps import FiniteMap, map_report, is_homeomorphism_map
from pytop.topology_builders import discrete_topology

# a) Ayrık hedef: |τ_Y| = 2^n
X4 = frozenset({1,2,3,4})
disc4 = TopologicalSpace(
    carrier=X4,
    topology=frozenset(frozenset(s) for r in range(5)
                       for s in itertools.combinations(X4, r)),
)
print(f"Ayrık topoloji |τ| = {len(disc4.topology)} = 2^4")
print(f"is_continuous testi → {len(disc4.topology)} ön görüntü hesabı")

# b) İki bölüm fonksiyonu
src = TopologicalSpace(
    carrier=X4,
    topology=frozenset([frozenset(), frozenset({1,2}), frozenset({3,4}), X4])
)
tgt2 = TopologicalSpace(
    carrier=frozenset({'a','b'}),
    topology=frozenset([frozenset(), frozenset({'a'}), frozenset({'b'}), frozenset({'a','b'})])
)
q1 = FiniteMap(domain=src, codomain=tgt2, mapping={1:'a',2:'a',3:'b',4:'b'})
rep1 = map_report(q1)
print(f"\nq1 (simetrik): cont={rep1['continuous']}, quotient={rep1['quotient']}, bij={rep1['bijective']}")

# c) Kompakt → Hausdorff bijeksiyon = homeomorfizma
sp_kh = TopologicalSpace(carrier=X4, topology=disc4.topology)  # ayrık = kompakt
fm_bij = FiniteMap(domain=sp_kh, codomain=sp_kh, mapping={1:1,2:2,3:3,4:4})
print(f"\nKimlik (ayrık → ayrık) homeomorfizma: {is_homeomorphism_map(fm_bij).is_true}")
````

### Alıştırma 3 — Karşılaştırma

Aynı iki uzay arasında iki farklı bijeksiyon oluşturun:
- $f: \{1,2,3\} \to \{1,2,3\}$, $f = \text{id}$
- $g: \{1,2,3\} \to \{1,2,3\}$, $g(1)=2, g(2)=1, g(3)=3$

Kaynak ve hedef topoloji: $\tau = \{\emptyset, \{1\}, \{2,3\}, X\}$.  
`map_report` ile karşılaştırın: hangi bijeksiyon homeomorfizma?

````python
X = frozenset({1, 2, 3})
tau = frozenset([frozenset(), frozenset({1}), frozenset({2,3}), X])
sp = TopologicalSpace(carrier=X, topology=tau)

f_id   = FiniteMap(domain=sp, codomain=sp, mapping={1:1, 2:2, 3:3})
f_swap = FiniteMap(domain=sp, codomain=sp, mapping={1:2, 2:1, 3:3})

for name, fm in [("id", f_id), ("swap(1↔2)", f_swap)]:
    rep = map_report(fm)
    print(f"{name}: continuous={rep['continuous']}, homeomorphism={rep['homeomorphism']}")
````

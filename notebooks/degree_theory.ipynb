# Derece Teorisi

**Modüller:** `pytop.degree_theory`, `pytop.degree_theory_applications`  
**Kaynak:** Adams & Franzosa *Introduction to Topology* §9.1–9.4

---

Bir $f : S^1 \to S^1$ sürekli haritası çemberi kaç kez ve hangi yönde sardığını sayan bir tam sayı belirler: bu sayıya **derece** denir.  
Derece, homotopi altında değişmeyen en temel topolojik değişmezlerden biridir ve retraksiyonların yokluğunu, polinomların köklerini, biyolojik döngülerin kararlılığını tek bir kavramla açıklar.

Bu notebook beş bölümden oluşur:
1. **Konu** — Derece tanımı, sarma sayısı, yönelim kavramları
2. **Teoremler** — Homotopi değişmezliği, surjektiflik, retraksiyon yokluğu, cebirin temel teoremi
3. **API** — `pytop.degree_theory` ve `pytop.degree_theory_applications` fonksiyon referansı
4. **Örnekler** — Dört senaryo: çember derecelerinden kalp atışı modeline
5. **Alıştırmalar** — Kodlama ve teori

## 1. Konu

### 1.1 Çember Haritasının Derecesi

**Tanım.** $f : S^1 \to S^1$ sürekli bir harita olsun.  
$S^1 = \{z \in \mathbb{C} : |z| = 1\}$ olarak alınırsa, $f$'nin **derecesi** $\deg(f) \in \mathbb{Z}$ şöyle yorumlanır:  
kaynak çemberi bir kez dolaşıldığında hedef çember $\deg(f)$ kez dolaşılır (negatif değer ters yönü belirtir).

Daha formel olarak, $f$ temel grup düzeyinde $f_* : \pi_1(S^1) \to \pi_1(S^1)$, yani $\mathbb{Z} \to \mathbb{Z}$ indükler; bu homomorfizm $n \mapsto d \cdot n$ biçimindedir. $d = \deg(f)$.

### 1.2 Standart Örnekler

| Harita | Formül | Derece |
|--------|--------|--------|
| Özdeşlik | $\operatorname{id}(z) = z$ | $1$ |
| Sabit | $c(z) = 1$ | $0$ |
| Kuvvet haritası | $p_n(z) = z^n$ | $n$ |
| Karmaşık eşlenik | $\overline{z}$ | $-1$ |
| Bileşke | $f \circ g$ | $\deg(f)\cdot\deg(g)$ |

### 1.3 Yönelim ve Sarma Sayısı

- **Yönelim-koruyan harita:** $\deg(f) > 0$; $f$ döngü yönünü korur.
- **Yönelim-tersleyen harita:** $\deg(f) < 0$; $f$ döngü yönünü çevirir.
- **Sarma sayısı (winding number):** Belirli bir noktanın etrafında bir eğrinin kaç kez döndüğünü sayar; çember derecesinin geometrik yorumudur.

```python
from pytop.degree_theory import (
    get_circle_degree_profiles,
    circle_degree_summary,
    circle_degree_by_value,
    get_retraction_degree_profiles,
    retraction_degree_summary,
    degree_theory_profile_registry,
)
from pytop.degree_theory_applications import (
    get_fundamental_theorem_algebra_profiles,
    get_heartbeat_degree_model_profiles,
    degree_theory_applications_registry,
)

print("=== Çember Derece Profilleri ===")
for p in get_circle_degree_profiles():
    print(f"  {p.display_name}")
    print(f"    Derece             : {p.degree}")
    print(f"    Yönelim davranışı  : {p.orientation_behavior}")
    print(f"    π₁ indükleme       : {p.induced_fundamental_group_map}")
    print()
```

## 2. Teoremler

### Teorem 1 — Homotopi Değişmezliği

> **Teorem.** $f, g : S^1 \to S^1$ homotopik ise $\deg(f) = \deg(g)$.

**Kanıt.** $H : S^1 \times [0,1] \to S^1$ homotopi olsun; $H(\cdot,0) = f$, $H(\cdot,1) = g$.  
Her $t$ için $H(\cdot,t)$'nin temel grup üzerinde indüklediği çarpma $d(t) \in \mathbb{Z}$ tamsayı değerlidir ve $t$'ye göre süreklidir; dolayısıyla sabittir.  
Bu yüzden $\deg(f) = d(0) = d(1) = \deg(g)$. $\square$

**Sonuç:** Derece, haritanın tam şeklini değil yalnızca homotopi sınıfını belirler.

---

### Teorem 2 — Surjektiflik

> **Teorem.** $f : S^1 \to S^1$ sürekli ve $\deg(f) \neq 0$ ise $f$ surjektiftir.

**Kanıt (taslak).** $f$ surjektif olmasaydı bir $y \in S^1$ ile $y \notin \operatorname{Im}(f)$ bulunurdu.  
$S^1 \setminus \{y\}$ büzülebilir olduğundan $f$, sabit haritaya homotopik olurdu; dolayısıyla $\deg(f) = 0$. Çelişki. $\square$

---

### Teorem 3 — Retraksiyon Yokluğu

> **Teorem.** $D^2$'den $S^1$'e sürekli bir retraksiyon yoktur; yani $r : D^2 \to S^1$ ile $r|_{S^1} = \operatorname{id}_{S^1}$ koşulunu sağlayan sürekli $r$ yoktur.

**Kanıt (derece ile).** Böyle bir $r$ var olsaydı $i : S^1 \hookrightarrow D^2$ dahil etme ile  
$r \circ i = \operatorname{id}_{S^1}$, dolayısıyla $\deg(r) \cdot \deg(i) = 1$.  
Ancak $D^2$ büzülebilir olduğundan $D^2 \to S^1$ olan her sürekli haritanın derecesi $0$'dır; $\deg(r) = 0$.  
$0 \cdot \deg(i) = 1$ imkânsız. $\square$

**Sonuç:** Bu kanıt, Brouwer sabit nokta teoreminin derece-teorik temelidir.

---

### Teorem 4 — Cebirin Temel Teoremi (Topolojik Kanıt)

> **Teorem.** Derecesi $\geq 1$ olan her karmaşık polinom en az bir kök taşır.

**Kanıt fikri.** $p(z) = z^n + a_{n-1}z^{n-1} + \cdots + a_0$ olsun.  
$p$ hiç kök taşımasaydı $p / |p| : \mathbb{C} \to S^1$ her yerde tanımlı sürekli bir harita olurdu.  
Büyük $R > 0$ için $p(Re^{i\theta}) \approx R^n e^{in\theta}$ olduğundan sınır çemberi üzerindeki normalize harita derece $n$ taşır.  
Ama bu harita $D_R^2$'nin içine uzar; büzülebilir bir uzaydan $S^1$'e uzayan her harita derece $0$ taşır. Çelişki. $\square$

```python
# Retraksiyon engelini profiller üzerinden göster
print("=== Retraksiyon Derece Profilleri ===")
for p in get_retraction_degree_profiles():
    exists = "VAR" if p.retraction_exists else "YOK"
    print(f"  [{exists}] {p.display_name}")
    print(f"    Ortam uzayı      : {p.ambient_space}")
    print(f"    Sınır uzayı      : {p.boundary_space}")
    print(f"    Derece engeli    : {p.degree_obstruction[:80]}...")
    print()

summary = retraction_degree_summary()
print("Özet:", summary)
```

## 3. API Referansı

### `pytop.degree_theory`

| Fonksiyon / Sınıf | Açıklama | Döndürür |
|-------------------|----------|----------|
| `CircleDegreeProfile` | $S^1 \to S^1$ haritası için derece profili | `dataclass(frozen=True)` |
| `get_circle_degree_profiles()` | Tüm çember derece profillerini döndürür | `tuple[CircleDegreeProfile, ...]` |
| `circle_degree_summary()` | Yönelim davranışına göre gruplar | `dict[str, list[str]]` |
| `circle_degree_by_value(degree)` | Belirli dereceye sahip profilleri filtreler | `tuple[CircleDegreeProfile, ...]` |
| `RetractionDegreeProfile` | Retraksiyon-derece engeli profili | `dataclass(frozen=True)` |
| `get_retraction_degree_profiles()` | Tüm retraksiyon profillerini döndürür | `tuple[RetractionDegreeProfile, ...]` |
| `retraction_degree_summary()` | Varlık bayrağına göre gruplar | `dict[str, list[str]]` |
| `degree_theory_profile_registry()` | Kategori başına profil sayısı | `dict[str, int]` |

### `pytop.degree_theory_applications`

| Fonksiyon / Sınıf | Açıklama | Döndürür |
|-------------------|----------|----------|
| `FundamentalTheoremAlgebraProfile` | Cebirin temel teoremi derece profili | `dataclass(frozen=True)` |
| `get_fundamental_theorem_algebra_profiles()` | FTA profillerini döndürür | `tuple[FundamentalTheoremAlgebraProfile, ...]` |
| `fta_profile_summary()` | FTA profil sayısı özeti | `dict[str, int]` |
| `HeartbeatDegreeModelProfile` | Kalp atışı faz modeli profili | `dataclass(frozen=True)` |
| `get_heartbeat_degree_model_profiles()` | Kalp atışı modellerini döndürür | `tuple[HeartbeatDegreeModelProfile, ...]` |
| `heartbeat_degree_summary()` | Kalp atışı profil sayısı özeti | `dict[str, int]` |
| `degree_theory_applications_registry()` | Tüm uygulama kategorilerinin sayısı | `dict[str, int]` |

```python
# Kayıt defteri özetleri
r1 = degree_theory_profile_registry()
r2 = degree_theory_applications_registry()

print("pytop.degree_theory profil kayıtları:")
for k, v in r1.items():
    print(f"  {k}: {v}")

print("\npytop.degree_theory_applications profil kayıtları:")
for k, v in r2.items():
    print(f"  {k}: {v}")

print(f"\nToplam profil sayısı: {sum(r1.values()) + sum(r2.values())}")
```

## 4. Örnekler

```python
# Örnek 1: Çember derecelerini pozitif / negatif / sıfır olarak ayır
profiles = get_circle_degree_profiles()

pos  = [p for p in profiles if p.degree > 0]
neg  = [p for p in profiles if p.degree < 0]
zero = [p for p in profiles if p.degree == 0]

print("Pozitif dereceli haritalar (yönelim-koruyan):")
for p in pos:
    print(f"  deg={p.degree:+d}  {p.display_name}")

print("\nNegatif dereceli haritalar (yönelim-tersleyen):")
for p in neg:
    print(f"  deg={p.degree:+d}  {p.display_name}")

print("\nSıfır dereceli haritalar (null-homotopik):")
for p in zero:
    print(f"  deg={p.degree:+d}  {p.display_name}")

# circle_degree_by_value kullanımı
deg1 = circle_degree_by_value(1)
print(f"\nDerece=1 profiller ({len(deg1)} adet):")
for p in deg1:
    print(f"  {p.key}  →  {p.homotopy_class}")
```

```python
# Örnek 2: Retraksiyon profillerini derece engeli ile karşılaştır
ret_profiles = get_retraction_degree_profiles()

print("=== Retraksiyon Engeli Analizi ===")
for p in ret_profiles:
    status = "✓ VAR" if p.retraction_exists else "✗ YOK"
    print(f"\n{status}  {p.display_name}")
    print(f"  Ortam  : {p.ambient_space}")
    print(f"  Sınır  : {p.boundary_space}")
    print(f"  Engel  : {p.degree_obstruction}")

# Grupla
summary = retraction_degree_summary()
print("\n--- Gruplama ---")
for group, keys in summary.items():
    print(f"  {group}: {keys}")
```

```python
# Örnek 3: Cebirin Temel Teoremi profillerini göster
fta_profiles = get_fundamental_theorem_algebra_profiles()

print("=== Cebirin Temel Teoremi — Derece Teorisi Kanıtları ===")
for p in fta_profiles:
    print(f"\n{p.display_name}")
    print(f"  Polinom ailesi    : {p.polynomial_family}")
    print(f"  Topolojik kurulum : {p.topological_setup[:90]}...")
    print(f"  Derece argümanı   : {p.degree_argument[:90]}...")
    print(f"  Sonuç             : {p.conclusion}")
    print(f"  Öğretim notu      : {p.teaching_note[:80]}...")
```

```python
# Örnek 4: Kalp atışı faz modeli profilleri
hb_profiles = get_heartbeat_degree_model_profiles()

print("=== Kalp Atışı Derece Modeli ===")
for p in hb_profiles:
    print(f"\n{p.display_name}")
    print(f"  Biyolojik bağlam : {p.biological_context[:80]}...")
    print(f"  Faz uzayı        : {p.phase_space}")
    print(f"  Derece sinyali   : {p.degree_signal[:80]}...")
    print(f"  Yorum            : {p.interpretation[:80]}...")

# Tüm uygulamalar birlikte özet
print("\n--- Tüm uygulama kayıtları ---")
for k, v in degree_theory_applications_registry().items():
    print(f"  {k}: {v}")
```

## Özet

| Kavram | Ana Sonuç |
|--------|-----------|
| Derece | $f : S^1 \to S^1$'in temel grup üzerindeki çarpma sayısı; homotopi altında değişmez |
| Homotopi değişmezliği | $f \simeq g \Rightarrow \deg(f) = \deg(g)$ |
| Surjektiflik | $\deg(f) \neq 0 \Rightarrow f$ surjektif |
| Retraksiyon yokluğu | $D^2 \to S^1$ retraksiyonu imkânsız; derece engeli $0 \neq 1$ |
| Brouwer teoremi | Retraksiyon yokluğunun doğrudan sonucu |
| Cebirin Temel Teoremi | Derece $n$ polinom → sınırda derece $n$ → köksüzlük çelişkisi |
| Kalp atışı modeli | Derece $1$ stabil periyodik ritim; bozulma derece değişmezliğiyle izlenir |

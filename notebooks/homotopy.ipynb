## 1. Konu

### 1.1 Sezgisel Açıklama

Homotopi, iki fonksiyonun **sürekli olarak birbirine dönüştürülebilip dönüştürülemeyeceğini** soran sorudur.

**Sezgi 1 — İp bükme:**  
$f, g: X \to Y$ iki sürekli fonksiyon olsun. Aralarında bir homotopi varsa, $f$'yi sanki bir ip gibi tutup deformasyona uğratarak $g$'ye dönüştürebilirsiniz — hiç koparmadan.

**Sezgi 2 — Temel grup ve sargı sayısı:**  
Dairenin etrafında dolaşan bir yol düşünün. Saat yönünde bir kez dolaşmak, iki kez dolaşmakla homotopik değildir — çünkü halkanın içinden geçemezsiniz. Bu "sargı sayısı" intuisyonu, $\pi_1(S^1) \cong \mathbb{Z}$ sonucunu verir.

**Sezgi 3 — Büzülebilirlik:**  
Bir uzay **büzülebilir** ise, tüm uzayı tek bir noktaya sürekli olarak çekebilirsiniz. $[0,1]$ büzülebilir: $H(x,t) = (1-t)x$ ile her noktayı $0$'a götürürsünüz. $S^1$ büzülebilir değildir: delik geçişini engeller.

**Neden önemli?**
- Uzayları homotopi eşdeğerliğine göre sınıflandırır (homeomorfizmden zayıf)
- Temel grup, uzayın "deliklerini" saymak için temel araçtır
- Sabit nokta teoremlerine (Brouwer, Lefschetz) zemin hazırlar
- Bağlama, demet teorisi ve örtü uzaylarının altyapısını oluşturur

---

### 1.2 Formal Tanımlar

**Homotopi:**  
$f, g: X \to Y$ sürekli. $f$ ile $g$ arasında bir **homotopi**, $H: X \times [0,1] \to Y$ sürekli fonksiyonudur; $H(x,0) = f(x)$ ve $H(x,1) = g(x)$ her $x \in X$ için. Gösterim: $f \simeq g$.

**Göreli Homotopi:**  
$A \subseteq X$ için $f \simeq g$ **$A$'ya göreli** (rel $A$) ise $H(a,t) = f(a) = g(a)$ her $a \in A$, $t \in [0,1]$ için sabit.

**Yol:**  
$\gamma: [0,1] \to X$ sürekli; $\gamma(0)$ başlangıç, $\gamma(1)$ bitiş noktası.

**Yol homotopisi:**  
$\gamma_0 \simeq \gamma_1$ rel $\{0,1\}$: uç noktaları sabit tutarak bükülme.

**Temel Grup $\pi_1(X, x_0)$:**  
$x_0 \in X$ tabanlı kapalı yolların (ilmeklerin) yol homotopi sınıflarının kümesi; birleştirme ile grup oluşturur.

**Deformasyon Retraksiyonu:**  
$r: X \to A$ ($A \subseteq X$) bir **deformasyon retraksiyonu**dur ancak ve ancak $r \circ i = \mathrm{id}_A$ ve $i \circ r \simeq \mathrm{id}_X$ (güçlü versiyonda rel $A$). Bu durumda $A$ bir **deformasyon retrakt**ıdır ve $X \simeq A$.

**Büzülebilirlik:**  
$X$ büzülebilir $\iff$ $X \simeq \{*\}$ (nokta uzayıyla homotopi eşdeğer)  
$\iff$ $\mathrm{id}_X \simeq$ sabit fonksiyon.

## 2. Teoremler

### Teorem 1 — Büzülebilir Uzayın Temel Grubu Trivialdir

**İfade:**  
$X$ büzülebilir $\Rightarrow$ $\pi_1(X, x_0) = \{e\}$ her $x_0 \in X$ için.

**İspat:**  
$X$ büzülebilir olduğundan bir $H: X \times [0,1] \to X$ mevcuttur; $H(x,0) = x$, $H(x,1) = x_0$ her $x$ için.  
$\gamma$ herhangi bir $x_0$ tabanlı ilmek olsun. Homotopi $K: [0,1] \times [0,1] \to X$:  
$$K(s,t) = H(\gamma(s), t)$$  
$K(s,0) = \gamma(s)$, $K(s,1) = x_0$, $K(0,t) = H(x_0,t)$, $K(1,t) = H(x_0,t)$.  
Bu sabit yol değil; uç noktaları sabit tutan bir homotopi değil. Fakat sabit yola homotopik olduğunu göstermek için şunu kullan:  
Büzülme sırasında $H(x_0, t) = c(t)$ bir yoldur; $c(0) = x_0$, $c(1) = x_0$ olması için $H(x_0, 0) = x_0$, $H(x_0, 1) = x_0$ gerekir — büzülmenin $x_0$'a olması bu koşulu sağlar.  
Standart dikdörtgen argümanıyla her $\gamma \simeq$ sabit yol rel $\{0,1\}$, dolayısıyla $[\gamma] = [e]$. $\square$

---

### Teorem 2 — $\pi_1(S^1) \cong \mathbb{Z}$

**İfade:**  
$S^1 = \{z \in \mathbb{C} : |z| = 1\}$'in temel grubu sonsuz devirlidir.

**İspat özeti (örtü uzayı argümanı):**  
Evrensel örtü $p: \mathbb{R} \to S^1$, $p(t) = e^{2\pi i t}$'dir. $\mathbb{R}$ büzülebilirdir, dolayısıyla $\pi_1(\mathbb{R}) = 0$.  
Uzun tam dizi (fiber $\mathbb{Z}$, $\pi_1(\mathbb{R})=0$):
$$\cdots \to \pi_1(\mathbb{R}) \to \pi_1(S^1) \xrightarrow{\partial} \pi_0(\mathbb{Z}) \to \pi_0(\mathbb{R}) \to \cdots$$
$\pi_0(\mathbb{Z}) = \mathbb{Z}$ (ayrık fiber), $\pi_0(\mathbb{R}) = 0$.  
Dolayısıyla $\partial: \pi_1(S^1) \xrightarrow{\sim} \mathbb{Z}$ izomorfizmi elde edilir.  
İzomorfizm **sargı sayısı** ile özdeştir: $[\gamma] \mapsto$ $\gamma$'nın $S^1$'i kaç kez sardığı. $\square$

---

### Teorem 3 — Çiçek Grafiğinin Temel Grubu Serbest Gruptur

**İfade:**  
$S^1 \vee S^1$ (iki dairenin takoz noktasında birleşimi) için $\pi_1(S^1 \vee S^1) \cong F_2$ (rütbe 2 serbest grup).

**İspat özeti (Seifert–van Kampen):**  
$U_1, U_2$: her biri bir daireyi içeren, takoz noktasını kapsayan açık. $U_1 \cap U_2$ büzülebilir (trivial $\pi_1$).  
$\pi_1(U_1) = \mathbb{Z}$, $\pi_1(U_2) = \mathbb{Z}$, $\pi_1(U_1 \cap U_2) = 0$.  
Seifert–van Kampen:
$$\pi_1(S^1 \vee S^1) \cong \mathbb{Z} *_{\{e\}} \mathbb{Z} = \mathbb{Z} * \mathbb{Z} = F_2$$
Üretici $a$ (ilk daire ilmeği), $b$ (ikinci daire ilmeği); ilişki yok. $\square$

## 3. API

```python
from pytop.homotopy import (
    HomotopyProfile,
    DeformationRetractionProfile,
    ContractibleProfile,
    homotopy_profile,
    homotopic,
    not_certified_homotopy,
    unknown_homotopy,
    deformation_retraction_profile,
    contractible_profile,
    known_contractible_profile,
    homotopy_summary,
    deformation_retraction_summary,
    contractible_summary,
    KNOWN_CONTRACTIBLE_EXAMPLES,
)
from pytop.fundamental_group import (
    FundamentalGroupProfile,
    fundamental_group_profile,
    trivial_group_profile,
    infinite_cyclic_group_profile,
    free_group_profile,
    unknown_fundamental_group_profile,
    known_fundamental_group_profile,
    fundamental_group_summary,
    KNOWN_FUNDAMENTAL_GROUP_MODELS,
)
print("pytop.homotopy ve pytop.fundamental_group yüklendi.")
```

### `pytop.homotopy` — Temel Yapılar

| Fonksiyon / Sınıf | Açıklama |
|---|---|
| `homotopic(source, target, ...)` | Onaylanmış homotopi profili oluşturur |
| `not_certified_homotopy(source, target, ...)` | Homotopi onaylanmamış; eksik sertifika |
| `unknown_homotopy(source, target, ...)` | Homotopi durumu bilinmiyor |
| `deformation_retraction_profile(name, space, subspace, ...)` | Deformasyon retraksiyonu profili |
| `contractible_profile(space, ...)` | Büzülebilirlik profili |
| `known_contractible_profile(key)` | Standart büzülebilir uzay kaydından profil al |
| `homotopy_summary(profile)` | Profili dict olarak özetle |

**Homotopi statüsleri:** `"homotopic"` | `"not_certified"` | `"unknown"`

### `pytop.fundamental_group` — Temel Yapılar

| Fonksiyon / Sınıf | Açıklama |
|---|---|
| `trivial_group_profile(space, ...)` | $\pi_1 = \{e\}$ profili |
| `infinite_cyclic_group_profile(space, ...)` | $\pi_1 \cong \mathbb{Z}$ profili |
| `free_group_profile(space, generators=..., ...)` | Serbest grup profili |
| `unknown_fundamental_group_profile(space, ...)` | Bilinmeyen $\pi_1$ profili |
| `known_fundamental_group_profile(key, ...)` | Kayıtlı standart modelden profil al |
| `fundamental_group_summary(profile)` | Profili dict olarak özetle |

**Grup türleri:** `"trivial"` | `"infinite_cyclic"` | `"free"` | `"profile"`

**Kayıtlı modeller (`KNOWN_FUNDAMENTAL_GROUP_MODELS`):**

```python
for key, profile in KNOWN_FUNDAMENTAL_GROUP_MODELS.items():
    print(f"  {key:25}  kind={profile.kind:16}  pres={profile.presentation}")
```

## 4. Örnekler

### Örnek 1 — Minimal: $S^1$'in Temel Grubu

```python
# S^1 -> pi_1 = Z (sonsuz devirli)
circle = known_fundamental_group_profile("circle")
s = fundamental_group_summary(circle)

print(f"Uzay       : {s['space']}")
print(f"Tür        : {s['kind']}")
print(f"Statü      : {s['status']}")
print(f"Üreteç     : {s['generators']}")
print(f"Sunum      : {s['presentation']}")
print(f"Sertifika  : {s['certification']}")
print()
print("Onaylı mı?:", circle.is_certified)
print("Trivial mi?:", circle.is_trivial)
```

### Örnek 2 — Orta Düzey: Büzülebilirlik ve Homotopi Profilleri

```python
# Kayıtlı büzülebilir uzaylar
print("=== Büzülebilir Uzay Kaydı ===")
for key in KNOWN_CONTRACTIBLE_EXAMPLES:
    cp = known_contractible_profile(key)
    s = contractible_summary(cp)
    print(f"  {key:20} onaylı={s['status']=='certified'}  tanık='{cp.witness}'")

print()

# Kapalı aralığın büzülebilirlik profili
interval_cp = known_contractible_profile("closed_interval")
print("[0,1] büzülebilir:", interval_cp.is_certified_contractible)

# İki fonksiyon arasında homotopi profili
print()
h = homotopic(
    source="identity map on [0,1]",
    target="constant map to 0",
    witness="H(x,t) = (1-t)*x",
    certification="straight-line homotopy",
)
s = homotopy_summary(h)
print(f"Homotopi adı  : {s['name']}")
print(f"Statü         : {s['status']}")
print(f"Tanık var mı? : {s['has_witness']}")
print(f"Onaylı mı?    : {h.is_certified_homotopy}")
```

### Örnek 3 — Gelişmiş: Deformasyon Retraksiyonu ve $\pi_1$

```python
# Halka (annulus) -> S^1'e deformasyon retraksiyonu
annulus_ret = deformation_retraction_profile(
    name="annulus_to_circle",
    space="annulus A = {x in R^2 : 1 <= |x| <= 2}",
    subspace="S^1 (unit circle)",
    status="certified",
    strong=True,
    certification="radial-projection",
    notes=("r(x) = x / |x|, H(x,t) = x / (1-t+t*|x|)",),
)
s = deformation_retraction_summary(annulus_ret)
print("Deformasyon retraksiyonu:")
for k, v in s.items():
    print(f"  {k:14}: {v}")

print()

# Annulus, S^1'e homotopi eşdeğer -> aynı pi_1
# pi_1(annulus) = pi_1(S^1) = Z
annulus_pi1 = infinite_cyclic_group_profile(
    space="annulus A",
    basepoint="(1,0)",
    generator="loop",
    certification="deformation-retract-to-S1",
    notes=("Annulus deformation retracts onto S^1, so pi_1 is the same.",),
)
print(f"pi_1(Annulus) = {annulus_pi1.presentation}  (onaylı: {annulus_pi1.is_certified})")
```

### Örnek 4 — Gerçekçi Senaryo: Seifert–van Kampen ile $\pi_1$ Hesabı

```python
# S^1 v S^1: iki dairenin takozda birleşimi
wedge = known_fundamental_group_profile("wedge_of_two_circles")
s = fundamental_group_summary(wedge)
print("=== S^1 v S^1 ===")
print(f"Uzay        : {s['space']}")
print(f"Tür         : {s['kind']}")
print(f"Rütbe       : {s['rank']}")
print(f"Üreteçler   : {s['generators']}")
print(f"Sunum       : {s['presentation']}")
print()

# Karşılaştırma: büzülebilir (trivial pi_1) vs dairesel (Z) vs çiçek (F_2)
scenarios = [
    known_fundamental_group_profile("point"),
    known_fundamental_group_profile("closed_interval"),
    known_fundamental_group_profile("circle"),
    known_fundamental_group_profile("wedge_of_two_circles"),
]

print(f"{'Uzay':30} {'Tür':18} {'Sunum':25} {'Onaylı'}")
print("-" * 85)
for p in scenarios:
    print(f"{str(p.space):30} {p.kind:18} {p.presentation:25} {p.is_certified}")

print()

# Kayıt dışı uzay: bilinmeyen pi_1
mystery = unknown_fundamental_group_profile(
    "some exotic 3-manifold",
    reason="No computation supplied; fundamental group can be arbitrarily complex.",
)
print(f"Bilinmeyen: {mystery.space} -> status={mystery.status}")
```

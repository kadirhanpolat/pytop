## 1. Konu

### 1.1 Sezgisel Açıklama

Kompaktlık, analizin **"sonsuzlukla başa çıkma"** aracıdır. İki temel sezgi:

**Sezgi 1 — Sonlu indirgenme:**  
Sonsuz sayıda açık kümeyle kaplasanız bile, sonlu bir alt kaplama yeterlidir. Bu, sonsuz argümanları sonluya indirgemenizi sağlar.

**Sezgi 2 — "Kaçma yok":**  
$[0,1]$ kompakttır çünkü bu aralıktaki noktalar "kaçıp gidemez" — sonsuz bir dizi her zaman yakınsak bir alt dizi içerir (Bolzano-Weierstrass). $(0,1)$ ise kompakt değildir: $\{1/n\}$ dizisi uç noktaya yaklaşır ama $(0,1)$'de bir limiti yoktur.

**Neden önemli?**
- Kompakt uzayda sürekli fonksiyon maksimum ve minimum alır (ekstremal değer teoremi)
- Kompakt uzayda her Cauchy dizisi yakınsaktır
- Kompakt Hausdorff uzayı en iyi davranan uzaydır: normal, ikinci sayılabilir → metrizlenebilir

---

### 1.2 Formal Tanımlar

**Kompakt Uzay:**  
$(X, \tau)$ kompakttır $\iff$ $X$'in her açık kaplama $\mathcal{U} = \{U_\alpha\}$ (yani $X = \bigcup U_\alpha$) sonlu bir alt kaplama içerir.

**Sayılabilir Kompaktlık:**  
$X$ sayılabilir kompakttır $\iff$ her sayılabilir açık kaplama sonlu bir alt kaplama içerir.

**Sıralı Kompaktlık:**  
$X$ sıralı kompakttır $\iff$ $X$'deki her dizi yakınsak bir alt dizi içerir.

**Lindelöf Özelliği:**  
$X$ Lindelöf'tür $\iff$ her açık kaplama sayılabilir bir alt kaplama içerir.

**Limit Nokta Kompaktlığı:**  
$X$ limit nokta kompakttır $\iff$ $X$'in her sonsuz alt kümesi bir limit noktasına sahiptir.

**İlişkiler (metrik uzaylarda):**
$$\text{kompakt} \Rightarrow \text{sıralı kompakt} \Rightarrow \text{sayılabilir kompakt} \Rightarrow \text{limit nokta kompakt}$$
Metrik uzaylarda tüm bu kavramlar denktir.

## 2. Teoremler

### Teorem 1 — Heine-Borel

**İfade:**  
$\mathbb{R}^n$'de bir $K \subseteq \mathbb{R}^n$ kümesi kompakttır ancak ve ancak kapalı ve sınırlıdır.

**İspat (⇒ Kompakt → Kapalı ve Sınırlı):**

_Kapalılık:_ $\mathbb{R}^n$ Hausdorff'tur. Hausdorff uzayda her kompakt küme kapalıdır. (Kanıt: $x \notin K$ için, Hausdorff özelliğiyle her $k \in K$'yi $x$'den ayıran $U_k, V_k$ açıkları bulunur. $\{U_k\}$ $K$'nın açık kapsamasıdır; sonlu alt kaplama $\{U_{k_1},\ldots,U_{k_m}\}$ var. $V = \bigcap V_{k_i}$ açıktır, $x \in V$, $V \cap K = \emptyset$. Dolayısıyla $K^c$ açıktır.)

_Sınırlılık:_ $\{B(0,n)\}_{n=1}^\infty$ $\mathbb{R}^n$'nin açık kapsamasıdır. Kompaktlıktan sonlu alt kaplama: $K \subseteq B(0,N)$ için bazı $N$. Dolayısıyla $K$ sınırlıdır.

**İspat (⇐ Kapalı ve Sınırlı → Kompakt):**  
$K$ sınırlı olduğundan bazı $[-M,M]^n$ hiperkübüne gömülüdür. $[-M,M]^n$ kompakttır (tek boyutlu $[a,b]$'nin Tychonoff ile çarpımı). $K$ bu kompakt kümenin kapalı alt kümesidir, dolayısıyla kompakttır. $\square$

---

### Teorem 2 — Ekstremal Değer Teoremi

**İfade:**  
$f: X \to \mathbb{R}$ sürekli, $X$ kompakt. O zaman $f$ maksimum ve minimum değerini alır.

**İspat:**  
_$f(X)$ sınırlı ve kapalıdır:_ Sürekli fonksiyon kompaktı kompakta götürür. $f(X) \subseteq \mathbb{R}$ kompakttır. Heine-Borel gereği kapalı ve sınırlıdır.

_Süpremum alınır:_ $s = \sup f(X)$ var ($f(X)$ sınırlı). $f(X)$ kapalı olduğundan $s \in f(X)$, yani $\exists x_0 \in X$ ile $f(x_0) = s$. $\square$

---

### Teorem 3 — Kompakt Uzayda Kapalı Alt Küme Kompakttır

**İfade:**  
$X$ kompakt, $F \subseteq X$ kapalı $\Rightarrow$ $F$ kompakttır.

**İspat:**  
$\{U_\alpha\}$ $F$'nin açık kapsaması olsun. $F$ kapalı olduğundan $F^c = X \setminus F$ açıktır. $\{U_\alpha\} \cup \{F^c\}$ $X$'in açık kapsamasıdır. $X$ kompakt olduğundan sonlu alt kaplama $\{U_{\alpha_1},\ldots,U_{\alpha_m}, F^c\}$ var. $F^c \cap F = \emptyset$ olduğundan $\{U_{\alpha_1},\ldots,U_{\alpha_m}\}$ $F$'nin sonlu kapsamasıdır. $\square$

## 3. API

````python
from pytop.spaces import TopologicalSpace
from pytop.compactness import (
    is_compact,
    is_countably_compact,
    is_sequentially_compact,
    is_limit_point_compact,
    is_lindelof,
    analyze_compactness,  # genel yüklem: property adını string olarak al
)
print("pytop.compactness yüklendi.")
````

### `analyze_compactness(space, property_name)` ile Sorgu

`property_name` seçenekleri:
```
"compact"             # tam kompaktlık
"countably_compact"   # sayılabilir kompaktlık
"sequentially_compact"# sıralı kompaktlık
"limit_point_compact" # limit nokta kompaktlık
"lindelof"            # Lindelöf özelliği
```

**Sembolik uzay tagları:**  
Sembolik uzaylarda kompaktlık, taglarla belirlenir:

| Tag | Anlamı |
|-----|--------|
| `"compact"` | Kompakt |
| `"lindelof"` | Lindelöf |
| `"second_countable"` | İkinci sayılabilir → Lindelöf |
| `"sequentially_compact"` | Sıralı kompakt |
| `"not_compact"` | Kompakt değil |

## 4. Örnekler

### Örnek 1 — Minimal: Kompakt Hausdorff Uzay

````python
# Kompakt Hausdorff uzay — sembolik
sp_kh = TopologicalSpace.symbolic(
    description="Kompakt Hausdorff uzay (örn. [0,1])",
    tags=["compact", "hausdorff"]
)

r = is_compact(sp_kh)
print("Kompakt:", r.is_true)
print("Gerekçe:", r.justification[0])
````

### Örnek 2 — Orta Düzey: İkinci Sayılabilir → Lindelöf Zinciri

````python
# İkinci sayılabilir uzay (örn. R ile standart topoloji)
sp_2c = TopologicalSpace.symbolic(
    description="İkinci sayılabilir uzay (örn. R)",
    tags=["second_countable"]
)

for prop in ["compact", "lindelof", "sequentially_compact"]:
    r = analyze_compactness(sp_2c, prop)
    icon = "✓" if r.is_true else ("✗" if r.is_false else "?")
    print(f"  {prop:22} {icon}  {r.justification[0][:65]}")
````

### Örnek 3 — Gelişmiş: Kompakt Olmayan Açık Aralık

````python
# (0,1) — kapalı değil, dolayısıyla kompakt değil
sp_open_interval = TopologicalSpace.symbolic(
    description="Açık aralık (0,1) ⊂ R",
    tags=["not_compact", "hausdorff", "second_countable"]
)

r_compact  = is_compact(sp_open_interval)
r_lindelof = is_lindelof(sp_open_interval)

print("(0,1) kompakt:", r_compact.is_false, "(false bekleniyor)")
print("Gerekçe:", r_compact.justification[0])
print()
print("(0,1) Lindelöf:", r_lindelof.is_true)
print("Gerekçe:", r_lindelof.justification[0])
````

### Örnek 4 — Gerçekçi Senaryo: Sonlu Uzay

````python
# Her sonlu topolojik uzay kompakttır
import itertools

X = frozenset({1, 2, 3})
all_subsets = frozenset(
    frozenset(s) for r in range(len(X)+1)
    for s in itertools.combinations(X, r)
)
finite_sp = TopologicalSpace(carrier=X, topology=all_subsets,
                              metadata={"name": "3 noktalı ayrık uzay"})

for prop in ["compact", "lindelof", "countably_compact", "sequentially_compact"]:
    r = analyze_compactness(finite_sp, prop)
    icon = "✓" if r.is_true else ("✗" if r.is_false else "?")
    print(f"  {prop:22} {icon}  {r.justification[0][:65]}")
````

### Alıştırma 3 — Karşılaştırma

Metrik uzaylarda `compact` ↔ `sequentially_compact` ↔ `countably_compact` hepsi denktir. Bunu aşağıdaki üç uzayda test edin ve sonuçları karşılaştırın:

1. `tags=["compact", "metrizable"]`

3. `tags=["countably_compact", "metrizable"]`

````python
scenarios = [
    ("compact+metrizable",             ["compact", "metrizable"]),
    ("seq_compact+metrizable",         ["sequentially_compact", "metrizable"]),
    ("countably_compact+metrizable",   ["countably_compact", "metrizable"]),
]

props = ["compact", "sequentially_compact", "countably_compact"]
for label, tags in scenarios:
    sp = TopologicalSpace.symbolic(description=label, tags=tags)
    results = {p: analyze_compactness(sp, p).is_true for p in props}
    print(f"{label}: {results}")
````

## 1. Konu

### 1.1 Sezgisel Açıklama

Örtü uzayı, bir uzayı **aynı yerel görünümle ama farklı global yapıyla** temsil eder.

**Sezgi 1 — Sarmalayan şerit:**  
$p: \mathbb{R} \to S^1$, $p(t) = e^{2\pi i t}$ düşünün. Gerçek doğru daireyi sonsuz kez örter; her küçük yay $S^1$'de $\mathbb{R}$'nin yerel bir parçasına homeomorftur. "Örtü" kelimesi bu fikirden gelir: bir uzayın üzerinde yerel homeomorf ama global farklı bir katman.

**Sezgi 2 — Yaprak sayısı:**  
$z \mapsto z^n$ ile $S^1 \to S^1$: her noktanın tam $n$ ön görüntüsü var. Bu $n$-yapraklı örtüdür.

**Sezgi 3 — Kaldırma:**  
$B$'de bir yol $\gamma$ varsa ve bir başlangıç noktası $\tilde{x}_0 \in p^{-1}(\gamma(0))$ seçilirse, $\gamma$'yı üst uzaya **tekil** biçimde kaldırabilirsiniz. Bu kaldırma, $\pi_1$'i hesaplamak için temel araçtır.

**$\pi_1$ ile bağlantı:**  
$p: \tilde{X} \to X$ evrensel örtü ($\tilde{X}$ basitçe bağlı) ise:
$$\pi_1(X, x_0) \cong \text{Deck}(\tilde{X}/X)$$
Deste dönüşüm grubu ($\tilde{X}$'i kendine götüren örtü izomorfizmaları) temel grupla izomorftır.

**Neden önemli?**
- $\pi_1(S^1) \cong \mathbb{Z}$'nin standart ispatı örtü uzayı kullanır
- Her bağlantı bileşenindeki örtüler $\pi_1$'in altgruplarıyla birebir karşılıklı
- Galois teorisiyle derin analoji: örtüler ↔ alan genişlemeleri

---

### 1.2 Formal Tanımlar

**Örtü Haritası:**  
$p: \tilde{X} \to X$ sürekli, örtü (evenly covered) özelliğiyle:  
Her $x \in X$ için $\exists$ açık $U \ni x$: $p^{-1}(U) = \bigsqcup_\alpha V_\alpha$ (ayrık açıklar) ve $p|_{V_\alpha}: V_\alpha \xrightarrow{\sim} U$ homeomorfizma.

**Yaprak Sayısı:** $|p^{-1}(x)|$ — path-connected bazda sabittir.

**Evrensel Örtü:** $\tilde{X}$ basitçe bağlı ($\pi_1(\tilde{X}) = 0$) örtü uzayı.

**Deste Dönüşümü (Deck Transformation):**  
$\phi: \tilde{X} \to \tilde{X}$ homeomorfizma; $p \circ \phi = p$.

**Normal (Regular) Örtü:**  
$p_*(\pi_1(\tilde{X}))$ $\pi_1(X)$'in normal altgrubu ise. Evrensel örtü her zaman normaldir.

## 2. Teoremler

### Teorem 1 — Yol Kaldırma Teoremi

**İfade:**  
$p: \tilde{X} \to X$ örtü, $\gamma: [0,1] \to X$ yol, $\tilde{x}_0 \in p^{-1}(\gamma(0))$.  
O zaman $\exists!$ yol $\tilde{\gamma}: [0,1] \to \tilde{X}$: $\tilde{\gamma}(0) = \tilde{x}_0$ ve $p \circ \tilde{\gamma} = \gamma$.

**İspat:**  
$[0,1]$ kompakt, $\gamma^{-1}(U_i)$ örtü özelliğindeki $\{U_i\}$'nin ön görüntüleri açık bir kaplama oluşturur. Lebesgue lemmasıyla sonlu bölüm $0 = t_0 < t_1 < \cdots < t_n = 1$ seçilir; her $\gamma([t_{k-1}, t_k]) \subseteq U_{i_k}$ için.  
Adım adım: $\tilde{\gamma}|_{[0,t_1]}$ — $p^{-1}(U_{i_1})$'ün ilgili yaprağı $V_{\alpha_1}$'da $(p|_{V_{\alpha_1}})^{-1} \circ \gamma$ ile kaldır. $\tilde{\gamma}(t_1)$ belirlendi; $\tilde{\gamma}|_{[t_1,t_2]}$ benzer şekilde.  
Teklik: her adımda başlangıç noktası ön görüntüsünde sabitlenmiş tek yol var. $\square$

---

### Teorem 2 — Evrensel Örtünün Temel Grubu

**İfade:**  
$p: \tilde{X} \to X$ evrensel örtü, $X$ yol-bağlı. O zaman $\text{Deck}(\tilde{X}/X) \cong \pi_1(X, x_0)$.

**İspat taslağı:**  
$[\gamma] \in \pi_1(X, x_0)$ için kaldırma $\tilde{\gamma}$ elde edilir; $\phi_{[\gamma]}: \tilde{x}_0 \mapsto \tilde{\gamma}(1)$.  
$\tilde{X}$ basitçe bağlı olduğundan $\phi_{[\gamma]}$ deste dönüşümünü tekil belirler (kaldırma tekliğiyle).  
Eşleşme $[\gamma] \mapsto \phi_{[\gamma]}$ bir grup izomorfizmasıdır:  
- Gruba saygı: $\phi_{[\gamma * \delta]} = \phi_{[\delta]} \circ \phi_{[\gamma]}$
- İnjektif: $\phi_{[\gamma]} = \mathrm{id} \Rightarrow \tilde{\gamma}$ kapalı $\Rightarrow [\gamma] = 0$ ($\tilde{X}$ basitçe bağlı)
- Surjektif: her deste dönüşümü bir ilmek kaldırmasından gelir. $\square$

---

### Teorem 3 — $\mathbb{R} \to S^1$ Evrensel Örtüdür

**İfade:** $p: \mathbb{R} \to S^1$, $p(t) = e^{2\pi i t}$, evrensel örtüdür.

**İspat:**  
_(Örtü)_ Her $z \in S^1$ için $U = S^1 \setminus \{-z\}$ (açık yay). $p^{-1}(U) = \bigsqcup_{n \in \mathbb{Z}} (n + (-1/2, 1/2))$: ayrık açık aralıklar; her biri $p$ altında $U$'ya homeomorf.  
_(Basitçe bağlı)_ $\mathbb{R}$ büzülebilir, dolayısıyla $\pi_1(\mathbb{R}) = 0$.  
_(Yol-bağlı)_ $\mathbb{R}$ yol-bağlı.  
Örtü + basitçe bağlı + yol-bağlı ⇒ evrensel örtü. $\square$  

**Sonuç:** Teorem 2'yi uygula: $\pi_1(S^1) \cong \text{Deck}(\mathbb{R}/S^1) \cong \mathbb{Z}$ (deste dönüşümleri $t \mapsto t+n$, $n \in \mathbb{Z}$).

## 3. API

```python
from pytop.covering_spaces import (
    CoveringMapProfile,
    CoveringSpaceProfileError,
    covering_map_profile,
    assumed_covering_map_profile,
    unknown_covering_map_profile,
    circle_degree_covering_profile,
    known_covering_profile,
    covering_profile_summary,
    KNOWN_COVERING_MODELS,
    COVERING_PROFILE_STATUSES,
)
print("pytop.covering_spaces yüklendi.")
print("Statü değerleri:", sorted(COVERING_PROFILE_STATUSES))
print("Kayıtlı modeller:", list(KNOWN_COVERING_MODELS.keys()))
```

### `CoveringMapProfile` Alanları

| Alan | Tür | Açıklama |
|------|-----|----------|
| `name` | `str` | Profil adı |
| `total_space` | `Any` | Örtü uzayı $\tilde{X}$ |
| `base_space` | `Any` | Baz uzay $X$ |
| `sheet_count` | `int\|str\|None` | Yaprak sayısı |
| `status` | `str` | `"certified"` \| `"assumed"` \| `"not_certified"` \| `"unknown"` |
| `covering_map` | `str` | Örtü haritası açıklaması |
| `local_homeomorphism_assumption` | `str` | Yerel homeomorfizm durumu |
| `fundamental_group_note` | `str` | $\pi_1$ ile bağlantı notu |

| Özellik | Açıklama |
|---------|----------|
| `.is_certified` | `status == "certified"` |
| `.has_local_homeomorphism_warning` | Uyarılarda yerel homeomorfizm referansı var mı |

### Yardımcı Fonksiyonlar

| Fonksiyon | Açıklama |
|-----------|----------|
| `circle_degree_covering_profile(n)` | $z \mapsto z^n$ için $n$-yapraklı $S^1 \to S^1$ |
| `known_covering_profile(key)` | Kayıttan profil al |
| `covering_profile_summary(profile)` | Profili dict olarak özetle |

## 4. Örnekler

### Örnek 1 — Minimal: $\mathbb{R} \to S^1$ Evrensel Örtü

```python
# R -> S^1 evrensel örtüsü
universal = known_covering_profile("real_line_to_circle")
s = covering_profile_summary(universal)

print("=== R → S¹ Evrensel Örtü ===")
for key, val in s.items():
    if key != "warnings":
        print(f"  {key:35}: {val}")

print()
print("Onaylı mı?:", universal.is_certified)
print("π₁ notu   :", universal.fundamental_group_note)
```

### Örnek 2 — Orta Düzey: $n$-Yapraklı Daire Örtüleri

```python
# z -> z^n: n-yapraklı S^1 -> S^1 örtüsü
print(f"{'Derece':8} {'Yaprak':10} {'Harita':18} {'Onaylı':8} {'π₁ notu'}")
print("-" * 80)
for n in [1, 2, 3, 5]:
    p = circle_degree_covering_profile(n)
    print(f"  {n:6}   {str(p.sheet_count):10} {p.covering_map:18} {str(p.is_certified):8} {p.fundamental_group_note[:45]}")

print()

# Genel "degree n" sembolik modeli
generic = known_covering_profile("circle_to_circle_degree_n")
print(f"Sembolik model: sheet_count='{generic.sheet_count}', status='{generic.status}'")
print(f"Not: {generic.warnings[0]}")
```

### Örnek 3 — Gelişmiş: Trivial İki Yapraklı Örtü ve Deste Dönüşümleri

```python
# Trivial iki yapraklı örtü: B x {0,1} -> B
trivial = known_covering_profile("trivial_two_sheet_cover")
s = covering_profile_summary(trivial)

print("=== Trivial 2-Yapraklı Örtü ===")
print(f"  Üst uzay     : {s['total_space']}")
print(f"  Baz uzay     : {s['base_space']}")
print(f"  Yaprak sayısı: {s['sheet_count']}")
print(f"  Harita       : {s['covering_map']}")
print(f"  Yerel homeo  : {s['local_homeomorphism_assumption']}")
print(f"  π₁ notu      : {s['fundamental_group_note']}")

print()

# Deste dönüşüm grubu yorumu
print("Deste dönüşümleri:")
print("  Trivial örtü: Deck = Z/2Z (iki yaprağı değiştiren tek non-trivial dönüşüm)")
print("  R -> S^1    : Deck = Z    (t ↦ t+n, n∈Z — sargı adımları)")
print("  z -> z^n   : Deck = Z/nZ (n'inci birimin kökleriyle çarpma)")
```

### Örnek 4 — Gerçekçi Senaryo: Özel Örtü Profili Oluşturma

```python
# S^3 -> SO(3) iki yapraklı örtüsü (spin cover)
spin_cover = covering_map_profile(
    "spin_cover_s3_to_so3",
    total_space="S^3 (unit quaternions)",
    base_space="SO(3)",
    sheet_count=2,
    status="certified",
    covering_map="q -> rotation by q (conjugation action on R^3)",
    local_homeomorphism_assumption="standard Lie group result",
    fundamental_group_note=(
        "pi_1(SO(3)) = Z/2Z; S^3 is simply connected (pi_1=0), "
        "so Deck = Z/2Z = {+1, -1} quaternions."
    ),
    certification="standard-lie-theory",
    warnings=("Teaching profile; Lie group details not verified by pytop.",),
)

print("=== S³ → SO(3) Spin Örtüsü ===")
print(f"  Yaprak sayısı: {spin_cover.sheet_count}")
print(f"  Harita       : {spin_cover.covering_map}")
print(f"  Onaylı       : {spin_cover.is_certified}")
print(f"  π₁ notu      : {spin_cover.fundamental_group_note}")

print()

# Tüm kayıtlı modellerin karşılaştırması
print("=== Kayıtlı Modeller Özeti ===")
print(f"{'Ad':35} {'Yaprak':12} {'Statü':12} {'Onaylı'}")
print("-" * 70)
for key, prof in KNOWN_COVERING_MODELS.items():
    print(f"  {key:33} {str(prof.sheet_count):12} {prof.status:12} {prof.is_certified}")
```

## 1. Konu

### 1.1 Sezgisel Açıklama

Ayrışma aksiyomları, bir topolojik uzaydaki **noktalar ve kapalı kümelerin ne kadar iyi birbirinden ayrılabildiğini** ölçer.

**Sezgi — Komşulukla ayırma:**  
İki farklı nokta $x$ ve $y$ verilsin. Ne kadar güçlü bir aksiyom varsa, o kadar "büyük" açık kümelerle (komşuluklarla) birbirinden ayırabilirsiniz:

| Aksiyom | Ne ayırır? | Nasıl? |
|---------|-----------|--------|
| **T0** | İki nokta | Birini içeren ama diğerini içermeyen bir açık var |
| **T1** | İki nokta | Her biri için diğerini dışlayan bir açık var |
| **T2 (Hausdorff)** | İki nokta | Ayrık açık komşuluklar |
| **T2½ (Urysohn)** | İki nokta | Ayrık **kapalı** komşuluklar |
| **T3 (Düzenli T1)** | Nokta & kapalı küme | Ayrık açık komşuluklar |
| **Tychonoff (T3½)** | Nokta & kapalı küme | Ayıran sürekli fonksiyon var |
| **T4 (Normal T1)** | İki kapalı küme | Ayrık açık komşuluklar |
| **T6 (Tamamen normal)** | İki kapalı küme | Ayıran sürekli fonksiyon + $G_\delta$ kapalılar |

**Neden önemli?**
- T2: dizilerin limitleri tektir; kompakt alt kümeler kapalıdır
- T3½: uzayı $[0,1]^I$'ya gömer (Tychonoff gömme teoremi)
- T4: Urysohn lemması ve Tietze uzatma teoremi geçerlidir
- Metrik uzaylar T4'tür (ve daha güçlüdür)

---

### 1.2 Formal Tanımlar

**T0 (Kolmogorov):**  
$x \neq y \Rightarrow$ $\exists$ açık $U$: ($x \in U, y \notin U$) veya ($y \in U, x \notin U$).

**T1 (Fréchet):**  
$x \neq y \Rightarrow$ $\exists$ açık $U \ni x$, $y \notin U$ ve $\exists$ açık $V \ni y$, $x \notin V$.  
Denkdeğeri: Her tekil küme $\{x\}$ kapalıdır.

**T2 (Hausdorff):**  
$x \neq y \Rightarrow$ $\exists$ açık $U \ni x$, $V \ni y$: $U \cap V = \emptyset$.

**T2½ (Urysohn):**  
$x \neq y \Rightarrow$ $\exists$ açık $U \ni x$, $V \ni y$: $\overline{U} \cap \overline{V} = \emptyset$.

**Düzenli (Regular):**  
$x \notin F$ (kapalı $F$) $\Rightarrow$ $\exists$ açık $U \ni x$, $V \supseteq F$: $U \cap V = \emptyset$.  
**T3** = Düzenli + T1.

**Tamamen Düzenli (Completely Regular):**  
$x \notin F$ (kapalı $F$) $\Rightarrow$ $\exists$ sürekli $f: X \to [0,1]$: $f(x)=0$, $f|_F = 1$.  
**Tychonoff (T3½)** = Tamamen Düzenli + T1.

**Normal:**  
Ayrık kapalı $E, F$ $\Rightarrow$ $\exists$ açık $U \supseteq E$, $V \supseteq F$: $U \cap V = \emptyset$.  
**T4** = Normal + T1.

**Tamamen Normal (Perfectly Normal, T6):**  
T4 + her kapalı küme bir $G_\delta$ kümesidir.

## 2. Teoremler

### Teorem 1 — Ayrışma Zinciri

**İfade:**
$$\text{T6} \Rightarrow \text{T4} \Rightarrow \text{T}3\tfrac{1}{2} \Rightarrow \text{T3} \Rightarrow \text{T}2\tfrac{1}{2} \Rightarrow \text{T2} \Rightarrow \text{T1} \Rightarrow \text{T0}$$
Hiçbir ok tersine çevrilemez (her adım için karşıörnek vardır).

**İspat (T4 ⇒ T3):**  
$X$ normal T1, $x \in X$, $F$ kapalı, $x \notin F$.  
T1 olduğundan $\{x\}$ kapalıdır. $\{x\} \cap F = \emptyset$ ve her ikisi kapalı.  
Normallikten ayrık açıklar $U \supseteq \{x\}$ ve $V \supseteq F$ mevcuttur, $U \cap V = \emptyset$.  
Yani $x \in U$, $F \subseteq V$, $U \cap V = \emptyset$ — düzenlilik sağlandı. T1 zaten var. $\square$

---

### Teorem 2 — Urysohn Lemması

**İfade:**  
$X$ normal, $E$ ve $F$ ayrık kapalı kümeler.  
O zaman $\exists$ sürekli $f: X \to [0,1]$: $f|_E = 0$ ve $f|_F = 1$.

**İspat (ana adımlar):**  
1. Dyadic rationals $D = \{k/2^n : 0 \leq k \leq 2^n, n \geq 0\} \cap [0,1]$.
2. Normallikle indüksiyon: $E \subseteq U_0 \subseteq \overline{U_0} \subseteq U_{1/2} \subseteq \overline{U_{1/2}} \subseteq U_1 = F^c$; genel adımda $U_r$ ailesini inşa et.
3. $f(x) = \inf\{r \in D : x \in U_r\}$ tanımla ($x \notin \bigcup U_r$ ise $f(x)=1$).
4. $f^{-1}([0,a))$ ve $f^{-1}((b,1])$ açık: sürekliliği garanti eder. $\square$

**Sonuç:** Normal T1 (= T4) $\Rightarrow$ Tamamen Düzenli T1 (= Tychonoff). Yani T4 ⇒ T3½.

---

### Teorem 3 — Tietze Uzatma Teoremi

**İfade:**  
$X$ normal, $F \subseteq X$ kapalı, $f: F \to [-1,1]$ sürekli.  
O zaman $\exists$ sürekli $g: X \to [-1,1]$: $g|_F = f$.

**İspat özeti:**  
Urysohn lemmasını tekrarlı olarak uygula:  
- $A_1 = f^{-1}([-1, -1/3])$, $B_1 = f^{-1}([1/3, 1])$ ayrık kapalı kümeler.
- Urysohn'dan $g_1: X \to [-1/3, 1/3]$, $g_1|_{A_1} = -1/3$, $g_1|_{B_1} = 1/3$.
- $f_1 = f - g_1|_F$: $\|f_1\|_{\infty} \leq 2/3$.
- Her adımda hata $2/3$ oranında küçülür; $g = \sum g_n$ üniform yakınsar. $\square$

| Tag | Anlamı |

## 4. Örnekler

### Örnek 1 — Minimal: Sierpiński Uzayı (T0 ama T1 değil)

```python
# Sierpinski uzayi: X={0,1}, tau={∅, {1}, {0,1}}
# {0} kapalı değil → T1 sağlanmaz
sierpinski = TopologicalSpace(
    carrier=frozenset({0, 1}),
    topology=frozenset([
        frozenset(),          # boş küme
        frozenset({1}),       # {1} açık
        frozenset({0, 1}),    # tam küme açık
    ]),
    metadata={"name": "Sierpiński uzayı"},
)

for prop in ["t0", "t1", "hausdorff"]:
    r = analyze_separation(sierpinski, prop)
    icon = "✓" if r.is_true else ("✗" if r.is_false else "?")
    print(f"  {prop:12} {icon}  {r.justification[0][:70]}")
```

### Örnek 2 — Orta Düzey: Tam Ayrışma Profili

```python
# Metrik uzay tagları: metrizable → T4 ve daha güçlü
metric_sp = TopologicalSpace.symbolic(
    description="Metrik uzay (örn. R^n)",
    tags=["metrizable", "hausdorff", "t4", "perfectly_normal", "second_countable"],
)

profile = separation_profile(metric_sp)

print(f"{'Aksiyom':20} {'Durum':10} Gerekçe")
print("-" * 80)
for prop, r in profile.items():
    status = "TRUE" if r.is_true else ("FALSE" if r.is_false else "?")
    print(f"  {prop:18} {status:10} {r.justification[0][:50]}")
```

### Örnek 3 — Gelişmiş: Zincir Kırılması — T3 ama T4 değil

```python
# Sorgenfrey düzlemi (RxR alt tabanla) T3 ama T4 değil
# (Klasik karşıörnek: normal olmayan T3 uzay)
sorgenfrey = TopologicalSpace.symbolic(
    description="Sorgenfrey düzlemi R_ℓ × R_ℓ",
    tags=["t3", "hausdorff", "t1", "not_t4", "not_normal"],
)

print("Sorgenfrey düzlemi ayrışma durumu:")
for prop in SEPARATION_CHAIN_ORDER:
    r = analyze_separation(sorgenfrey, prop)
    icon = "✓" if r.is_true else ("✗" if r.is_false else "?")
    print(f"  {prop:18} {icon}")

print()
print("Zincirin nerede koptuğunu görüyoruz: T3 ✓  ama T4 ✗")
```

### Örnek 4 — Gerçekçi Senaryo: Dört Uzayı Karşılaştır

```python
import itertools

scenarios = [
    ("Sierpiński",       TopologicalSpace(
        carrier=frozenset({0,1}),
        topology=frozenset([frozenset(), frozenset({1}), frozenset({0,1})]),
    )),
    ("Ayrık (sonlu)",   TopologicalSpace(
        carrier=frozenset({1,2,3}),
        topology=frozenset(
            frozenset(s) for r in range(4)
            for s in itertools.combinations({1,2,3}, r)
        ),
        metadata={"name": "3 noktalı ayrık uzay"},
    )),
    ("Tychonoff",        TopologicalSpace.symbolic(
        description="Tychonoff uzay (örn. tamamen düzenli T1)",
        tags=["tychonoff", "t1"],
    )),
    ("Kompakt Hausdorff", TopologicalSpace.symbolic(
        description="Kompakt Hausdorff uzay (örn. S^n)",
        tags=["compact", "hausdorff", "t4"],
    )),
]

props = ["t0", "t1", "hausdorff", "t3", "tychonoff", "t4"]
header = f"{'Uzay':22}" + "".join(f"{p:12}" for p in props)
print(header)
print("-" * (22 + 12 * len(props)))
for label, sp in scenarios:
    row = f"{label:22}"
    for p in props:
        r = analyze_separation(sp, p)
        icon = "✓" if r.is_true else ("✗" if r.is_false else "?")
        row += f"{icon:12}"
    print(row)
```

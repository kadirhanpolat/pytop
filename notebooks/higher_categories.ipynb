## 1. Konu: (∞,1)-Kategori Tanımı ve Modeller

Bir **(∞,1)-kategori**, her dereceden morfizmlere sahip olan, ancak k≥2 için tüm k-morfizmlerin
homotopik olarak tersinir olduğu bir matematiksel yapıdır. Bu yapıyı temsil eden dört temel model vardır:

### (∞,1)-Kategori Modelleri

| Model | Temsil | Temel Koşul | Referans |
|-------|--------|------------|----------|
| Quasi-kategori | sSet (simplicial küme) | İç horn doldurma (Λⁿₖ → X, 0 < k < n) | Joyal 2002 |
| Kan kompleksi | sSet | Tüm hornlar dolar (∞-grupoid) | Quillen 1967 |
| Tam Segal uzayı | bisSet | Segal koşulu + tamlık | Rezk 2001 |
| sSet-zenginleştirilmiş kategori | sSet-Cat | Bergner model yapısı | Bergner 2007 |

### Stabilite ve Sunulabilirlik

| Özellik | Anlam |
|---------|-------|
| **Kararlı (stable)** | Sıfır nesnesi var, tüm pushout/pullback var, pushout ⟺ pullback |
| **Sunulabilir (presentable)** | Erişilebilir + tüm küçük eşlimit var; P(C)'nin lokal bir hali |
| **∞-topos** | Sunulabilir + Giraud aksiyomları (evrensel eşlimitler, ayrık koproduktlar, efektif grupoidler) |
| **t-yapı** | (C_{≥0}, C_{≤0}) çifti; C^♥ = C_{≥0} ∩ C_{≤0} bir abelyen kategorisi |

### Bergner'in Eşdeğerlik Teoremi

Aşağıdaki dört model kategorisi birbiriyle Quillen eşdeğerdir:
- Joyal model yapısı (quasi-kategoriler üzerinde)
- Rezk model yapısı (bisimplicial kümeler üzerinde, fibrant = tam Segal uzayı)
- Segal kategoriler üzerindeki model yapısı
- Bergner model yapısı (sSet-zenginleştirilmiş kategoriler üzerinde)

## 1.2 Joyal Model Yapısı ve Homotopi Kategorisi

### İç Horn Doldurma

Bir simplicial küme X, aşağıdaki koşulu sağlarsa **quasi-kategori** adını alır:
$$\text{Her } \Lambda^n_k \to X \text{ için } 0 < k < n \implies \Delta^n \to X \text{ uzantısı var.}$$

- **0-simpleksler**: Nesneler
- **1-simpleksler**: Morfizmler  
- **Bileşim**: Λ²₁ horn dolgusu (benzersiz değil, homotopik olarak benzersiz)

### Homotopi Kategorisi hX

Bir quasi-kategori X'in homotopi kategorisi hX:
- **Nesneler**: X_0 (0-simpleksler)
- **Morfizmler**: π₀(X₁) / homotopi (1-simpleksler arası homotopi)
- **Bileşim**: Horn dolgularının π₀ görüntüsü

### Joyal vs Kan-Quillen Model Yapısı

| Özellik | Joyal | Kan-Quillen |
|---------|-------|-------------|
| Fibrant nesneler | Quasi-kategoriler | Kan kompleksleri |
| Zayıf eşdeğerlikler | Kategorik eşdeğerlikler | Zayıf homotopi eşdeğerlikleri |
| Modeller | (∞,1)-kategoriler | Homotopi tipleri (∞-grupoidler) |

## 2. Teoremler

### Teorem 1 (Joyal İç Horn Koşulu, 2002)
Bir simplicial küme X, her iç horn Λⁿₖ → X (0 < k < n) için Δⁿ → X uzantısı varsa
ve yalnızca bu durumda bir **quasi-kategori**dir. Quasi-kategoriler, (∞,1)-kategorilerin
başlıca modelidir; nesneler 0-simpleksler, morfizmler 1-simplekslerdir. Bileşim
homotopik olarak benzersizdir (ama kesin değil). Joyal model yapısında fibrant nesneler
quasi-kategorilerdir.

### Teorem 2 (Grothendieck Homotopi Hipotezi — Kan Kompleksleri)
Bir simplicial küme X, tüm hornlar için (Λⁿₖ → X, 0 ≤ k ≤ n) uzantı varsa bir
**Kan kompleksi**dir; bu ∞-grupoid modelidir (tüm morfizmler homotopi anlamında tersinir).
Grothendieck hipotezi: ∞-grupoidler ≃ homotopi tipleri (CW-kompleksleri).
Teknik versiyon: tekil kompleks Sing: Top → sSet Kan-Quillen model yapısında Quillen
eşdeğerliği kurar.

### Teorem 3 (Bergner Eşdeğerlikleri, 2007)
Dört model kategorisi birbiriyle Quillen eşdeğerdir:
(a) Joyal model yapısı (quasi-kategoriler),
(b) Rezk CSS model yapısı (tam Segal uzayları),
(c) Segal kategoriler üzerindeki model yapısı,
(d) Bergner model yapısı (sSet-kategoriler).
Bu dört yapının tamamı aynı (∞,2)-kategorisini sunar.

### Teorem 4 (Lurie ∞-Topos Giraud Aksiyomları, HTT 2009)
Sunulabilir bir ∞-kategori X, aşağıdaki koşulların tamamını sağlıyorsa bir **∞-topos**tur:
(1) Eşlimitler evrenseldir (pullback eşlimitleri korur),
(2) Koprodüktlar ayrıktır (X ×_{X+Y} Y = ∅),
(3) Her grupoid nesnesi etkilidir (Cech siniri doğru bölümü verir).
Eşdeğer koşul: X, bir P(C) presheaf ∞-kategorisinin erişilebilir sol kesin lokalizasyonudur.

````python
# Teoremler — profil alanlarından doğrulama
def check_fields(key, expected_fields):
    p = next(x for x in profiles if x.key == key)
    results = {}
    for field, expected in expected_fields.items():
        actual = getattr(p, field)
        results[field] = ("OK" if actual == expected else f"FAIL(got {actual!r})", actual)
    return results

print("Teorem 1 — Quasi-kategori (iç horn, sSet modeli):")
r = check_fields("quasi_category", {"category_type": "quasi_category",
                                     "model_type": "simplicial_set",
                                     "is_stable": False})
for f, (st, v) in r.items(): print(f"  {f}: {st} ({v!r})")

print("\nTeorem 2 — Kan kompleksi / ∞-grupoid:")
r = check_fields("kan_complex_groupoid", {"category_type": "kan_complex",
                                           "is_stable": False, "is_presentable": False})
for f, (st, v) in r.items(): print(f"  {f}: {st} ({v!r})")

print("\nTeorem 3 — Tam Segal uzayı (bisimplicial model):")
r = check_fields("complete_segal_space", {"category_type": "complete_segal_space",
                                           "model_type": "bisimplicial_set"})
for f, (st, v) in r.items(): print(f"  {f}: {st} ({v!r})")

print("\nTeorem 4 — ∞-topos (sunulabilir, Giraud aksiyomları):")
r = check_fields("infinity_topos", {"category_type": "infinity_topos",
                                     "is_presentable": True,
                                     "has_all_limits": True, "has_all_colimits": True})
for f, (st, v) in r.items(): print(f"  {f}: {st} ({v!r})")
````

## 3. Fonksiyon Analizi

Bu bölümde 4 analiz fonksiyonunun çok katmanlı karar mantığı incelenmektedir.
Her fonksiyon `Result` nesnesi döndürür; `r.is_true` / `r.is_false` ile sorgulanır.

### 3.1 `is_infinity_categorical(space)`

**Sözde kod:**
```
Katman 1: QUASI_CATEGORY_TAGS | KAN_COMPLEX_TAGS → true  (açık ∞-kategori modeli)
Katman 2: SEGAL_SPACE_TAGS | STABLE_INFINITY_TAGS | INFINITY_TOPOS_TAGS → true  (tanınan modeller)
Katman 3: MODEL_CATEGORY_TAGS | ENRICHED_CATEGORY_TAGS → true  (lokalizasyon yoluyla sunar)
Katman 4: {strict_1_category, ordinary_category, set_category} → false
Katman 5: Bilinmiyor
```

**Karmaşıklık:** O(|tags|). **Avantaj:** 5 katman, geniş kapsam. 
**Not:** Model kategoriler ∞-kategori değil, ∞-kategoriyi *sunar* (Dwyer-Kan lokalizasyonu ile).

````python
class _S:
    def __init__(self, *tags): self.tags = set(tags)

print("=== is_infinity_categorical ===")
cases = [
    ("quasi_category",         "Katman 1: quasi-kategori → true"),
    ("inner_horn_filling",     "Katman 1: iç horn doldurma → true"),
    ("kan_complex",            "Katman 1: Kan kompleksi → true"),
    ("infinity_groupoid",      "Katman 1: ∞-grupoid → true"),
    ("complete_segal_space",   "Katman 2: tam Segal uzayı → true"),
    ("stable_infinity_category","Katman 2: kararlı ∞-kategori → true"),
    ("infinity_topos",         "Katman 2: ∞-topos → true"),
    ("model_category",         "Katman 3: model kategorisi (sunar) → true"),
    ("dg_category",            "Katman 3: dg-kategori (dg-sinir ile sunar) → true"),
    ("strict_1_category",      "Katman 4: katı 1-kategori → false"),
    ("ordinary_category",      "Katman 4: sıradan kategori → false"),
]
for tag, desc in cases:
    r = is_infinity_categorical(_S(tag))
    status = "TRUE" if r.is_true else ("FALSE" if r.is_false else "UNKNOWN")
    print(f"  [{tag:<30s}] → {status:<8s} | {desc}")

r_unk = is_infinity_categorical(_S("compact_manifold"))
status = "TRUE" if r_unk.is_true else ("FALSE" if r_unk.is_false else "UNKNOWN")
print(f"  [{'compact_manifold':<30s}] → {status:<8s} | Katman 5: etiket yok → bilinmiyor")
````

### 3.2 `is_stable_infinity_category(space)`

**Sözde kod:**
```
Katman 1: STABLE_INFINITY_TAGS → true  (kararlı ∞-kategori etiketleri)
Katman 2: {spectra_category, derived_infinity_category, dg_category, stable_category} → true
Katman 3: {infinity_topos, segal_space, kan_complex} → false  (bunlar kararlı değil)
Katman 4: Bilinmiyor
```

**Avantaj:** ∞-toposes kararlı değil (sıfır nesnesi yoktur, ∅ ≠ *). 
**Not:** Bir quasi-kategori kararlı olabilir ya da olmayabilir — etiket yoksa bilinmiyor.

````python
print("=== is_stable_infinity_category ===")
cases = [
    ("stable_infinity_category", "Katman 1: açık etiket → true"),
    ("spectra_category",          "Katman 1: spektra kategorisi → true"),
    ("triangulated_infinity",     "Katman 1: üçgenlenmiş ∞ → true"),
    ("pushout_pullback_stable",   "Katman 1: pushout=pullback → true"),
    ("derived_infinity_category", "Katman 1: türevli ∞-kategori → true"),
    ("dg_category",               "Katman 2: dg-kategori (pretriangulated) → true"),
    ("infinity_topos",            "Katman 3: ∞-topos (sıfır nesnesi yok) → false"),
    ("segal_space",               "Katman 3: Segal uzayı → false"),
    ("kan_complex",               "Katman 3: Kan kompleksi → false"),
    ("quasi_category",            "Katman 4: belirsiz (kararlı olabilir) → bilinmiyor"),
]
for tag, desc in cases:
    r = is_stable_infinity_category(_S(tag))
    status = "TRUE" if r.is_true else ("FALSE" if r.is_false else "UNKNOWN")
    print(f"  [{tag:<30s}] → {status:<8s} | {desc}")
````

### 3.3 `has_all_limits_and_colimits(space)`

**Sözde kod:**
```
Katman 1: INFINITY_TOPOS_TAGS | ADJUNCTION_TAGS → true  (∞-toposes ve adjoint fonksiyon teoremi)
```

````python
print("=== has_all_limits_and_colimits ===")
cases = [
    ("infinity_topos",           "Katman 1: ∞-topos → true"),
    ("left_exact_localization",  "Katman 1: sol kesin lokalizasyon → true"),
    ("infinity_adjunction",      "Katman 1: ∞-adjoint → true"),
    ("adjoint_functor_theorem",  "Katman 1: adjoint fonksiyon teoremi → true"),
    ("yoneda_infinity",          "Katman 1: Yoneda → true"),
    ("presentable_infinity",     "Katman 2: sunulabilir → true"),
    ("accessible_infinity",      "Katman 2: erişilebilir → true"),
    ("quasi_category",           "Katman 2: quasi-kategori → true"),
    ("dg_category",              "Katman 3: dg-kategori → false"),
    ("not_complete",             "Katman 3: tamamlanmamış → false"),
    ("no_limits",                "Katman 3: limit yok → false"),
]
for tag, desc in cases:
    r = has_all_limits_and_colimits(_S(tag))
    status = "TRUE" if r.is_true else ("FALSE" if r.is_false else "UNKNOWN")
    print(f"  [{tag:<30s}] → {status:<8s} | {desc}")
````

### 3.4 `is_presentable_infinity_category(space)`

**Sözde kod:**
```
Katman 1: INFINITY_TOPOS_TAGS → true  (∞-toposes sunulabilirdir)
Katman 2: {presentable_infinity, accessible_infinity} | ADJUNCTION_TAGS → true
Katman 3: {dg_category, not_presentable, small_category} → false
Katman 4: Bilinmiyor
```

**Avantaj:** ∞-toposların sunulabilirliği Lurie'nin HTT §5.5 teorisinin temelidir.
dg-kategoriler zenginleştirilmiş kategoriler olduğundan sunulabilir ∞-kategori değildir.

````python
print("=== is_presentable_infinity_category ===")
cases = [
    ("infinity_topos",          "Katman 1: ∞-topos (sunulabilir) → true"),
    ("lurie_topos",             "Katman 1: Lurie toposu → true"),
    ("infinity_giraud",         "Katman 1: Giraud aksiyomları → true"),
    ("higher_topos",            "Katman 1: yüksek topos → true"),
    ("presentable_infinity",    "Katman 2: açık etiket → true"),
    ("accessible_infinity",     "Katman 2: erişilebilir → true"),
    ("infinity_adjunction",     "Katman 2: adjoint (AFT) → true"),
    ("dg_category",             "Katman 3: dg-kategori (sunulabilir değil) → false"),
    ("not_presentable",         "Katman 3: açık ret → false"),
    ("small_category",          "Katman 3: küçük kategori → false"),
    ("stable_infinity_category","Katman 4: kararlı ∞-kategori (belirsiz) → bilinmiyor"),
]
for tag, desc in cases:
    r = is_presentable_infinity_category(_S(tag))
    status = "TRUE" if r.is_true else ("FALSE" if r.is_false else "UNKNOWN")
    print(f"  [{tag:<30s}] → {status:<8s} | {desc}")
````

## 6. Alıştırmalar

Aşağıdaki alıştırmalar öğrencilere yüksek kategori teorisini pekiştirmek için tasarlanmıştır.

**Alıştırma 1.** Bir Kan kompleksi (∞-grupoid) ile bir quasi-kategorinin farkını açıklayın.
Hangi horn doldurmak zorunda? `KAN_COMPLEX_TAGS` ve `QUASI_CATEGORY_TAGS`'teki etiketleri
inceleyerek bu farkı Python koduyla örnekleyin.

**Alıştırma 2.** Tam Segal uzayının iki koşulunu (Segal koşulu + tamlık koşulu) tanımlayın.
Bir bisimplicial küme W için Segal koşulunu sembolik olarak yazın:
W_n ≃ W_1 ×_{W_0} ... ×_{W_0} W_1. `SEGAL_SPACE_TAGS` etiketlerini listeleyin.

**Alıştırma 3.** Bir kararlı ∞-kategorinin üç özelliğini listeleyin. Neden ∞-toposes kararlı
değildir? `is_stable_infinity_category` fonksiyonunu `infinity_topos` etiketiyle çağırın
ve sonucu yorumlayın.

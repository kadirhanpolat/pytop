# Boyut Teorisi

**Modül:** `pytop.dimension_theory`  
**Konu:** Küçük tümevarım boyutu `ind`, büyük tümevarım boyutu `Ind`, kaplama boyutu `dim`, sıfır boyutluluk

---

Bu notebook beş bölümden oluşur:
1. **Konu** — Boyut sezgisi ve üç farklı topolojik boyut tanımı
2. **Teoremler** — $\text{ind} \leq \text{Ind}$, metrik uzaylarda $\text{dim} = \text{ind} = \text{Ind}$, Cantor kümesi boyutu
3. **API** — `dim`, `ind`, `Ind`, sıfır boyutluluk yüklemleri
4. **Örnekler** — Dört senaryo: Cantor'dan $\mathbb{R}^n$'ye
5. **Alıştırmalar** — Kodlama, teori ve karşılaştırma

## 1. Konu

### 1.1 Sezgisel Açıklama

Topolojik boyut, uzayın "kaç boyutlu göründüğünü" sezgisel kavramdan titiz bir şekilde yakalar.

**Sezgi 1 — Noktanın boyutu:**  
Bir nokta $\{*\}$ 0 boyutludur. Bir doğru parçası 1 boyutludur. Düzlem 2 boyutludur. Ama topolojik boyut neden zaten bildiğimiz şeyle örtüşüyor ve nasıl ispatlıyoruz?

**Sezgi 2 — Sıfır boyutluluk = clopen taban:**  
Bir uzay **sıfır boyutlu** ise her noktanın clopen (hem açık hem kapalı) kümelerden oluşan bir komşuluk tabanı vardır. Rasyoneller $\mathbb{Q}$ sıfır boyutludur — her komşuluğun içinde hem açık hem kapalı kümeler bulabilirsiniz. Cantor kümesi de sıfır boyutludur.

**Sezgi 3 — Tümevarım (inductive) boyut:**  
$\text{ind}(X) = n$ ise her sınır $\partial U$ en fazla $(n-1)$ boyutludur. Yani $\mathbb{R}$'nin boyutu 1'dir çünkü sınırları (nokta çiftleri) 0 boyutludur.

**Üç eşdeğer boyut tanımı (normal metrik uzaylarda):**

| Boyut | Tanım | Sezgi |
|-------|-------|-------|
| `ind` (küçük tümevarım) | Sınır boyutuna göre | Yerel |
| `Ind` (büyük tümevarım) | Kapalı küme separatörüne göre | Global |
| `dim` (kaplama boyutu) | Sonlu örtülerin küçüklüğüne göre | Kombinatorik |

**Neden önemli?**
- Sıfır boyutlu kompakt Hausdorff = Stone uzayı (Boolean cebirlerinin dual'ı)
- Boyut, homeomorfizm invariantıdır ($\mathbb{R}^m \not\cong \mathbb{R}^n$ için $m \neq n$)
- Fraktalların boyutu bu teorinin genelleştirilmesiyle (Hausdorff boyutu) tanımlanır

---

### 1.2 Formal Tanımlar

**Küçük Tümevarım Boyutu `ind`:**  
- $\text{ind}(\emptyset) = -1$
- $\text{ind}(X) \leq n$ $\iff$ her $x \in X$ için her $U \ni x$ açıkta, $V \ni x$ açık $V \subseteq U$ ve $\text{ind}(\partial V) \leq n-1$
- $\text{ind}(X) = n$ $\iff$ $\text{ind}(X) \leq n$ ve $\text{ind}(X) \leq n-1$ değil

**Büyük Tümevarım Boyutu `Ind`:**  
- $\text{Ind}(\emptyset) = -1$
- $\text{Ind}(X) \leq n$ $\iff$ her $A \subseteq X$ kapalı, $U \supseteq A$ açık için, $A \subseteq V \subseteq U$ ile $\text{Ind}(\partial V) \leq n-1$ olan $V$ açığı var

**Kaplama Boyutu `dim`:**  
$\text{dim}(X) \leq n$ $\iff$ $X$'in her sonlu açık örtüsünün, örtü derecesi $\leq n+1$ olan bir sonlu açık iyileştirmesi var.

**Sıfır Boyutluluk:**  
$\text{ind}(X) = 0$ $\iff$ $X$ clopen küme tabanına sahip $\iff$ $\text{dim}(X) = 0$ (normal uzayda).

---
## 2. Teoremler

### Teorem 1 — Normal Metrik Uzaylarda $\text{ind} = \text{Ind} = \text{dim}$

**İfade (Hurewicz-Wallman, 1941):**  
$X$ ayrılabilir metrik uzay $\Rightarrow$ $\text{ind}(X) = \text{Ind}(X) = \text{dim}(X)$.

**İspat taslağı (yalnızca $\text{ind} \leq \text{dim}$ yönü):**  
$\text{dim}(X) \leq n$ varsayalım. Her $x$ ve komşuluk $U$ için, kaplama boyutu tanımıyla $x$'i bir $V \subseteq U$ içinde izole eden açık küme bulunabilir; $\partial V$'nin kaplama boyutu $\leq n-1$ gösterilebilir (tümevarım).  
Ters yön ve $\text{Ind}$ ile eşitlik için tam cebirsel argüman gerekir; referans: Hurewicz & Wallman, "Dimension Theory" (1941). $\square$

**Not:** Normal olmayan uzaylarda üç boyut tanımı farklı değer verebilir.

---

### Teorem 2 — Cantor Kümesi Sıfır Boyutludur

**İfade:** $C$ (standart üçte-bir Cantor kümesi) için $\text{dim}(C) = 0$.

**İspat:**  
$C$ kompakt, metrik, tamamen bağlantısız ($C$'nin her bağlantı bileşeni bir noktadır).  
Kompakt Hausdorff uzayda tamamen bağlantısızlık $\Leftrightarrow$ clopen taban $\Leftrightarrow$ $\text{ind} = 0$.  
$C$'yi inşa eden her adımda kalan aralıklar clopen kümeler oluşturur; bunlar taban oluşturur.  
Kaplama boyutu = 0 (her açık kaplama, çapı 0'a yaklaşan clopen kümelerle iyileştirilebilir). $\square$

**Önemli not:** $C$ 0 boyutlu olmasına rağmen $|C| = |\mathbb{R}|$ (süreklilik gücünde). Boyut sayısallıkla aynı şey değildir.

---

### Teorem 3 — $\text{dim}(\mathbb{R}^n) = n$

**İfade:** $\text{dim}(\mathbb{R}^n) = n$.

**İspat (üst sınır $\leq n$):**  
$\mathbb{R}^n$'de her açık kaplama, herhangi bir noktanın en fazla $n+1$ açık kümenin ortak noktasında bulunduğu şekilde iyileştirilebilir. Bu, $\delta$-ızgara örtüsü ve kombinatorik argümanla gösterilir.

**İspat (alt sınır $\geq n$):**  
$\text{dim}(\mathbb{R}^n) \leq n-1$ olsaydı, cebirsel topoloji (Brouwer sabit nokta teoremi) ile çelişki doğar — $n$-boyutlu Öklid uzayında her sürekli $f: D^n \to D^n$'nin sabit noktası olması $\text{dim} \geq n$ gerektirir. $\square$

## 3. API

````python
from pytop.dimension_theory import (
    dim,
    ind,
    Ind,
    has_clopen_base,
    is_zero_dimensional,
    is_totally_disconnected,
    ZERO_DIMENSION_TAGS,
    TOTALLY_DISCONNECTED_TAGS,
    NEGATIVE_ZERO_DIMENSION_TAGS,
)
print("pytop.dimension_theory yüklendi.")
print("Sıfır boyut tagları  :", ZERO_DIMENSION_TAGS)
print("Tam bağlantısız taglar:", TOTALLY_DISCONNECTED_TAGS)
````

### API Kullanım Şekilleri

`dim(space)`, `ind(space)`, `Ind(space)` fonksiyonları şu girdileri kabul eder:

1. **Tag tabanlı sembolik dict:**
   ```python
   {"tags": ["zero_dimensional"], "metadata": {}}
   ```
2. **İsimden benchmark:** `{"metadata": {"name": "R^3"}}` → `dim = 3`
3. **Açık alan:** `{"dim": 2, "ind": 2, "Ind": 2}` ya da `{"metadata": {"dim": 2}}`
4. **Tag tabanlı `TopologicalSpace.symbolic`:**
   ```python
   TopologicalSpace.symbolic(description="...", tags=["zero_dimensional"])
   ```

**Döndürülen değer:** `int | None`  
- `0` → sıfır boyutlu  
- `n` → $n$-boyutlu  
- `None` → belirlenemeyen (yetersiz bilgi)

**Yüklemler (bool döner):**

| Fonksiyon | Açıklama |
|-----------|----------|
| `is_zero_dimensional(space)` | `ind = 0` veya `zero_dimensional` tag |
| `has_clopen_base(space)` | Clopen taban var mı? |
| `is_totally_disconnected(space)` | Tamamen bağlantısız mı? |

## 4. Örnekler

### Örnek 1 — Minimal: İsimden Otomatik Boyut

````python
# Öklid uzayları: isimden benchmark
euclidean_spaces = [
    {"metadata": {"name": "R^1"}},
    {"metadata": {"name": "R^2"}},
    {"metadata": {"name": "R^3"}},
    {"metadata": {"name": "R^10"}},
]

print("Öklid uzaylarında kaplama boyutu (isimden):")
print(f"{'Uzay':10} {'dim':5} {'ind':5} {'Ind'}")
print("-" * 28)
for sp in euclidean_spaces:
    name = sp["metadata"]["name"]
    print(f"  {name:8}  {str(dim(sp)):5}  {str(ind(sp)):5}  {str(Ind(sp))}")

print()

# Cantor kümesi: tag tabanlı
cantor = {"tags": ["cantor_set"], "metadata": {"name": "Cantor set"}}
print(f"Cantor kümesi: dim={dim(cantor)}, ind={ind(cantor)}, sıfır boyutlu={is_zero_dimensional(cantor)}")
````

### Örnek 2 — Orta Düzey: Sıfır Boyutlu Uzaylar

````python
from pytop.spaces import TopologicalSpace

# Çeşitli sıfır boyutlu uzaylar
zero_dim_spaces = [
    ("Cantor kümesi",   {"tags": ["cantor_set"]}),
    ("Stone uzayı",     {"tags": ["stone_space"]}),
    ("Clopen taban",    {"tags": ["clopen_base"]}),
    ("Zero-dim tag",    {"tags": ["zero_dimensional"]}),
    ("Tam bağlantısız", {"tags": ["totally_disconnected"]}),
]

print(f"{'Uzay':22} {'dim':5} {'is_zero_dim':12} {'has_clopen':12} {'is_totally_disc'}")
print("-" * 70)
for label, sp in zero_dim_spaces:
    d = dim(sp)
    z = is_zero_dimensional(sp)
    c = has_clopen_base(sp)
    t = is_totally_disconnected(sp)
    print(f"  {label:20}  {str(d):5}  {str(z):12}  {str(c):12}  {t}")

print()

# TopologicalSpace.symbolic ile
rationals = TopologicalSpace.symbolic(
    description="Q (rasyoneller)",
    tags=["zero_dimensional", "totally_disconnected"]
)
print(f"Q: dim={dim(rationals)}, is_zero_dim={is_zero_dimensional(rationals)}, "
      f"totally_disc={is_totally_disconnected(rationals)}")
````

### Örnek 3 — Gelişmiş: Açık Alan ile Boyut

````python
# Doğrudan dim/ind/Ind alanı verilen uzaylar
explicit_spaces = [
    ("1-manifold (R)",   {"dim": 1, "ind": 1, "Ind": 1}),
    ("2-manifold (S^2)", {"dim": 2, "ind": 2, "Ind": 2}),
    ("Cantor (explicit)",{"dim": 0, "ind": 0, "Ind": 0}),
    ("Sınır bilinmiyor", {}),  # None bekleniyor
]

print("Açık alan ile boyut sorgusu:")
print(f"{'Uzay':22} {'dim':6} {'ind':6} {'Ind'}")
print("-" * 42)
for label, sp in explicit_spaces:
    print(f"  {label:20}  {str(dim(sp)):6}  {str(ind(sp)):6}  {str(Ind(sp))}")

print()

# metadata içinde de aranır
meta_space = {"metadata": {"dim": 4, "name": "exotic R^4"}}
print(f"Metadata içinden: dim={dim(meta_space)}, isim='exotic R^4'")
````

### Örnek 4 — Gerçekçi Senaryo: Boyut Tablosu

````python
# Kapsamlı karşılaştırma tablosu
space_table = [
    ("Tek nokta {*}",     {"dim": 0, "ind": 0, "Ind": 0}),
    ("R (gerçek doğru)",  {"metadata": {"name": "R^1"}}),
    ("R^2 (düzlem)",      {"metadata": {"name": "R^2"}}),
    ("R^n (n=5)",         {"metadata": {"name": "R^5"}}),
    ("Cantor kümesi",     {"tags": ["cantor_set"]}),
    ("Q (rasyoneller)",   {"tags": ["zero_dimensional"]}),
    ("Sonsuz ayrık uzay", {"tags": ["totally_disconnected"], "ind": 0}),
    ("Bilinmeyen",        {}),
]

print(f"{'Uzay':26} {'dim':6} {'ind':6} {'Ind':6} {'sıfır-d':8} {'clopen-t'}")
print("-" * 68)
for label, sp in space_table:
    d  = dim(sp)
    i  = ind(sp)
    I  = Ind(sp)
    z  = is_zero_dimensional(sp)
    cb = has_clopen_base(sp)
    print(f"  {label:24}  {str(d):6}  {str(i):6}  {str(I):6}  "
          f"{'✓' if z else '✗':8}  {'✓' if cb else '✗'}")
````

### Alıştırma 3 — Karşılaştırma: Boyut ve Diğer Özellikler

Aşağıdaki ilişkileri `pytop.dimension_theory` ve `pytop.connectedness` birlikte kullanarak test edin:

1. `totally_disconnected` tag → `is_zero_dimensional` doğru mu?
2. `zero_dimensional` tag → `is_totally_disconnected` doğru mu?
3. `dim = n > 0` → `is_zero_dimensional` yanlış mı?

Sonuçları bir doğruluk tablosu olarak yazdırın.

````python
from pytop.connectedness import is_totally_disconnected as conn_is_totally_disconnected

test_cases = [
    ("totally_disconnected",   {"tags": ["totally_disconnected"]}),
    ("zero_dimensional",       {"tags": ["zero_dimensional"]}),
    ("cantor_set",             {"tags": ["cantor_set"]}),
    ("stone_space",            {"tags": ["stone_space"]}),
    ("dim=2 (düzlem)",         {"dim": 2, "ind": 2}),
    ("dim=1 (doğru)",          {"metadata": {"name": "R^1"}}),
]

print(f"{'Uzay':24} {'dim':5} {'zero_dim':10} {'clopen':8} {'tot_disc(dim)':14} {'tot_disc(conn)'}")
print("-" * 78)
for label, sp in test_cases:
    d   = dim(sp)
    zd  = is_zero_dimensional(sp)
    cb  = has_clopen_base(sp)
    td1 = is_totally_disconnected(sp)          # dimension_theory
    td2 = conn_is_totally_disconnected(sp)     # connectedness
    print(f"  {label:22}  {str(d):5}  {str(zd):10}  {str(cb):8}  {str(td1):14}  {td2}")
````

### Alıştırma 4 — Boyut–Bağlılık Çapraz Kontrolü

`pytop.dimension_theory` ve `pytop.connectedness` modüllerini birlikte kullanarak şu ilişkileri doğrulayın:
1. `cantor_set` ve `stone_space` taglı uzaylar: `dim=0` ve `is_totally_disconnected=True` tutarlı mı?
2. `R^1`, `R^2` uzayları: `dim>0` ve `is_connected=True` tutarlı mı?
3. Bağlı bir uzayın sıfır boyutlu olamayacağını (`dim=0` + `connected=True` çelişkisini) beş farklı tag kombinasyonu için tablolayın

````python
from pytop.connectedness import (
    is_connected as conn_connected,
    is_path_connected as conn_path_connected,
)

test_spaces = [
    ("Cantor kümesi",      {"tags": ["cantor_set"]}),
    ("Stone uzayı",        {"tags": ["stone_space"]}),
    ("R^1 (bağlı)",        {"metadata": {"name": "R^1"}}),
    ("R^2 (yol-bağlı)",   {"metadata": {"name": "R^2"}}),
    ("Ayrık (sonsuz)",     {"tags": ["totally_disconnected"], "ind": 0}),
]

print(f"{'Uzay':22} {'dim':5} {'zero_d':8} {'bağlı':8} {'yol-b':7} {'tutarlı'}")
print("-" * 68)
for label, sp in test_spaces:
    d  = dim(sp)
    zd = is_zero_dimensional(sp)
    co = conn_connected(sp)
    pc = conn_path_connected(sp)
    # Tutarlılık: sıfır-d ve bağlı aynı anda True olmamalı
    consistent = "✓"
    if zd and co.is_true:
        consistent = "⚠ sıfır-d+bağlı?"
    print(f"  {label:20}  {str(d):5}  {str(zd):8}  "
          f"{'✓' if co.is_true else '✗':8}  {'✓' if pc.is_true else '✗':7}  {consistent}")

print()
print("Beklenen ilişkiler:")
print("  dim=0 → totally_disconnected (sıfır boyutlu uzay bağlı olamaz)")
print("  dim>0 + bağlı → zero_dimensional değil")
print("  Cantor/Stone: dim=0, bağlı=✗ — tutarlı")
````

## Özet

| Kavram | Açıklama |
|--------|----------|
| `ind(X)` | Küçük tümevarım boyutu; sınır boyutuna göre |
| `Ind(X)` | Büyük tümevarım boyutu; separator boyutuna göre |
| `dim(X)` | Kaplama boyutu; örtü derecesine göre |
| $\text{ind}=\text{Ind}=\text{dim}$ | Ayrılabilir metrik uzaylarda eşit (Hurewicz–Wallman) |
| Sıfır boyutluluk | Clopen taban; tamamen bağlantısız; Stone uzayı |
| Cantor kümesi | $\text{dim}=0$; ama $|C|=|\mathbb{R}|$ — boyut $\neq$ kardinalite |
| $\text{dim}(\mathbb{R}^n)=n$ | Brouwer + kombinatorik ızgara argümanı |
| `is_zero_dimensional(sp)` | `ind=0` veya `zero_dimensional` tag |
| `has_clopen_base(sp)` | Clopen taban varlığı |
| `is_totally_disconnected(sp)` | Tamamen bağlantısız mı? |

## 1. Konu

### 1.1 Sezgisel Açıklama

Düğüm teorisi, $\mathbb{R}^3$'te (ya da $S^3$'te) gömülü kapalı eğrileri çalışır. Temel soru: **iki düğüm birbirine izotop mu, yoksa gerçekten farklı mı?**

**Sezgi 1 — Kesintisiz dönüşüm:**  
Bir düğümü bir ip olarak düşünün; uçları yapıştırılmış. Kesip yeniden yapıştırmadan diğerine dönüştürebiliyorsanız izotoptur (aynı düğüm tipidir). Çözük düğüm (unknot) ile üç yapraklı düğüm (trefoil) arasında böyle bir dönüşüm yoktur.

**Sezgi 2 — Diyagram ve kesişimler:**  
$\mathbb{R}^2$'ye yansıtılmış düğüm diyagramında kesişimler görünür. Kesişim sayısı düğüm tipini sınıflandırmaya yardımcı olur; ama aynı sayıyla birden fazla düğüm olabilir.

**Sezgi 3 — Değişmez (invariant):**  
Düğüm tipini ayırt etmek için değişmezler kullanılır: iki izotop düğümün değişmezleri aynıdır. Değişmez farklıysa düğümler farklıdır. Örnekler: kesişim sayısı, Alexander polinomu, Jones polinomu.

**Neden önemli?**
- DNA topolojisi: süpersarmal ve topoisomeraz reaksiyonları
- Kimyasal topoloji: kiral moleküler düğümler
- 3-manifold topolojisi: Dehn cerrahisi
- Kuantum alan teorisi (Jones polinomu ↔ Chern-Simons)

---

### 1.2 Formal Tanımlar

**Düğüm (Knot):**  
Gömme $f: S^1 \hookrightarrow S^3$ (veya $\mathbb{R}^3$). İki düğüm $K_1, K_2$ **eşdeğerdir** (izotop) $\iff$ $S^3$'ün bir ortam izotopisi $K_1$'i $K_2$'ye götürür.

**Link:**  
$n$ bileşenli link: $n$ ayrık çemberin $S^3$'e gömülmesi. Her bileşen bir düğümdür.

**Düğüm Diyagramı:**  
$\mathbb{R}^2$'ye genel konum yansıtması; her kesişimde üst/alt ip bilgisi eklenir.

**Reidemeister Hamleleri:**  
Üç yerel diyagram operasyonu (R1, R2, R3) — her ikisi de düğüm tipini korur.

**Bağlantı Sayısı (Linking Number):**  
İki bileşenli link için: $\ell k(K_1, K_2) = $ yönlü kesişimlerin işaretli toplamının yarısı.

**Alexander Polinomu $\Delta_K(t)$:**  
Seifert matrisi üzerinden tanımlanan $\mathbb{Z}[t, t^{-1}]$ elemanı. İzotopi değişmezidir.

## 2. Teoremler

### Teorem 1 — Reidemeister Teoremi

**İfade:**  
İki düğüm diyagramı aynı düğüm tipini temsil eder ancak ve ancak biri diğerine sonlu sayıda R1, R2, R3 hamlesi uygulanarak dönüştürülebilir.

**İspat özeti:**  
_(⇐)_ Her Reidemeister hamlesi bir ortam izotopisi gerçekleştirir — diyagram değişse de $\mathbb{R}^3$'teki eğri aynı kalır.

_(⇒)_ $H: S^3 \times [0,1] \to S^3$ ortam izotopisi verilsin. Genel konum argümanıyla $H$'yi sonlu sayıda "elementar" adıma böl; her adım diyagram üzerinde en fazla bir Reidemeister hamlesiyle temsil edilebilir.  
Bu ayrışım, transversallik ve katmanlaştırma (stratification) tekniklerine dayanır — her adımda diyagram geçişi R1/R2/R3'ten biridir. $\square$

**Sonuç:** Düğüm değişmezi inşa etmek için Reidemeister hamleleri altında korunan bir nesne tanımlamak yeterlidir.

---

### Teorem 2 — Alexander Polinomu İzotopi Değişmezidir

**İfade:**  
$K_1 \simeq K_2$ (izotop) $\Rightarrow$ $\Delta_{K_1}(t) = \Delta_{K_2}(t)$ ($t$ kuvveti ve birim çarpana kadar).

**İnşa (Seifert matrisi yoluyla):**  
1. $K$ için bir Seifert yüzeyi $F$ seç ($\partial F = K$, yönlendirilebilir).
2. $H_1(F)$ üzerinde $V_{ij} = \ell k(a_i, a_j^+)$ Seifert matrisi $V$ tanımla ($a_i$ 1-zincirler, $a_j^+$ normal yönde itilmiş kopyası).
3. $\Delta_K(t) = \det(V - tV^T)$.

**İspat taslağı:**  
Farklı Seifert yüzeyleri, stabilizasyon hamleleriyle birbirine bağlıdır. Stabilizasyon $\Delta_K$'yı $t^{\pm 1}$ ve birim çarpana kadar değiştirmez (bu, Seifert'in orijinal hesabından izlenir). $\square$

---

### Teorem 3 — Trefoil Çözük Düğümden Farklıdır

**İfade:** Trefoil $\not\simeq$ unknot.

**İspat (Alexander polinomuyla):**  
$\Delta_{\text{unknot}}(t) = 1$  
$\Delta_{\text{trefoil}}(t) = t^{-1} - 1 + t \neq 1$  
Alexander polinomu izotopi değişmezi olduğundan iki düğüm izotop olamaz. $\square$

## 3. API

```python
from pytop.knots import (
    KnotProfile,
    LinkProfile,
    ReidemeisterMoveProfile,
    KnotInvariantProfile,
    KnotApplicationProfile,
    get_knot_profiles,
    get_link_profiles,
    get_reidemeister_move_profiles,
    get_knot_invariant_profiles,
    get_knot_application_profiles,
    knot_crossing_summary,
    knot_invariant_kind_summary,
    knot_theory_profile_registry,
    knot_application_domain_summary,
)
print("pytop.knots yüklendi.")
print("Kayıt özeti:", knot_theory_profile_registry())
```

### Modülün Yapısı

| Fonksiyon | Döndürür | İçerik |
|-----------|----------|--------|
| `get_knot_profiles()` | `tuple[KnotProfile]` | Unknot, trefoil, sekiz, cinquefoil |
| `get_link_profiles()` | `tuple[LinkProfile]` | Çözük link, Hopf link |
| `get_reidemeister_move_profiles()` | `tuple[ReidemeisterMoveProfile]` | R1, R2, R3 |
| `get_knot_invariant_profiles()` | `tuple[KnotInvariantProfile]` | Alexander, Jones (trefoil + unknot) |
| `get_knot_application_profiles()` | `tuple[KnotApplicationProfile]` | DNA, kimyasal topoloji |
| `knot_crossing_summary()` | `dict[str, int]` | Düğüm → kesişim sayısı |
| `knot_invariant_kind_summary()` | `dict[str, list[str]]` | Değişmez türü → profil anahtarları |
| `knot_theory_profile_registry()` | `dict[str, int]` | Tüm ailelerin kayıt sayısı |

## 4. Örnekler

### Örnek 1 — Minimal: Temel Düğüm Katalogu

```python
knots = get_knot_profiles()

print(f"{'Anahtar':25} {'Görünen Ad':25} {'Kesişim'}")
print("-" * 65)
for k in knots:
    print(f"  {k.key:23} {k.display_name:25} {k.crossing_count}")

print()
print("Kesişim sayısı indeksi:", knot_crossing_summary())
```

### Örnek 2 — Orta Düzey: Reidemeister Hamleleri ve Değişmezler

```python
# Reidemeister hamleleri — hepsi düğüm tipini korur
moves = get_reidemeister_move_profiles()
print("=== Reidemeister Hamleleri ===")
for m in moves:
    icon = "✓" if m.preserves_knot_type else "✗"
    print(f"  {m.move_type:5} {m.display_name:20} korur={icon}  etki: {m.local_effect}")

print()

# Değişmez profilleri — türe göre grupla
print("=== Değişmez Aileleri ===")
groups = knot_invariant_kind_summary()
for kind, keys in groups.items():
    print(f"  {kind}: {keys}")

print()

# Trefoil vs Unknot — Alexander polinomu karşılaştırması
print("=== Alexander Polinomu Karşılaştırması ===")
for inv in get_knot_invariant_profiles():
    if inv.invariant_kind == "alexander_polynomial":
        print(f"  {inv.applies_to:25} Δ(t) = {inv.symbolic_value}")
```

### Örnek 3 — Gelişmiş: Jones Polinomu ve Ayna Simetrisi

```python
# Jones polinomu: trefoil sağ ve sol elli olarak farklılaşır
# (alexander bunu yapamazken jones yapabilir — daha güçlü değişmez)
print("=== Jones Polinomu Profilleri ===")
for inv in get_knot_invariant_profiles():
    if inv.invariant_kind == "jones_polynomial":
        print(f"  {inv.display_name}")
        print(f"    Sembolik değer : {inv.symbolic_value}")
        print(f"    Not            : {inv.normalization_note}")
        print(f"    Öğretim notu   : {inv.teaching_note}")
        print()

# Link profilleri — Hopf link bağlantı sayısı
print("=== Link Profilleri ===")
for lnk in get_link_profiles():
    print(f"  {lnk.display_name:30} bileşen={lnk.component_count}")
    print(f"    Sinyal: {lnk.basic_linking_signal}")
    print()
```

### Örnek 4 — Gerçekçi Senaryo: DNA Topolojisi Uygulamaları

```python
apps = get_knot_application_profiles()

print("=== Uygulama Alanı Dağılımı ===")
domains = knot_application_domain_summary()
for domain, count in domains.items():
    print(f"  {domain:35} {count} profil")

print()
print("=== DNA Topolojisi Profilleri ===")
for app in apps:
    if app.application_domain == "dna_topology":
        print(f"  {app.display_name}")
        print(f"    Topolojik nesne: {app.topological_object}")
        print(f"    Sinyal         : {app.application_signal}")
        print(f"    Öğretim notu   : {app.teaching_note}")
        print()

print("=== Kimyasal Topoloji Profilleri ===")
for app in apps:
    if app.application_domain == "chemical_topology":
        print(f"  {app.display_name}")
        print(f"    Topolojik nesne: {app.topological_object}")
        print(f"    Sinyal         : {app.application_signal}")
        print()
```

### Alıştırma 1 — Kodlama Görevi

Aşağıdaki soruları `pytop.knots` profilleriyle yanıtlayın:

1. Kesişim sayısı ≤ 4 olan tüm düğümleri listeleyin.
2. Her değişmez türü için kaç profil kayıtlı olduğunu yazdırın.
3. Hem trefoil hem unknot için Alexander ve Jones polinomlarını yan yana karşılaştırın.

```python
# 1. Kesişim sayısı <= 4
print("Kesişim sayısı ≤ 4 olan düğümler:")
for k in get_knot_profiles():
    if k.crossing_count <= 4:
        print(f"  {k.display_name} (c={k.crossing_count})")

print()

# 2. Değişmez türü sayıları
print("Değişmez türü → profil sayısı:")
for kind, keys in knot_invariant_kind_summary().items():
    print(f"  {kind}: {len(keys)}")

print()

# 3. Trefoil vs Unknot karşılaştırma
invs = {(i.applies_to, i.invariant_kind): i.symbolic_value
        for i in get_knot_invariant_profiles()}

print(f"{'Düğüm':30} {'Alexander':20} {'Jones'}")
print("-" * 70)
for knot_key in ["unknot_baseline", "trefoil_knot"]:
    alex = invs.get((knot_key, "alexander_polynomial"), "—")
    jones = invs.get((knot_key, "jones_polynomial"), "—")
    print(f"  {knot_key:28} {alex:20} {jones}")
```

| Düğüm | Kesişim | Alexander Δ(t) | Jones V(q) |
|-------|---------|---------------|------------|

### Alıştırma 4 — Uygulama ve Reidemeister Profil Özeti

Tüm profil ailelerini birleştiren kapsamlı bir rapor oluşturun:
1. `get_knot_application_profiles()`'dan her uygulama alanı için profil sayısını ve topolojik nesneleri listeleyin
2. `get_reidemeister_move_profiles()`'dan her hamlenin yerel etkisini ve tip koruyuculuğunu tablolayın
3. Trefoil için hem değişmez profillerini hem de ilgili uygulama profillerini (`applies_to` ve `topological_object` alanları üzerinden) eşleştirin

```python
from collections import defaultdict

# 1. Uygulama profilleri — alana göre grupla
print("=== Uygulama Alanları ===")
app_by_domain = defaultdict(list)
for app in get_knot_application_profiles():
    app_by_domain[app.application_domain].append(app)

for domain, apps in app_by_domain.items():
    print(f"  {domain} ({len(apps)} profil):")
    for app in apps:
        print(f"    [{app.display_name}] topolojik nesne: {app.topological_object}")

print()

# 2. Reidemeister hamleler tablosu
print("=== Reidemeister Hamleleri ===")
print(f"{'Hamle':6} {'Ad':22} {'Korur':6} {'Yerel etki'}")
print("-" * 70)
for m in get_reidemeister_move_profiles():
    icon = "✓" if m.preserves_knot_type else "✗"
    print(f"  {m.move_type:4}  {m.display_name:20}  {icon:4}  {m.local_effect}")

print()

# 3. Trefoil için değişmez + uygulama eşlemesi
print("=== Trefoil Değişmez Profilleri ===")
for inv in get_knot_invariant_profiles():
    if inv.applies_to == "trefoil_knot":
        print(f"  {inv.invariant_kind:25} = {inv.symbolic_value}")

print("\nTüm uygulama profillerinde düğüm nesneleri:")
for app in get_knot_application_profiles():
    print(f"  [{app.application_domain}] {app.display_name}: {app.topological_object[:55]}")
```

|--------|----------|
| Düğüm | $S^1 \hookrightarrow S^3$ gömülmesi; izotopi sınıfları |

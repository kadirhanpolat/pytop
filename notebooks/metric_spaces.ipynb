## 1. Konu

### 1.1 Sezgisel Açıklama

Günlük hayatta "mesafe" sezgiseldir: iki şehir arasındaki kilometre, iki ses dosyası arasındaki benzerlik skoru, iki gen dizisi arasındaki edit mesafesi. Topoloji bu sezgiyi soyutlar: "mesafe" ne olursa olsun, **yakın olmak** ve **sınır** kavramları aynı şekilde çalışır.

**Metrik uzay = mesafe + 4 kural:**
- Bir nokta kendine mesafesi sıfırdır
- Mesafe simetriktir
- Farklı noktalara mesafe pozitiftir
- Üçgen eşitsizliği: dolaylı yol doğrudan yoldan uzundur

**Metrik → Topoloji:**  
Her metrik uzay doğal bir topoloji üretir: açık topların ürettiği topoloji. Bu sayede metrik uzay = topolojik uzay + ekstra yapı.

---

### 1.2 Formal Tanımlar

**Metrik:**  
$d: X \times X \to [0, \infty)$ bir metrik (ya da uzaklık fonksiyonu) olmak için:
1. $d(x,y) = 0 \iff x = y$ \hspace{2em}(öz-eşitlik)
2. $d(x,y) = d(y,x)$ \hspace{3.4em}(simetri)
3. $d(x,z) \le d(x,y) + d(y,z)$ \hspace{1em}(üçgen eşitsizliği)

$(X, d)$ çiftine **metrik uzay** denir.

**Açık Top:**  
$B(x, r) = \{y \in X : d(x,y) < r\}$ merkez $x$, yarıçap $r > 0$.

**Kapalı Top:**  
$\bar{B}(x, r) = \{y \in X : d(x,y) \le r\}$.

**İndüklenen Topoloji:**  
$(X, d)$ metrik uzayının indüklediği topoloji:
$$\tau_d = \{U \subseteq X : \forall x \in U,\, \exists r > 0,\, B(x,r) \subseteq U\}$$
Diğer bir deyişle: $U$ açık $\iff$ $U$'nun her noktasının etrafında tamamen $U$ içinde bir açık top var.

**Cauchy Dizisi:**  
$(x_n) \subseteq X$ Cauchy $\iff$ $\forall \varepsilon > 0,\, \exists N,\, \forall m,n \ge N: d(x_m, x_n) < \varepsilon$.

**Tamamlı Metrik Uzay:**  
Her Cauchy dizisi yakınsayan bir diziyse $(X, d)$ tamamlıdır.

## 2. Teoremler

### Teorem 1 — Her Metrik Uzay Hausdorff'tur

**İfade:**  
$(X, d)$ metrik uzay $\Rightarrow$ $X$, $\tau_d$ topolojisiyle Hausdorff'tur.

**İspat:**  
$x \ne y$ olsun. $r = d(x,y) > 0$. Yarıçap $\varepsilon = r/2$ alalım. $U = B(x, \varepsilon)$ ve $V = B(y, \varepsilon)$ açıktır.

İddia: $U \cap V = \emptyset$.  
Aksine $z \in U \cap V$ olsun. O zaman $d(x,z) < \varepsilon$ ve $d(y,z) < \varepsilon$. Üçgen eşitsizliğiyle:
$$d(x,y) \le d(x,z) + d(z,y) < \varepsilon + \varepsilon = r = d(x,y)$$
Bu çelişkidir. Dolayısıyla $U \cap V = \emptyset$ ve $X$ Hausdorff'tur. $\square$

---

### Teorem 2 — Baire Kategorisi Teoremi

**İfade:**  
$(X, d)$ tamamlı metrik uzay. Eğer $X = \bigcup_{n=1}^\infty F_n$ ve her $F_n$ kapalıysa, o zaman en az bir $F_N$'nin içi ($\operatorname{int}(F_N)$) boş değildir.

**Eşdeğer ifade (açık yoğun kümeler):**  
Tamamlı metrik uzayda, her birinin tamamlayıcısı yoğun açık olmak üzere sayılabilir sayıda açık yoğun kümenin kesişimi yoğundur.

**İspat (ana fikir):**  
Her $F_n$'nin içinin boş olduğunu varsayalım (çelişki için). İç boş demek: $X \setminus F_n$ yoğun açık demek. Tamamlılığı kullanarak iç içe geçmiş açık toplar inşa ederiz:
$$B_1 \supseteq \bar{B}_1 \supseteq B_2 \supseteq \bar{B}_2 \supseteq \cdots$$
yarıçapları $r_n \to 0$ ile. Merkezler $(x_n)$ Cauchy dizisi oluşturur (her $m,n \ge N$ için $d(x_m,x_n) < 2r_N \to 0$). Tamamlılıktan $x_n \to x$. Bu $x$, $\bigcap \bar{B}_n \subseteq X \setminus \bigcup F_n$'dedir, bu ise $\bigcup F_n \ne X$ anlamına gelir — çelişki. $\square$

---

### Teorem 3 — İndüklenen Topoloji Topoloji Oluşturur

**İfade:**  
$(X, d)$ metrik uzay. $\tau_d$ (açık toplarla üretilen aile) bir topoloji oluşturur.

**İspat:**

_(T1)_ $\emptyset \in \tau_d$ (boşluk koşulu trivial). $X \in \tau_d$: her $x \in X$ için $B(x,1) \subseteq X$.

_(T2) Sonlu kesişim:_ $U, V \in \tau_d$, $x \in U \cap V$. $\exists r_1, r_2 > 0$ ile $B(x,r_1) \subseteq U$ ve $B(x,r_2) \subseteq V$. $r = \min(r_1, r_2) > 0$ alalım. $B(x,r) \subseteq B(x,r_1) \cap B(x,r_2) \subseteq U \cap V$.

_(T3) Keyfi birleşim:_ $U = \bigcup_\alpha U_\alpha$, $x \in U$. O zaman $x \in U_\alpha$ bazı $\alpha$ için. $\exists r > 0$ ile $B(x,r) \subseteq U_\alpha \subseteq U$. $\square$

## 4. API

````python
from pytop.metric_spaces import (
    FiniteMetricSpace, open_ball, validate_metric, induced_topological_space
)

coords = {i: float(i) for i in range(6)}
def eucl(x, y): return abs(coords[x] - coords[y])
fms6 = FiniteMetricSpace(carrier=frozenset(range(6)), distance=eucl)

# open_ball: O(n * D)
print("B(3, 1.5):", sorted(open_ball(fms6, 3, 1.5)))   # {2,3,4}
print("B(3, 2.5):", sorted(open_ball(fms6, 3, 2.5)))   # {1,2,3,4,5}

# validate_metric: O(n³) — 6³=216 üçlü
v = validate_metric(fms6)
print(f"\nMetrik geçerli: {v.is_true}  (n=6 → {6**3} üçlü kontrol)")

# induced_topological_space: O(2^n * n²)
induced = induced_topological_space(fms6)
print(f"\nIndüklenen topoloji |τ| = {len(induced.topology)}")
print(f"  (2^6 = {2**6} alt küme arasından seçildi)")
````

### Örnek 3 — Gelişmiş: Ayrık Metrik

````python
# Ayrık metrik: d(x,y) = 0 ise x=y, 1 ise x≠y
X = frozenset({'a', 'b', 'c'})

def discrete_d(x, y):
    return 0 if x == y else 1

disc_metric = FiniteMetricSpace(carrier=X, distance=discrete_d)

# r < 1 → tek nokta; r >= 1 → tüm uzay
print("B('a', 0.5):", open_ball(disc_metric, 'a', 0.5))   # sadece {'a'}
print("B('a', 1.5):", open_ball(disc_metric, 'a', 1.5))   # {'a','b','c'}

# İndüklenen topoloji ayrık mı?
induced_disc = induced_topological_space(disc_metric)
print(f"\nAçık küme sayısı: {len(induced_disc.topology)} (2^3=8 bekleniyor)")
````

### Alıştırma 4 — Karmaşıklık: `validate_metric` ve `induced_topological_space`

a) `validate_metric` neden O(n³) karmaşıklığa sahiptir? `n = 3, 4, 5, 6` için kaç üçlü (triple) kontrol edildiğini hesaplayın ve karşılaştırın.

````python
from pytop.metric_spaces import (
    FiniteMetricSpace, validate_metric, induced_topological_space,
    normalized_metric, capped_metric
)

# a) validate_metric: O(n³) üçlü sayısı
print("n → n³ (validate_metric üçlü sayısı)")
for n in [3, 4, 5, 6]:
    print(f"  n={n}: n³ = {n**3}")

# b) Ayrık metrik — induced topoloji boyutu
print()
def disc(x, y): return 0 if x == y else 1

for n in [2, 3, 4, 5]:
    carrier_n = frozenset(range(n))
    fms_n = FiniteMetricSpace(carrier=carrier_n, distance=disc)
    tau_size = len(induced_topological_space(fms_n).topology)
    print(f"  n={n}: |τ|={tau_size:4}  2^n={2**n:4}  (ayrık metrik → ayrık topoloji)")

# c) normalize/cap ile topoloji boyutu değişiyor mu?
coords5 = {i: float(i) for i in range(5)}
def eucl5(x, y): return abs(coords5[x] - coords5[y])
fms5 = FiniteMetricSpace(carrier=frozenset(range(5)), distance=eucl5)

norm_d = normalized_metric(fms5)
cap_d  = capped_metric(fms5, cap=1.5)
fms_norm = FiniteMetricSpace(carrier=frozenset(range(5)), distance=norm_d)
fms_cap  = FiniteMetricSpace(carrier=frozenset(range(5)), distance=cap_d)

tau_orig = len(induced_topological_space(fms5).topology)
tau_norm = len(induced_topological_space(fms_norm).topology)
tau_cap  = len(induced_topological_space(fms_cap).topology)
print(f"\nÖklid:      |τ| = {tau_orig}")
print(f"Normalize:  |τ| = {tau_norm}")
print(f"Cap(1.5):   |τ| = {tau_cap}")
````

## 1. Konu

### 1.1 Sezgisel Açıklama

Bağlılık, bir uzayın **tek parçadan mı oluştuğunu** soran sorudur.

**Sezgi 1 — İki parça yok:**  
$X$ bağlıdır $\iff$ $X$'i iki ayrık, boş olmayan açık kümeye **bölmek imkânsızdır**. $\mathbb{R}$ bağlıdır: $(-\infty,0) \cup (0,+\infty)$ bir bölme gibi görünse de $0$ noktasını kapsamaz. $\mathbb{R} \setminus \{0\}$ ise bağlı değildir.

**Sezgi 2 — Yol ile ulaşmak:**  
$X$ yol bağlıdır $\iff$ herhangi iki nokta arasında $X$'in içinde kalan sürekli bir yol vardır. Yol bağlılık, bağlılıktan güçlüdür; ters her zaman doğru değildir (topşist eğri).

**Sezgi 3 — Yerel bağlılık:**  
Her noktanın bağlı bir komşuluk tabanı varsa uzay yerel bağlıdır. $\mathbb{R}$ hem bağlı hem yerel bağlıdır; ama bazı uzaylar biri olmadan diğeri olabilir.

**Özellik hiyerarşisi:**
$$\text{Yay bağlı} \Rightarrow \text{Yol bağlı} \Rightarrow \text{Bağlı}$$
Hiçbir ok tersine çevrilemez.

**Neden önemli?**
- Ara değer teoremi: bağlılık + sürekli fonksiyon → ara değer
- Homeomorfizm invariantı: $\mathbb{R}$ ve $\mathbb{R}^2$'nin homeomorf olmadığını kanıtlar
- Bileşenler: uzayın "parçalarını" anlamak
- Sabit nokta teoremlerine zemin (Brouwer)

---

### 1.2 Formal Tanımlar

**Bağlılık:**  
$(X, \tau)$ **bağlıdır** $\iff$ $X = U \sqcup V$ ($U, V$ açık, boş olmayan) şeklinde yazılamaz.  
Denkdeğeri: $X$'in tek clopen (hem açık hem kapalı) kümeleri $\emptyset$ ve $X$'dir.

**Yol Bağlılığı:**  
$x, y \in X$ için $\exists$ sürekli $\gamma: [0,1] \to X$, $\gamma(0) = x$, $\gamma(1) = y$.  
Yol bağlılık ⇒ bağlılık (ama tersi değil).

**Yay Bağlılığı (Arc-connected):**  
$\gamma: [0,1] \to X$ homeomorfizma — yolun enjektif olmasını ister.  
Hausdorff uzaylarda yol bağlılık = yay bağlılık.

**Yerel Bağlılık:**  
Her $x \in X$ ve her $U \ni x$ açık için, $x$'i içeren bağlı bir açık $V \subseteq U$ var.

**Tam Bağlantısızlık (Totally Disconnected):**  
Her bağlantı bileşeni bir nokta kümesidir. Örnek: $\mathbb{Q}$, Cantor kümesi.

**Dağınık (Scattered):**  
Her boş olmayan alt uzayın izole bir noktası vardır. Örnek: sonlu T0 uzaylar, sayılabilir sıra uzayları.

## 2. Teoremler

### Teorem 1 — $[0,1]$ Bağlıdır

**İfade:** Standart topolojisiyle $[0,1]$ bağlıdır.

**İspat:**  
$[0,1] = U \sqcup V$, $U, V$ açık, boş olmayan olsun. $a \in U$, $b \in V$ alalım, $a < b$ diyelim.  
$c = \sup(U \cap [a,b])$ tanımla. $U$ açık olduğundan $c \notin U$ (yoksa $c$'nin bir komşuluğu $U$'da olurdu ve $c < b$ varsayımıyla $c = \sup$ ile çelişki çıkar).  
Yani $c \in V$. $V$ açık olduğundan $(c-\epsilon, c+\epsilon) \subseteq V$ için bazı $\epsilon > 0$.  
Ama $c = \sup(U \cap [a,b])$ olduğundan $(c-\epsilon, c) \cap U \neq \emptyset$. $U \cap V = \emptyset$ ile çelişki. $\square$

---

### Teorem 2 — Ara Değer Teoremi

**İfade:**  
$f: X \to \mathbb{R}$ sürekli, $X$ bağlı. $f(a) < c < f(b)$ için $\exists x \in X$: $f(x) = c$.

**İspat:**  
$A = f^{-1}((-\infty, c))$, $B = f^{-1}((c, +\infty))$ açıktır (süreklilik).  
$A \cup B \subseteq X$; $f(a) < c$ ⇒ $a \in A$ (boş değil); $f(b) > c$ ⇒ $b \in B$ (boş değil).  
$A \cap B = \emptyset$. Eğer $A \cup B = X$ olsaydı, $X = A \sqcup B$ bağlılıkla çelişirdi.  
Dolayısıyla $\exists x \in X \setminus (A \cup B)$, yani $f(x) = c$. $\square$

---

### Teorem 3 — Yol Bağlı ⇒ Bağlı

**İfade:** Yol bağlı her uzay bağlıdır.

**İspat:**  
$X$ bağlı olmadığını varsay: $X = U \sqcup V$, $U, V$ açık, boş olmayan.  
$a \in U$, $b \in V$ al. Yol bağlılıktan $\gamma: [0,1] \to X$, $\gamma(0)=a$, $\gamma(1)=b$.  
$\gamma^{-1}(U)$ ve $\gamma^{-1}(V)$ açık, boş olmayan ($0 \in \gamma^{-1}(U)$, $1 \in \gamma^{-1}(V)$), ayrık ve birleşimleri $[0,1]$.  
$[0,1]$ bağlıdır — çelişki. $\square$

**Not (Ters Neden Yanlış):**  
**Topşist eğri (Topologist's sine curve):** $S = \{(x, \sin(1/x)) : x > 0\} \cup \{(0,y) : y \in [-1,1]\}$ bağlıdır ama yol bağlı değildir — dikey segment $\{0\} \times [-1,1]$'e yol götürmek imkânsızdır.

```python
from pytop.spaces import TopologicalSpace
from pytop.connectedness import (
    analyze_connectedness,
    is_connected,
    is_path_connected,
    is_arc_connected,
    is_locally_connected,
    is_totally_disconnected,
    is_scattered,
    normalize_connectedness_property,
    TRUE_TAGS,
    FALSE_TAGS,
)
print("pytop.connectedness yüklendi.")
print("Desteklenen özellikler:", sorted(TRUE_TAGS.keys()))
```

### `analyze_connectedness(space, property_name)` ile Sorgu

| `property_name` | Alias | Açıklama |
|----------------|-------|----------|
| `"connected"` | `"connectedness"` | Bağlılık |
| `"path_connected"` | `"pathconnected"` | Yol bağlılığı |

| `"locally_connected"` | `"local_connectedness"` | Yerel bağlılık |

| `"totally_disconnected"` | Tam bağlantısız |
| `"disconnected"` | Bağlı değil |

## 4. Örnekler

### Örnek 1 — Minimal: Sonlu Uzayda Bağlılık

```python
# Bağlı sonlu uzay: topoloji sadece ∅ ve X (indiskret)
sp_indiscrete = TopologicalSpace(
    carrier=frozenset({1, 2, 3}),
    topology=frozenset([frozenset(), frozenset({1, 2, 3})]),
    metadata={"name": "İndiskret uzay (3 nokta)"},
)

# Bağlantısız sonlu uzay: {1,2} ⊔ {3,4}
sp_disconnected = TopologicalSpace(
    carrier=frozenset({1, 2, 3, 4}),
    topology=frozenset([
        frozenset(),
        frozenset({1, 2}),
        frozenset({3, 4}),
        frozenset({1, 2, 3, 4}),
    ]),
    metadata={"name": "Bağlantısız: {1,2} ⊔ {3,4}"},
)

for label, sp in [("İndiskret", sp_indiscrete), ("Bağlantısız", sp_disconnected)]:
    r = is_connected(sp)
    icon = "✓" if r.is_true else "✗"
    print(f"  {label:15} bağlı={icon}  {r.justification[0][:65]}")
```

### Örnek 2 — Orta Düzey: Yol Bağlı → Bağlı Zinciri

```python
# Öklid uzayı: yol bağlı → yay bağlı → bağlı → yerel bağlı
euclidean = TopologicalSpace.symbolic(
    description="R^n (Öklid uzayı)",
    tags=["path_connected", "arc_connected", "connected", "locally_connected"],
)

# Topşist eğri: bağlı ama yol bağlı değil
topologist_sine = TopologicalSpace.symbolic(
    description="Topşist sinüs eğrisi",
    tags=["connected", "not_path_connected", "not_arc_connected", "not_locally_connected"],
)

props = ["connected", "path_connected", "arc_connected", "locally_connected"]

print(f"{'Özellik':22} {'R^n':10} {'Topşist sinüs'}")
print("-" * 50)
for prop in props:
    re = analyze_connectedness(euclidean, prop)
    rs = analyze_connectedness(topologist_sine, prop)
    ie = "✓" if re.is_true else ("✗" if re.is_false else "?")
    is_ = "✓" if rs.is_true else ("✗" if rs.is_false else "?")
    print(f"  {prop:20} {ie:10} {is_}")
```

### Örnek 3 — Gelişmiş: Tam Bağlantısızlık ve Dağınıklık

```python
import itertools

# Ayrık topoloji (T1, sonlu) → totally disconnected + scattered
discrete_3 = TopologicalSpace(
    carrier=frozenset({1, 2, 3}),
    topology=frozenset(
        frozenset(s) for r in range(4)
        for s in itertools.combinations({1, 2, 3}, r)
    ),
    metadata={"name": "3 noktalı ayrık uzay"},
)

# Sierpiński (T0, non-T1) → not totally_disconnected
sierpinski = TopologicalSpace(
    carrier=frozenset({0, 1}),
    topology=frozenset([frozenset(), frozenset({1}), frozenset({0, 1})]),
    metadata={"name": "Sierpiński uzayı"},
)

# Rasyoneller (sembolik): totally disconnected, not scattered
rationals = TopologicalSpace.symbolic(
    description="Q (rasyonel sayılar)",
    tags=["totally_disconnected", "not_scattered", "not_connected"],
)

scenarios = [
    ("Ayrık (3 nokta)", discrete_3),
    ("Sierpiński",       sierpinski),
    ("Q (rasyoneller)", rationals),
]
props2 = ["connected", "totally_disconnected", "scattered"]

print(f"{'Uzay':22}" + "".join(f"{p[:15]:18}" for p in props2))
print("-" * (22 + 18 * len(props2)))
for label, sp in scenarios:
    row = f"{label:22}"
    for p in props2:
        r = analyze_connectedness(sp, p)
        icon = "✓" if r.is_true else ("✗" if r.is_false else "?")
        row += f"{icon:18}"
    print(row)
```

### Örnek 4 — Gerçekçi Senaryo: Altı Uzayın Tam Profili

```python
all_scenarios = [
    ("R (Öklid doğrusu)",
     TopologicalSpace.symbolic(description="R",
         tags=["path_connected", "arc_connected", "locally_connected"])),
    ("S^1 (çember)",
     TopologicalSpace.symbolic(description="S^1",
         tags=["path_connected", "arc_connected", "locally_connected"])),
    ("Cantor kümesi",
     TopologicalSpace.symbolic(description="Cantor set",
         tags=["totally_disconnected", "not_connected", "not_locally_connected"])),
    ("Topşist sinüs",
     TopologicalSpace.symbolic(description="Topologist's sine curve",
         tags=["connected", "not_path_connected", "not_locally_connected"])),
    ("Ayrık (sonsuz)",
     TopologicalSpace.symbolic(description="Infinite discrete space",
         tags=["totally_disconnected", "scattered", "disconnected"])),
    ("İndiskret (sonsuz)",
     TopologicalSpace.symbolic(description="Infinite indiscrete space",
         tags=["path_connected", "connected"])),
]

all_props = ["connected", "path_connected", "arc_connected",
             "locally_connected", "totally_disconnected", "scattered"]

header = f"{'Uzay':22}" + "".join(f"{p[:6]:9}" for p in all_props)
print(header)
print("-" * (22 + 9 * len(all_props)))
for label, sp in all_scenarios:
    row = f"{label:22}"
    for p in all_props:
        r = analyze_connectedness(sp, p)
        icon = "✓" if r.is_true else ("✗" if r.is_false else "?")
        row += f"{icon:9}"
    print(row)
```

## 5. Alıştırmalar

### Alıştırma 1 — Kodlama Görevi

Aşağıdaki topolojiyi elle oluşturun ve `is_connected` ile `is_totally_disconnected` sonuçlarını doğrulayın:

- **Uzay A:** $X = \{1,2,3,4\}$, $\tau = \{\emptyset, \{1\}, \{2,3\}, \{1,2,3\}, X\}$
  - Bağlı mı? (İpucu: $\{1,2,3\}$ clopen mu diye kontrol edin)
- **Uzay B:** $X = \{a,b,c\}$, $\tau$ = ayrık topoloji
  - `totally_disconnected` ve `scattered` için sonuçları yazdırın

```python
# Uzay A
X_A = frozenset({1, 2, 3, 4})
tau_A = frozenset([
    frozenset(),
    frozenset({1}),
    frozenset({2, 3}),
    frozenset({1, 2, 3}),
    X_A,
])
sp_A = TopologicalSpace(carrier=X_A, topology=tau_A)

# Uzay B: ayrık
X_B = frozenset({"a", "b", "c"})
tau_B = frozenset(
    frozenset(s) for r in range(4)
    for s in itertools.combinations(X_B, r)
)
sp_B = TopologicalSpace(carrier=X_B, topology=tau_B)

print("Uzay A:")
for prop in ["connected", "totally_disconnected"]:
    r = analyze_connectedness(sp_A, prop)
    icon = "✓" if r.is_true else ("✗" if r.is_false else "?")
    print(f"  {prop:22} {icon}  {r.justification[0][:60]}")

print()
print("Uzay B:")
for prop in ["connected", "totally_disconnected", "scattered"]:
    r = analyze_connectedness(sp_B, prop)
    icon = "✓" if r.is_true else ("✗" if r.is_false else "?")
    print(f"  {prop:22} {icon}  {r.justification[0][:60]}")
```

### Alıştırma 3 — Karşılaştırma: Homeomorfizm İnvaryantı Olarak Bağlılık

$\mathbb{R}$ ve $\mathbb{R}^2$'nin homeomorf olmadığını bağlılık argümanıyla gösterin:

1. $\mathbb{R} \setminus \{p\}$ bağlı mı? (`tags=["disconnected"]`)
2. $\mathbb{R}^2 \setminus \{p\}$ bağlı mı? (`tags=["path_connected", "connected"]`)
3. Bu fark neden homeomorfizmi engeller?

Kodu çalıştırın ve sonuçları yorumlayın.

```python
R_minus_pt = TopologicalSpace.symbolic(
    description="R \ {0} = (-inf,0) ∪ (0,+inf)",
    tags=["disconnected", "not_connected"],
)
R2_minus_pt = TopologicalSpace.symbolic(
    description="R^2 \ {(0,0)}",
    tags=["path_connected", "arc_connected", "connected"],
)

for label, sp in [("R \ {p}", R_minus_pt), ("R^2 \ {p}", R2_minus_pt)]:
    r = is_connected(sp)
    icon = "✓" if r.is_true else "✗"
    print(f"  {label:12} bağlı={icon}  {r.justification[0][:65]}")

print()
print("Yorum: R \ {p} iki parçaya ayrılır ama R^2 \ {p} bağlı kalır.")
print("Homeomorfizma bağlantı bileşeni sayısını korur → R ≇ R^2.")
```

### Alıştırma 4 — Bağlılık Hiyerarşisi Doğrulaması

`path_connected ⇒ connected` ve `arc_connected ⇒ path_connected` içermelerini sembolik uzaylarla test edin:
1. `path_connected` taglı 5 uzay oluşturun; hepsinde `is_connected` doğru mu?
2. `connected` + `not_path_connected` taglı 5 uzay oluşturun; hiçbirinde `is_path_connected` doğru olmamalı
3. Her iki grubu connected/path_connected/arc_connected sütunlarıyla tablolayın ve ihlal sayısını yazdırın

```python
path_connected_spaces = [
    TopologicalSpace.symbolic(description=f"PC-{i+1}", tags=["path_connected", "arc_connected"])
    for i in range(5)
]
connected_only_spaces = [
    TopologicalSpace.symbolic(description=f"C-{i+1}", tags=["connected", "not_path_connected"])
    for i in range(5)
]

props = ["connected", "path_connected", "arc_connected"]

print("=== Yol-Bağlı Uzaylar (path_connected ⇒ connected) ===")
violations = 0
print(f"{'Ad':8}" + "".join(f"  {p[:4]:6}" for p in props))
print("-" * 32)
for i, sp in enumerate(path_connected_spaces):
    row = f"PC-{i+1}   "
    res = {p: analyze_connectedness(sp, p) for p in props}
    row += "  ".join(f"{'✓' if r.is_true else '✗':6}" for r in res.values())
    if not res["connected"].is_true:
        violations += 1
        row += "  ** İHLAL **"
    print("  " + row)
print(f"\nyol-bağlı ⇒ bağlı ihlali: {violations}")

print()
print("=== Yalnızca Bağlı Uzaylar (connected, not path_connected) ===")
violations2 = 0
print(f"{'Ad':8}" + "".join(f"  {p[:4]:6}" for p in props))
print("-" * 32)
for i, sp in enumerate(connected_only_spaces):
    row = f"C-{i+1}    "
    res = {p: analyze_connectedness(sp, p) for p in props}
    row += "  ".join(f"{'✓' if r.is_true else '✗':6}" for r in res.values())
    if res["path_connected"].is_true:
        violations2 += 1
        row += "  ** İHLAL **"
    print("  " + row)
print(f"\nnot-yol-bağlı ama yol-bağlı çıkan ihlal: {violations2}")
```

| Hiyerarşi | Yay $\Rightarrow$ Yol $\Rightarrow$ Bağlı; hiçbiri tersine çevrilemez |

| `analyze_connectedness(sp, prop)` | Tek özellik için `Result` döner |
